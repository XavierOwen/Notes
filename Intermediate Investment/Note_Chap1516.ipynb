# Options Markets

**Derivatives**: securities whose prices are determined by, or "derived from", the prices of other securities, including options, futures, and swaps contracts, etc.

**Call Option**: The right to buy an asset at a specified exercise price ($X$) on or before a specified expiration date ($T$).

**Put Option** The right to sell an asset at a specified exercise price ($X$) on or before a specified expiration date ($T$).

- The buyers holding call options wishes stock price ↑
- The buyers holding put. options wishes stock price ↓

There are **3** situations at exercising time. 
1. In the money: Exercising the option is profitable
2. Out of money: Exercising the option is NOT profitable
3. At the money: The exercise price and asset price are equal.

## Option Contract

- American Option: Can be exercised **on or before** its expiration date.
- European Option: Can be exercised **only on** its expiration date.

## Values
### Call
Here $X$ is the expiration price and $S_T$ is the stock price at expiration. Then the payoff to **call options *holder*** is

$$
\max{\left( 0,S_T-X \right)} = \begin{cases}
S_t - X & \text{ if } S_T > X \\ 
0 & \text{ if } S_T \leq X
\end{cases}
$$

![](./Raw/callOptionHolderProfit.png)

And the payoff to **call options *writer*** is 

$$
\min{\left( 0,S_T-X \right)} = \begin{cases}
-\left( S_t - X \right) & \text{ if } S_T > X \\ 
0 & \text{ if } S_T \leq X
\end{cases}
$$

![](./Raw/callOptionWriterProfit.png)

So actually the profit plot is always symmetric along $x\text{-axis}$ between the holder and the writer of an options

### Put
Payoff to put option holder

$$
\max{\left( 0,X-S_T \right)} = \begin{cases}
0 & \text{ if } S_T \geq X \\ 
-\left( S_t - X \right) & \text{ if } S_T < X
\end{cases}
$$

Payoff to put option writer

$$
\min{\left( 0,X-S_T \right)} = \begin{cases}
0 & \text{ if } S_T \geq X \\ 
S_t - X & \text{ if } S_T < X
\end{cases}
$$

## Complex Strategy
### Protective put
buy a stock, then buy a put option. The payoff would be 

$$S_T - S_0 + \max{\left( 0,X-S_T \right)} = \max{\left( S_T - S_0,X-S_0 \right)}$$

![](./Raw/protectivePut.png)

We make a limit downside risk, and so that we will have to pay a premium $P$ which will lower our final profit $P$ less than our actual payoff.

### Covered Call
write a call option, then buy a stock. The payoff would be

$$\min{\left( 0,X-S_T \right)} + S_T - S_0 = \min{\left( S_T - S_0,X-S_0 \right)}$$

![](./Raw/coveredCall.png)

But still you can also earn the premium who buy the call you write, so that the investment will not just to put money in the stock using $S_0$ but get a paid call premium $C$ so that in all, we invest $S_0-C$.

### Straddle
Buy a call option, then buy a put option with same expiration date $T$ and exercise price $X$. The payoff is

$$\max{\left( 0,X-S_T \right)} + \max{\left( 0,S_T - X \right)} = \max{\left( S_T - X,X-S_T \right)}$$

![](./Raw/Straddle.png)

And since we buy two options, we need to pay a call premium $C$ and a put premium $P$.

Or inversely, we sell a call option then sell a put option with same date and price.

### Spread

- Money spread: buy one option and sell another with a different **exercise price**.
- Time spread: sell and purchase of options with differing **expiration date**.

# Option Valuation

**Intrinsic Value**: Stock price minus exercise price, or the profit that could be attained by immediate exercise of an in-the-money call option, $S_0 - X$.

**Time Value**: Difference between option’s price and its intrinsic value.

![](./Raw/callOptionValue.png)

**Determinants**: The price of a call option || put optionwill go up $\Uparrow$, when
1. Stock price, $S\;\;\;\;\;\;\;\;\;\;\;\;\;\;\:\:\:\:\uparrow \;\;\; \| \;\;\; \downarrow$
2. Exercise price, $X\;\;\;\;\;\;\;\;\;\;\:\downarrow \;\;\; \| \;\;\; \uparrow$
3. Volatility, $\sigma\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\:\,\,\uparrow \;\;\; \| \;\;\; \uparrow$
4. Time to expiration, $T\;\;\;\uparrow \;\;\; \| \;\;\; \uparrow$
5. Interest rate, $r_f\;\;\;\;\;\;\;\;\;\;\;\;\,\uparrow \;\;\; \| \;\;\; \downarrow$
6. Dividend payouts, $d\;\;\;\;\;\;\:\,\downarrow \;\;\; \| \;\;\; \downarrow$

## Binomial Option Pricing
![](./Raw/binomialOption Pricing.png)

>**e.g.** using the hedge ratio, $H = \displaystyle\frac{C_u - C_d} {uS_0 - dS_0}$, the change in the price
of an option for a one dollar increase in the stock price, also called Delta: $\Delta$
>
>where $C_u$ and $C_d$ refers to the call option’s value when the stock goes up or down, respectively, and $uS_0$ and $dS_0$ are the stock prices in the two states. Call options always have positive Delta, while put options' are negative.
>1. Given the possible end-of-year stock prices, $uS_0 = \$120$ and $dS_0=\$90$, and the exercise price of $\$110,$ calculate that $C_u = \$10$ and $C_d = \$0$. The stock price range is $\$30$, while the option price range is $\$10$.
2. Hedge ratio: $\$10/\$30 = 1/3$
3. With that, we can say there's a certainty that a portfolio made up of $1/3$ share with one written option would have an end-of-year value of $\$30$
4. Discount the profit: the present value of $\$30$ with a one-year interest rate of $10\%$ is $\$27.27$.
5. So that we have the equation: $1/3 S_0 - C_0 = \$27.27 \Longrightarrow C_0 = \$6.06$

**option elasticity**:  The percentage change in option price per percent change in stock price.

**Dynamic hedging**: constant updating of hedge positions as market conditions change.

## BS Model
- $C_0$: current option value
- $S_0$: current stock price
- $N(d)$: cumulative normal distribution
- $X$: Exercise price
- $\delta$: annual dividend yield of the underlying stock
- $r$: risk free interest rate
- $T$: time remaining till expiration
- $\sigma$: stock price volatility

$$C_0 = S_0 \cdot e^{-\delta t} \cdot N(d_1) \cdot - X\cdot e^{-r\cdot t} N(d_2)\\
d_1 =\frac{\ln{(S_0/X)} + (r - \delta +\sigma^2 / 2)/T} {\delta \sqrt{T}}\\
d_2 = d_1 - \delta \sqrt{T}$$

And we use the parity relationship between the call option and put option to find the price for the put option

$$\begin{array}{ccccccc}
C &+& X \cdot e^{-r\cdot t} &=& P &+& S_0 \\
\text{Call} & & \text{Cash} &  & \text{Put} & & \text{Underlying Asset}
\end{array}$$

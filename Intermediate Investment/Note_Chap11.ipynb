**Duration**: a measure of the effective maturity of a bond, defined as the **weighted average** of the times until each payment, with **weights proportional to the present value** of the payments.

$$D = \sum_{t=1}^{T}w_t \times t \\
w_t = \frac{\frac{\text{Cash Flow}} {(1+y)^t}} {\text{Bond Price}} \\
D_{Modified} = D^*= \frac{D} {1+\text{YTM}}$$

>**e.g.** $8\%$ coupon, annully payment, $\text{YTM} = 10\%$
>
>$$\begin{align}
\text{Bond Price} &= \frac{\text{Cash Flow}_1} {(1+y)} + \frac{\text{Cash Flow}_2} {(1+y)^2} + \frac{\text{Cash Flow}_3} {(1+y)^3} \\
&= \frac{80} {(1+0.1)} + \frac{80} {(1+0.1)^2} + \frac{1080} {(1+0.1)^3} \\
&= 950.26
\end{align}$$

>$$\begin{align}
\text{Duration} &= D \\
&= \frac{1} {\text{Bond Price}} \left( 1\text{ year } \times \frac{\text{Cash Flow}_1} {(1+y)} + 2\text{ year } \times \frac{\text{Cash Flow}_2} {(1+y)^2} + 3\text{ year } \times \frac{\text{Cash Flow}_3} {(1+y)^3} \right)\\
&= \frac{1} {950.26} \left( 1\times \frac{80} {(1+0.1)} + 2 \times \frac{80} {(1+0.1)^2} + 3 \times \frac{1080} {(1+0.1)^3} \right)\\
&= 2.7774\text{ year }
\end{align}$$

## Passive Bond Management
### Immunization

**Immunization**: A duration-matching strategy to shield net worth from interest rate movement.

**Interest rate risk**: mismatch between the maturities of assets and liabilities;
1. Liabilities: customers’ deposits, short-term, short duration.
2. Assets: commercial loans, long-term, long duration.

So that when interest rate goes up, your liability goes down and your asset goes up. However if it goes down, their net worth goes down, the assets falls more in the value than their liabilities.

>**e.g.** Interest rate initially $8\%$
>
>- Liability: $5$ years guaranteed investment contract, just like the zero-coupon bond, $D=5$; and promise to pay $10,000 \times (1.08)^5 = 14,693.28$ after that
- Asset: $6\text{-year}$ coupon bond, $8\%$ coupon rate and $\text{YTM} = 8\%$. After some calculation, $D=4.99$.
>
>So now no matter how rate changes, the accumulated value of invested payment actually doesn't change that much.
>
>- If the interest rate keeps at a $8\%$ level, the accumulated investment payment is
>
>$$ {800} \times {(1.08)^4} + {800} \times {(1.08)^3} + {800} \times {(1.08)^2} + {800} \times {(1.08)^1} + {800} \times {(1.08)^0} + \frac{10800} {(1.08)^1} = 14693.28$$
>
>- And if the interest rate falls to $7\%$, the accumulated investment payment is
>
>$$ {800} \times {(1.07)^4} + {800} \times {(1.07)^3} + {800} \times {(1.07)^2} + {800} \times {(1.07)^1} + {800} \times {(1.07)^0} + \frac{10800} {(1.07)^1} = 14694.05$$
>
>- And if the interest rate increase to $9\%$, the accumulated investment payment is
>
>$$ {800} \times {(1.09)^4} + {800} \times {(1.09)^3} + {800} \times {(1.09)^2} + {800} \times {(1.09)^1} + {800} \times {(1.09)^0} + \frac{10800} {(1.09)^1} = 14696.02$$

How to make that portfolio?
**Rebalancing**: Realigning the proportions of assets in a portfolio to rematch the durations of assets and liabilities.

>An insurance company must pay $\$19,487$ in $7$ years. The market interest rate is $10\%$. So the PV of the obligation is $\$10,000$. The manager wishes to fund the obligation using $3\text{-year}$ zero-coupon bonds and perpetuities paying annual coupons.
>
>$$D_{perpetuity} = \frac{1+y} {y} = 11 \\[1em]
D_{portfolio} = w \times 3 + (1-w) \times 11 = D_{liability} = 7 \Rightarrow w = 0.5$$
>
>The manager purchases $\$5,000$ PV of zero-coupon bonds and $\$5,000$ PV of perpetuity. Note that the face value of the zeros will be $\$5,000 \times 1.1^3 = \$6,655$; the perpetuity pays $\$5,000 \times 10\% = \$500$ annually.

### Cash Flow Matching and Dedication
**cash flow matching**: Matching cash flows from a fixed-income portfolio with those of an obligation.

**dedication strategy**: Refers to multiperiod cash flow matching.

The advantage of dedication over immunization is no need for rebalancing.

However it is not widely persued, because it's constraint on bonds while perpetuity is not available.

## Convexity

Duration only apply to small change, so when there's a big change in interest rate, **convexity** is required to be taken into account, which is the curvature of the price-yield relationship of a bond.

$$\frac{\Delta P} {P} = -D^* \times \Delta y + \frac{1} {2} \times Convex \times \left( \Delta y \right)^2 \\[1em]
D^* = \frac{1} {1+y} \sum_{t=1}^{T} w_t \times t \\[1em]
Convex = \frac{1} {\left( 1+y \right)^2} \sum_{t=1}^{T} w_t \times t(t+1) \\[1em]
w_t = \frac{CF_t} {(1+y)^t} \Big/ \text{Price}
$$

>**e.g.** $T=3$, Annual coupon rate: $10\%$, YTM: $10\%$, Par value: $100$.
>
>$$D^* = \frac{1} {1.1} \times \frac{\ffrac{10} {1.1} \times 1 + \ffrac{10} {1.1^2} \times 2 + \ffrac{110} {1.1^3} \times 3 }  {\ffrac{10} {1.1} + \ffrac{10} {1.1^2} + \ffrac{110} {1.1^3}} = 2.487$$
>
>$$Convex = \frac{1} {1.1^2} \times\frac{\ffrac{10} {1.1} \times 1 \times 2 + \ffrac{10} {1.1^2} \times 2 \times 3 + \ffrac{110} {11.1^3} \times 3 \times 4 } {\ffrac{10} {1.1} + \ffrac{10} {1.1^2} + \ffrac{110} {1.1^3}} = 8.756$$
>
>> $P_0 = \ffrac{10} {1.1} + \ffrac{10} {1.1^2} + \ffrac{110} {1.1^3} = 100$ 
>>
> small changes in YTM: $10\% \to 9.9\%$
>>
>> $P_1 = \ffrac{10} {1.099} + \ffrac{10} {1.099^2} + \ffrac{110} {1.099^3} = 100.2491$
>>
>> Real price change: $\ffrac{P_1 - P_0} {P_0} = 0.2491\%$
>>
>> Duration Rule: $\ffrac{\Delta P} {P} = -D^* \times \Delta y = -2.487 \times (-0.001) = 0.2487\%$

>> With Convexity: $\ffrac{\Delta P} {P} = -D^* \times \Delta y + \frac{1} {2} \times Convex \times \left( \Delta y \right)^2 = -2.487 \times (-0.001) + \ffrac{1} {2} \times 8.756 \times (-0.001)^2 = 0.2491\% $.
>>***
big changes in YTM: $10\% \to 5\%$
>>
>> $P_1 = \ffrac{10} {1.05} + \ffrac{10} {1.05^2} + \ffrac{110} {1.05^3} = 113.62$
>>
>> Real price change: $\ffrac{P_1 - P_0} {P_0} = 13.62\%$
>>
>> Duration Rule: $\ffrac{\Delta P} {P} = -D^* \times \Delta y = -2.487 \times (-0.05) = 12.44\%$

>> With Convexity: $\ffrac{\Delta P} {P} = -D^* \times \Delta y + \frac{1} {2} \times Convex \times \left( \Delta y \right)^2 = -2.487 \times (-0.05) + \ffrac{1} {2} \times 8.756 \times (-0.05)^2 = 13.53\% $.

## Active Bond Management
### Sources of Potential Profit
1. interest rate forecasting
2. identification of relative mispricing within the fixed-income market

**substitution swap**: Exchange of one bond for a bond with similar attributes but more attractively priced.

**intermarket spread swap**: Switching from one segment of the bond market to another.

**rate anticipation swap**: A switch made in response to forecasts of interest rate changes.

**pure yield pickup swap**: Moving to higher yield bonds, usually with longer maturities.

**tax swap**: Swapping two similar bonds to receive a tax benefit.


## SUMMARY
- Even default-free bonds such as Treasury issues are subject to interest rate risk. Longer-term bonds generally are more sensitive to interest rate shifts than short-term bonds. A measure of the average life of a bond is Macaulay’s duration, defined as the weighted average of the times until each payment made by the security, with weights proportional to the present value of the payment.
- Modified duration is a direct measure of the sensitivity of a bond’s price to a change in its yield. The proportional change in a bond’s price approximately equals the negative of modified duration times the change in the bond’s yield.
- Immunization strategies are characteristic of passive bond portfolio management. Such strategies attempt to render the individual or firm immune from movements in interest rates. This may take the form of immunizing net worth or, instead, immunizing the future accumulated value of a bond portfolio.
- Immunization of a fully funded plan is accomplished by matching the durations of assets and liabilities. To maintain an immunized position as time passes and interest rates change, the portfolio must be periodically rebalanced.
- Convexity refers to the curvature of a bond’s price-yield relationship. Accounting for convexity can substantially improve on the accuracy of the duration approximation for the response of bond prices to changes in yields.
- A more direct form of immunization is dedication or cash flow matching. If a portfolio is perfectly matched in cash flow with projected liabilities, rebalancing will be unnecessary.
- Active bond management can be decomposed into interest rate forecasting techniques and intermarket spread analysis. One popular taxonomy classifies active strategies as substitution swaps, intermarket spread swaps, rate anticipation swaps, and pure yield pickup swaps.

## Key Equations
- Macaulay's Duration
$$D = \sum_{t=1}^{T} t \times w_t$$

- Modified duration and duration rule
$$\frac{\Delta P} {P} = -D \times\left[ \frac{\Delta(1+y)} {(1+y)}\right] = -D^* \times \Delta y$$

- Duration of perpetuity
$$D_{perpetuity} = \frac{1+y} {y}$$

- Convexity
$$\frac{\Delta P} {P} = -D^* \times \Delta y + \frac{1} {2} \times Convex \times \left( \Delta y \right)^2$$

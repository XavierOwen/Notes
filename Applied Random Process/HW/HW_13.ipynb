## Question 2

$$\begin{align}
P_{ij}\P{t+s} &= P\CB{X\P{t+s} = j \mid X\P0 = i}\\
&= \sum_{k=0}^\infty P\CB{X\P{t+s} = j,X\P t = k \mid X\P0 = i}\\
&= \sum_{k=0}^\infty P\CB{X\P{t+s} = j \mid X\P0 = i,X\P t = k}\cdot P\CB{X\P t = k \mid X\P0 = i}\\
&= \sum_{k=0}^\infty P\CB{X\P{t+s} = j \mid X\P t = k}\cdot P\CB{X\P t = k \mid X\P0 = i}\\
&= \sum_{k=0}^\infty P_{kj}\P s \cdot P_{ik}\P t
\end{align}$$

$$\begin{align}
P_{ij}'\P t &= \lim_{h \to 0} \ffrac{P_{ij}\P{t+h} - P_{ij}\P t}{h}\\
&= \lim_{h \to 0} \ffrac{ \P{\sum\limits_{k\neq i} P_{ik}\P h \cdot P_{kj} \P t} - \P{1-P_{ii}\P h} P_{ij}\P t}{h}\\
&= \lim_{h \to 0}\P{ \sum_{k\neq i} \ffrac{P_{ik}\P h}{h}P_{kj} \P t - \P{\ffrac{1-P_{ii}\P h}{h}}P_{ij}\P t }\\
&= \sum_{k\neq i}\P{ \lim_{h \to 0} \ffrac{P_{ik}\P h}{h}}P_{kj} \P t - \lim_{h \to 0}\P{\ffrac{1-P_{ii}\P h}{h}}P_{ij}\P t \\
&= \sum_{k\neq i} q_{ik} P_{kj}\P t - v_i P_{ij}\P t
\end{align}$$

## Question 3

$\bspace$The four stats are $\P{1,1}$, $\P{1,0}$, $\P{0,1}$, and $\P{0,0}$, where $1$ indicates the machine at that position is working and $0$ otherwise. It is a continuous Markov Chain and we denote the four states $A$, $B$, $C$ and $D$ respectively for notation convenience.

$$v_A = \lambda_1 + \lambda_2,v_B = \lambda_1 + \mu_2,v_C = \mu_1 + \lambda_2,v_D = \mu_1 + \mu_2$$

$\bspace$And thus for the first machine, its transition probabilities are

$$
\begin{align}
P_{11}\P t &= \ffrac{\lambda_1 e^{-\P{\mu_1 + \lambda_1}t} + \mu_1}{\lambda_1 + \mu_1}\\
P_{10}\P t &= \ffrac{\lambda_1 - \lambda_1 e^{-\P{\mu_1 + \lambda_1}t}}{\lambda_1 + \mu_1} \\
P_{01}\P t &= \ffrac{\mu_1 - \mu_1 e^{-\P{\mu_1 + \lambda_1}t}}{\lambda_1 + \mu_1} \\
P_{00}\P t &= \ffrac{\mu_1 e^{-\P{\mu_1 + \lambda_1}t} + \lambda_1}{\lambda_1 + \mu_1}\\
\end{align}
$$

$\bspace$Then for two machines, its transition probabilities are

$$
\begin{align}
P_{AA}\P t &= P_{11}\P t P_{11}^{\P2}\P t = \ffrac{\lambda_1 e^{-\P{\mu_1 + \lambda_1}t} + \mu_1}{\lambda_1 + \mu_1}\cdot \ffrac{\lambda_2 e^{-\P{\mu_2 + \lambda_2}t} + \mu_2}{\lambda_2 + \mu_2}\\
P_{AB}\P t &= P_{11}\P t P_{10}^{\P2}\P t = \ffrac{\lambda_1 e^{-\P{\mu_1 + \lambda_1}t} + \mu_1}{\lambda_1 + \mu_1} \cdot\ffrac{\lambda_2 - \lambda_2 e^{-\P{\mu_2 + \lambda_2}t}}{\lambda_2 + \mu_2} \\
&\vdots\\
P_{DD}\P t &= P_{00}\P t P_{00}^{\P2}\P t = \ffrac{\mu_1 e^{-\P{\mu_1 + \lambda_1}t} + \lambda_1}{\lambda_1 + \mu_1}\cdot\ffrac{\mu_2 e^{-\P{\mu_2 + \lambda_2}t} + \lambda_2}{\lambda_2 + \mu_2}\\
\end{align}
$$

$\bspace$Now we check the backward and forward Kolmogorov's Equations for $P_{AA}$.

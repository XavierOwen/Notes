## Question 2

$\P{1}$

$$\begin{align}
P\CB{\tau \leq 2} &= P\CB{\tau = 1} + P\CB{\tau = 2 \mid \tau \neq 1} \\
&= P_0 + P\CB{X_1 = j,X_2 = 0,j\geq1}\\
&= p + \sum_{j=1}^{\infty} p\P{1-p}^j \cdot p^j\\
&= p\P{\sum_{j=0}^{\infty} \P{p\P{1-p}}^j} \\
&= \ffrac{p}{1-p+p^2}
\end{align}$$

$$\begin{align}
P\CB{\tau = 2} &= P\CB{X_1 = j,X_2 = 0,j\geq1} \\
&= P\CB{\tau \leq 2} - p\\
&= \ffrac{p^2 - p^3}{1-p+p^2}
\end{align}$$

$\P{2}$

$\bspace$First we need to find the $\mu$

$$\begin{align}
\mu &= \sum_{j=0}^{\infty} j \cdot p\P{1-p}^j \\
&= p\P{\sum_{j=1}^{\infty} j\cdot\P{1-p}^j }\\
\P{1-p}\mu &= p\P{\sum_{j=2}^{\infty} \P{j-1}\cdot\P{1-p}^j} 
\end{align}$$

$$\begin{align}
\mu &= \ffrac{1}{p}\P{\mu - \P{1-p}\mu}\\
&= \ffrac{1}{p}\cdot p\cdot\P{ \sum_{j=1}^{\infty}\P{1-p}^j }\\
&= \ffrac{1-p} {p}
\end{align}$$


$\bspace$So that when $p\geq0.5$, $\mu \leq 1$ and consequently $\pi_0 = 1$. Then when $0<p <0.5$, we have the equation for $\pi_0$

$$\begin{align}\pi_0 &= p + p\P{1-p}\pi_0 + p\P{1-p}^2\pi_0^2 + \cdots + p\P{1-p}^n\pi_0^n+\cdots\\
&= p\P{\sum_{j=0}^{\infty}\P{\P{1-p}\pi_0}^j}\\
&\bspace \text{given that }\P{1-p}\pi_0<1,\\
\pi_0 &=  \ffrac{p} {1-\P{1-p}\pi_0}\\
\Rightarrow \pi_0 &= \ffrac{p} {1-p} < 1
\end{align}$$

And $\pi_0 = \ffrac{p} {1-p}$ is not contradictory to the assumption that $\P{1-p}\pi_0<1$. So that's the answer when $0<p < 0.5$.

## Question 3

$\bspace$We can write the distribution of $P_j$ first

$$\begin{array}{c|c|c} \hline
P_0 & P_1 & P_2 \\ \hline
\P{1-p}^2 & 2p\P{1-p} & p^2 \\ \hline
\end{array}$$

$\bspace$Then $\mu = 2p\P{1-p} + 2p^2 = 2p$, greater than $1$ only when $p > 0.5$.

$\P{1.a}$

$\bspace$As discussed, when $0 < p\leq 0.5$, the extinction probability $\pi_0$ will be $1$. However when $p > 0.5$, we have the equation

$$\pi_0 = \P{1-p}^2 + 2p\P{1-p} \pi_0 + p^2\pi_0^2 \Rightarrow \pi_0 = \P{\ffrac{1-p} {p}}^2 <1$$

$\P{1.b}$

$\bspace$We first focus on the second generation, where we have the probability table of the individuals left.

$$\begin{array}{c|c|c|c} \hline
4 & 3 & 2 & 1 \\ \hline
p^6 & 2p^5\P{1-p} & 2p^5\P{1-p}^3 + 6p^4\P{1-p^2} & 4p^2\P{1-p}^2 + 6p^3\P{1-p}^3 \\ \hline
\end{array}$$

$\bspace$Then the probability of first extinction in the third generation can be obtained by

$$\begin{align}
&\P{1-p}^8\cdot p^6 + \P{1-p}^6\cdot 2p^5\P{1-p} + \P{1-p}^4\cdot \P{2p^5\P{1-p}^3 + 6p^4\P{1-p^2}}\\
&\bspace+ \P{1-p}^2\cdot \P{4p^2\P{1-p}^2 + 6p^3\P{1-p}^3}\\
=&\;\P{1-p}^8\cdot p^6 + 4\P{1-p}^7\cdot p^5 + 6\P{1-p}^6\cdot p^4 + 6\P{1-p}^5\cdot p^3 + 4\P{1-p}^4\cdot p^2
\end{align}$$

$\P{2}$

$\bspace$Here when $0<p\leq0.5$ we still have $\pi_0 = 1$. So when $p>0.5$, we can calculate the extinction probability condition on $\lambda$:

$$\begin{align}
&P\CB{\text{extinction}}\\
=&\; \sum_{j = 0}^{\infty} P\CB{\text{extinction}\mid \lambda = j}\cdot P\CB{\lambda = j}\\
=&\; \sum_{j = 0}^{\infty} \P{\ffrac{1-p} {p}}^{2j} e^{-\lambda} \ffrac{\lambda^j} {j!}\\
=&\; e^{-\lambda} \sum_{j = 0}^{\infty} \left.\P{\P{\ffrac{1-p}{p}}^2\lambda}^j\right/ {j!}
\end{align}$$

$\bspace$Since $e^x = 1+x+\ffrac{x^2} {2!} + \cdots+\ffrac{x^n} {n!}+\cdots$, we have the preceding results simplified as

$$P\CB{\text{extinction}} = \exp\CB{\P{\ffrac{1-p}{p}}^2\lambda - \lambda}$$

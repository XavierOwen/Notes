$\P{1}$

$\bspace$Let $N$ denote the number needed. And let $X_i$ denote the trials thus we have

$$\begin{align}
\EE{N} &= \EE{\EE{N \mid X_1}} \\
&= \EE{N \mid X_1 = H} \cdot P\CB{X_1 = H} + \EE{N \mid X_1 = T} \cdot P\CB{X_1 = T}
\end{align}$$

$\bspace$For $\EE{N \mid X_1 = H}$, since we already have a head toss, what we need is one tail toss, and that's nothing but a geometric variable, similar for $\EE{N \mid X_1 = T}$. Therefore,

$$\begin{align}
\EE{N} &= \EE{N \mid X_1 = H} \cdot P\CB{X_1 = H} + \EE{N \mid X_1 = T} \cdot P\CB{X_1 = T} \\
&= \P{1 + \ffrac{1} {1-p}} \cdot p + \P{1 + \ffrac{1} {p}} \cdot \P{1-p} \\
&= 1 + \ffrac{p} {1-p} + \ffrac{1-p} {p} = \ffrac{p} {1-p} + \ffrac{1} {p}
\end{align}$$

## Question 2

$\bspace$First we need to calculate the marginal distribution of $Y$

$$\begin{align}
f_Y\P{y} &= \int_{-\infty}^{\infty} f\P{x,y} \;\dd{x} \\
&= 4ye^{-y} \int_{y}^{\infty} xe^{-x} - y e^{-x} \;\dd{x}\\
&= 4ye^{-y}\P{\big.-\P{1+x}e^{-x}\big|_y^{\infty} - y \P{\big.-e^{-x}\big|_{y}^{\infty}}}\\
&= 4ye^{-2y}, \bspace y \geq 0
\end{align}$$

$\bspace$and $f_Y\P{y} = 0$ elsewhere. Thus we have

$$\begin{align}
\EE{X \mid Y = y} &= \int_{-\infty}^{\infty} x \cdot \ffrac{f\P{x,y}} {f_Y\P{y}} \; \dd{x}\\
&= \int_{y}^{\infty} x \cdot \P{x-y}e^{y-x} \;\dd{x} \\
&= e^{y} \P{\Big. -\P{2 + 2x + x^2}e^{-x} \Big|_{y}^{\infty}-y\P{-\big.\P{1+x}e^{-x}\big|_{y}^{\infty}}}\\
&= 2+y, \bspace y \geq 0
\end{align}$$

## Question 3

$\bspace$Since $X$ and $Y$ are non-negative independent $r.v.$s, $F_Z\P{z} = P\CB{X+Y \leq z} = 0$ and thus $f_Z\P{z} = 0$ for $z < 0$.

$\bspace$Now when $Z \geq 0$, we condition the probability on $Y$ which gives:

$$\begin{align}
F_Z\P{z} &= P\CB{Z \leq z} = P\CB{X + Y \leq z} \\
&= \int_{-\infty}^{\infty} P\CB{X + Y \leq z \mid Y = y} \cdot f_Y\P{y} \;\dd{y} \\
&= \int_{-\infty}^{\infty} P\CB{X \leq z-y}\cdot f_Y\P{y} \;\dd{y} \\
&\using{z \geq y \geq 0} \int_{0}^{z} F_X\P{z-y}\cdot f_Y\P{y} \;\dd{y} \\
\end{align}$$

$\bspace$take the derivative on $z$ we have

$$
\begin{align}
f_Z\P{z} &= F_Z'\P{z}\\
&= \int_{0}^{z} f_X\P{z-y}\cdot f_Y\P{y} \;\dd{y} + F_X\P{z-z}f_Y\P{z}\cdot 1-F_X\P{z-0}f_Y\P{0}\cdot 0 \\
&= \int_{0}^{z} f_X\P{z-y}\cdot f_Y\P{y} \;\dd{y}, \bspace z \geq 0
\end{align}$$

$\bspace$then combine these two parts we reach our final conclusion that

$$f_Z\P{z} = \begin{cases}
\d{\int_{0}^{z} f_X\P{z-y}\cdot f_Y\P{y} \;\dd{y}}, &\text{if }z \geq 0\\[0.6em]
0, &\text{if } z<0
\end{cases}$$

$\bspace$The limiting distribution exists because the Markov Chain here is irreducible, of course, and also aperiodic, from the fact that $P_{00}^{\P{1}} = \P{1-a}>0$.

$$\mathbf{P}^2 = \begin{Vmatrix}
\ffrac{1} {2} & \ffrac{3} {8} & \ffrac{1} {8}\\

$\bspace$So that we have $P\CB{X_2=S\mid X_0=C} = 3/8$; $P\CB{X_2=C\mid X_0=C} = 7/16$; $P\CB{X_2=R\mid X_0=C} = 3/16$.

$\P{2}$

$\bspace$**irreducible**: $S \to C \to R \to C \to S$  
$\bspace$**positive recurrent**: this Markov Chain is also finite  
$\bspace$**aperiodic**: $P_{SS}^{\P{1}}, P_{CC}^{\P{1}}, P_{RR}^{\P{1}}>0$

$\bspace$And thus, **erogodic**.

$\P{3}$

$\bspace$Since this Markov Chain is erogodic, we have the equations for its limiting probabilities $\P{\pi_1,\pi_2,\pi_3}$:

$$\begin{cases}
\pi_1 = 0.5\pi_1 + 0.5\pi_2 \\
\pi_2 = 0.5\pi_1 + 0.25\pi_2 + 0.5\pi_3\\
\pi_3 = 0.25\pi_2 + 0.5\pi_3\\
\pi_1+\pi_2+\pi_3 = 1
\end{cases} \Longrightarrow \begin{cases}
\pi_1 = 0.4\\
\pi_2 = 0.4\\
\pi_3 = 0.2
\end{cases}$$

$\bspace$Thus in the long run, the percentages of the days are sunny, cloudy and rainy are $40\%$, $40\%$ and $20\%$ respectively.

## Question 5

$\P{1}$

$$P\P{i} = \alpha_i\cdot P\P{i+1} + \P{1-\alpha_i}\cdot P\P{i-1}$$

$\P{2}$

$$P\P{i+1}-P\P{i} = \ffrac{1-\alpha_i} {\alpha_i}\P{P\P{i}-P\P{i-1}}\Rightarrow P\P{i} - P\P{i-1} = P\P{1} \cdot \prod_{j=1}^{i-1} \ffrac{1-\alpha_j} {\alpha_j} \\
\Rightarrow P\P{i} = P\P{1}\SB{1+\sum_{k=2}^{i} \P{\prod_{j=1}^{k-1}\ffrac{1-\alpha_j} {\alpha_j}}}$$

$\P{3}$

$\bspace$The required probability is the same with the probability that the number of balls in urn $2$ reaches $N$ before $0$. And for urn $2$, the winning probability is $\alpha_i=\ffrac{i} {N}$ where it gets one more ball. Then 

$$\begin{align}
P\P{i} &= P\P{1}\SB{1+\sum_{k=2}^{i} \P{\prod_{j=1}^{k-1}\ffrac{1-\ffrac{j} {N}} {\ffrac{j} {N}}}}\\
&= P\P{1}\SB{1+\sum_{k=2}^{i} \ffrac{\P{N-1}!} {\P{N-k}!\P{k-1}!}}\\
&= P\P{1}\SB{\sum_{k=1}^{i} \binom{N-1} {k-1}}
\end{align}$$

$\bspace$And consider when $i=N$ then $P\P{N} = p\P{1}\cdot 2^{N-1} = 1$, then we have

$$P\P{i} = 2^{1-N}\cdot\SB{\sum_{k=1}^{i} \binom{N-1} {k-1}}$$

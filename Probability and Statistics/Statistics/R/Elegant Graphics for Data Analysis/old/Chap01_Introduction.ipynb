```r
getdata <- function(id, directory, summarize=FALSE, FileEncoding='UTF-8-BOM') {
    filelist <- dir(directory);
    file <- paste(directory,filelist,sep='/')
    data <- read.csv(file, fileEncoding = FileEncoding)
    if(summarize == "TRUE") {
        mySum <- summary(data)
        print(mySum)
    }
    return(data)
}

getdata("mydata.csv", "assets", TRUE)
```

\newcommand{\SSD}{\text{SSD}}$Considering the **symmetric random walk**, which *in each time unit* is *equally likely* to *take a unit step* either to the *left* or to the *right*. That is, it's a Markov Chain with $P_{i,i+1} = P_{i,i-1} = \ffrac{1}{2}$ for $i=0,\pm 1,\dots$. Now suppose that each $\Delta t$ time unit we take a step of size $\Delta x$ either to the left or the right with equal probabilities. Let $X\P t$ denote the position at time $t$, then

$\bspace X\P t = \Delta x \P{X_1 + X_2 + \cdots + X_{\SB{t/\Delta t}}},\bspace X_i = \begin{cases}
+1,& \text{if the $i$th step of length $\Delta x$ is to the right}\\
-1,& \text{if it is to the left}
\end{cases}$

and $\SB{t/\Delta t}$ is the largest integer less than or equal to $\ffrac{t}{\Delta t}$, and the $X_i$ are assumed to be independent with $P\CB{X_i = 1} = P\CB{X_i = -1} = 0.5$. Then

$\bspace \Exp{X\P t} = \Delta x \cdot \P{0+0+\cdots+0} = 0,\bspace \Var{X\P t} = \P{\Delta x}^2 \P{1+1+\cdots+1} = \P{\Delta x}^2 \SB{\ffrac{t}{\Delta t}}$

Now let $\Delta t$ and $\Delta x$ reach $0$, and in the purpose of obtaining a limit, the path is $\Delta x = \sigma \sqrt{\Delta t}$ and thus as $t\to \infty$,

$\bspace \Exp{X\P t} = 0,\bspace \Var{X\P t} \to \sigma^2 t$

We then derive some of its properties using the **central limit theorem**:

1. $X\P t$ is normal with mean $0$ and variance $\sigma^2 t$. 

2. Since the changes of value of the random walk in nonoverlapping time intervals are independent, $\CB{X\P t, t\geq 0}$ has independent increments, in that for all $t_1 < t_2 < \cdots < t_n\\[0.5em]$
$$X\P{t_n} - X\P{t_{n-1}},X\P{t_{n-1}} - X\P{t_{n-2}},\dots,X\P{t_2} - X\P{t_1},X\P{t_1}$$
are independent.

3. Since the distribution of the change in position of the random walk over any time interval depends only on the length of that interval, it would appear that $\CB{X\P t , t\geq 0}$ has stationary increments, in that the distribution of $X\P{t+s} - X\P t$ does not depend on $t$.

$Def.1$ **Brownian Motion Process**

>A stochastic process $\CB{X\P t , t\geq 0}$ is said to be a ***Brownian motion process*** if 
>
>1. $X\P 0 = 0$
>2. $\CB{X\P t , t\geq 0}$ has *stationary* and *independent* increments
>3. $\forall \: t>0$, $X\P t$ is **normally distributed** with mean $0$ and variance $\sigma^2 t$
>
>And when $\sigma=1$ we call this a ***standard Brownian motion*** denoted as $B\P t = \ffrac{X\P t}{\sigma}$

In addition, $X\P t$ should be a continuous function of $t$. To prove this, we must show that

$\bspace w.p. 1,\bspace \d{\lim_{h \to 0} \P{X\P{t+h} - X\P t} = 0}$

And this seems reasonable since $X\P{t+h} - X\P t$ is a $r.v.$ with mean $0$ and variance $h$. Interestingly, although $X\P t$ will, *with probability* $1$ be a continuous function of $t$, it possesses the property of being ***nowhere differentiable***, since $\ffrac{X\P{t+h} - X\P t}{h}$ has mean $0$ and variance $1/ h$ which goes to $\infty$ as $h$ goes to $0$.

$Remark$

>The path of Brownian motion is continous (*integrable*) while NOT *derivable*, in mathematical analysis sense.
***

We now derive the joint density function of $X\P{t_1},X\P{t_2},\dots,X\P{t_n}$ for $t_1<t_2<\cdots<t_n$. The density function of $X\P t$, a normal $r.v.$ with mean $0$ and variance $t$, is

$\bspace f_t\P x = \ffrac{1}{\sqrt{2\pi t}} e^{-x^2/2t}$

Then note that 

$\bspace \begin{align}
X\P{t_1}&=x_1\\
X\P{t_2}&=x_2\\
&\;\vdots\\
X\P{t_n}&=x_n
\end{align}$$\;\iff$$\begin{align}
X\P{t_1}&=x_1\\
X\P{t_2}-X\P{t_1}&=x_2-x_1\\
&\;\vdots\\
X\P{t_n} - X\P{t_{n-1}}&=x_n-x_{n-1}
\end{align}$

By the independent increment assumption, $X\P{t_1},X\P{t_2}-X\P{t_1},\dots,X\P{t_n} - X\P{t_{n-1}}$ are independent. And by the stationary increment assumption, $X\P{t_k} - X\P{t_{k-1}}$ is **normal** with mean $0$ and variance $t_k - t_{k-1}$. Hence, the joint density of $X\P{t_1},X\P{t_2},\dots,X\P{t_n}$ is given by

$\bspace\begin{align}
f\P{x_1,x_2,\dots,x_n} &= f_{t_{\void_1}}\P{x_1} \cdot f_{t_{\void_2} - t_{\void_1}}\P{x_2-x_1}\cdots f_{t_n - t_{n-1}}\P{x_{n}-x_{n-1}}\\
&= \ffrac{\exp\CB{-\ffrac{1}{2} \SB{\ffrac{x_1^2}{t_1} + \ffrac{\P{x_2-x_1}^2}{t_2-t_1} + \cdots + \ffrac{\P{x_n - x_{n-1}}^2}{t_n-t_{n-1}}}}}{\P{2\pi}^{n/2} \sqrt{t_1\P{t_2-t_1}\cdots\P{t_n-t_{n-1}}}}
\end{align}$

And with this, in principle, we can compute any desired probabilities. For instance, the conditional distribution of $X\P s$ given that $X\P t =s$ where $s<t$:

$\bspace\begin{align}
f_{s\mid t}\P{x\mid B}&= \ffrac{f_s\P x f_{t-s}\P{B-x}}{f_t\P B}\\
&= \ffrac{\exp\CB{-\ffrac{x^2}{2s} - \ffrac{\P{B-x}^2}{2\P{t-s}}}}{2\pi\sqrt{s\P{t-s}} \cdot f_t\P B}\\
&= \ffrac{\exp\CB{-\ffrac{B^2}{2\P{t-s}}}}{2\pi\sqrt{s\P{t-s}} \cdot f_t\P B}\exp\CB{-x^2\P{\ffrac{1}{2s}+\ffrac{1}{2\P{t-s}}} +\ffrac{Bx}{t-s}}\\
&= \ffrac{\exp\CB{-\ffrac{B^2}{2\P{t-s}}}}{2\pi\sqrt{s\P{t-s}} \cdot f_t\P B}\exp\CB{-\ffrac{t}{2s\P{t-s}}\P{x^2 - 2\ffrac{sB}{t}x + \ffrac{s^2B^2}{t^2} }} \exp\CB{\ffrac{t}{2s\P{t-s}}\ffrac{s^2B^2}{t^2}}\\
&= \ffrac{\exp\CB{-\ffrac{B^2}{2t}}}{2\pi\sqrt{s\P{t-s}} \cdot f_t\P B}\exp\CB{-\ffrac{t}{2s\P{t-s}}\P{x-\ffrac{sB}{t}}^2}
\end{align}$

Thus, $\Exp{X\P s \mid X\P t = B} = \ffrac{sB}{t}$ and $\Var{X\P s \mid X\P t = B} = \ffrac{s}{t}\P{t-s}$

## Hitting Times, Maximum Variable, and the Gamblerâ€™s Ruin Problem
Let $T_a$ denote the *first time the Brownian Motion process hits* $a$, named, ***hitting time***. When $a>0$, we compute $P\CB{T_a \leq t}$ by conditioning on whether or not $T_a \leq t$, which leads to

$\bspace \begin{align}
P\CB{X\P t\geq a} &= P\CB{X\P t\geq a \mid T_a \leq t} \cdot P\CB{T_a \leq t} + P\CB{X\P t\geq a \mid T_a > t} \cdot P\CB{T_a > t}\\
&= P\CB{X\P t\geq a \mid T_a \leq t} \cdot P\CB{T_a \leq t}
\end{align}$

By symmetry, we have $P\CB{X\P t \geq a \mid T_a \leq t} = 0.5$ and then, based on this,

$$\begin{align}
P\CB{T_a \geq t} &= 2P\CB{X\P t \geq a}\\
&= \ffrac{2}{\sqrt{2\pi t}}\int_a^\infty \exp\CB{-\ffrac{x^2}{2t}}\;\dd x\\
&= \ffrac{2}{\sqrt{2\pi}} \int_{a/\sqrt t}^\infty \exp\CB{-\ffrac{y^2}{2}}\;\dd y, \bspace a>0
\end{align}$$

if $a<0$, then the distribution of $T_a$ is, by symmetry, the same as that of $T_{-a}$. Hence, 

$\bspace P\CB{T_a \geq t} = \ffrac{2}{\sqrt{2\pi}} \int_{\abs a/\sqrt t}^\infty \exp\CB{-\ffrac{y^2}{2}}\;\dd y$

***
And the maximum value the process attains in $\SB{0,t}$, named the **maximum variable**. We obtain its distribution as follows:

$$\begin{align}
P\CB{\max_{0\leq s\leq t}X\P s\geq a} &= P\CB{T_a \leq t} \bspace\text{by continuity}\\
&= 2P\CB{X\P t \geq a}\\
&= \ffrac{2}{\sqrt{2\pi}} \int_{a/\sqrt t}^\infty \exp\CB{-\ffrac{y^2}{2}}\;\dd y
\end{align}$$

***
Now consider the probability that Brownian motion hits $A$ before $B$, where $A,B > 0$. Since Brownian motion is a limit of symmetric random walk, we start from the Gambler's ruin problem. With $N = \ffrac{A+B}{\Delta x}$ and $i = \ffrac{B}{\Delta x}$, the probability that the symmetric random walk goes up $A$ before going down $B$ is equal to $B\Delta x/\P{A+B}\Delta x = \ffrac{B}{A+B}$. Hence upon letting $\Delta x \to 0$, we see that

$\bspace P\CB{\text{up $A$ before down $B$}} = \ffrac{B}{A+B}$

## Variations on Brownian Motion
### Brownian Motion with Drift
Consider Brownian Motion $\CB{X\P t, t\geq 0}$. It's with ***drift parameter*** $\mu$ and ***variance parameter*** $\sigma^2$ if

1. $X\P 0 = 0$
2. $\CB{X\P t, t\geq 0}$ has stationary and independent increments
3. $X\P t$ is *normally* distributed with mean $\mu t$ and variance $\sigma^2 t$. Not just $\mu$ but $\mu \mathbf t$

Or equivalently, first denote $B\P t$ as the standard Brownian motion, $\CB{B\P t, t\geq 0}$, and then define $X\P t = \sigma B\P t + \mu t$

And clearly, from this representation, $B\P t$ with drift $\mu t$ is still a continous function of $t$.

### Geometric Brownian Motion
The **geomatric Brownian Motion** is defined as follows: suppose $Y\P t$ is a Brownian motion process with **drift coefficient** $\mu$ and **variance coefficient** $\sigma^2$, then $X\P t=e^{Y\P t}$ is the ***geomatric Brownian Motion***.

We now compute the expected value of the process at time $t$ given the history of the process up to time $s$ with $s<t$.

$\bspace\begin{align}
\Exp{X\P t\mid X\P u, 0\leq u\leq s} &= \Exp{e^{Y\P t}\mid Y\P u, 0\leq u\leq s}\\
&= \Exp{e^{Y\P s + Y\P t - Y\P s}\mid Y\P u, 0\leq u\leq s}\\
&= e^{Y\P s}\Exp{e^{Y\P t - Y\P s}\mid Y\P u, 0\leq u\leq s}\\
&= X\P s\Exp{e^{Y\P{t-s}}}\\
&= X\P s \cdot \exp\CB{\Exp{Y\P{t-s}} + \ffrac{1}{2}\Var{Y\P{t-s}}}\\
&= X\P s e^{\P{t-s}\P{\mu + \sigma^2/2}}
\end{align}$

$Remark$

>$\Exp{e^{Y\P t - Y\P s}\mid Y\P u, 0\leq u\leq s} = \Exp{e^{Y\P t - Y\P s}}$ is by independent increments and then $\Exp{e^{Y\P t - Y\P s}} = \Exp{e^{Y\P{t-s}}}$ is by stationary increments. And that's just the *moment generating function of a Normally Distributed* $r.v.$. And since $Y\P{t-s}\sim\N{\mu\P{t-s},\sigma^2\P{t-s}}$, we finish the calculation.

Also, it's reasonable to suppose that $X_n/X_{n-1}$, $n\geq 1$ are $i.i.d.$. Thus letting $Y_n = \dfrac{X_{n}}{X_{n-1}}$ leads to

$\bspace X_n = Y_n Y_{n-1}\cdots Y_{1}Y_0 \Rightarrow \log Y_n = \d{\sum\nolimits_{i=1}^n\log Y_i + \log X_0}$

And $Y_i$ are $i.i.d.$, then so are the $\log Y_i$. So, $\log Y_n$ is a Brownian motion with a drift after suitable normalization. Thus, $\CB{X_n}$ will also be approximately geometric Brownian motion.

## White Noise
Let $\CB{X\P t, t\geq0}$ denote a standard Brownian Motion process and let $f$ be a function having a continuous derivative in the region $\SB{a,b}$. The stochastic integral is defined as

$\bspace\d{\int_a^b f\P t \;\dd X\P t \equiv \underset{\max \P{t_{\void_i} - t_{\void_{i-1}}\:\to\: 0}}{\lim_{n\to\infty} \sum_{i=1}^n} f\P{t_{i-1}}\SB{X\P{t_i} - X\P{t_{i-1}}}}$

where $a=t_0 <t_1<\cdots t_n = b$ is a partition of the region $\SB{a,b}$. Using the identity

$\bspace\d{\sum_{i=1}^n f\P{t_{i-1}} \SB{X\P{t_{i}} - X\P{t_{i-1}}}= f\P b X\P b - f\P a X\P a - \sum_{i=1}^n X\P{t_i} \SB{f\P{t_i} - f\P{t_{i-1}}}}$

we see that

$\bspace\d{\int_a^b f\P t \;\dd X\P t = f\P b X\P b - f\P a X\P a - \int_a^b X\P t\;\dd f\P t}$

And this can be seen as the definition of $\d{\int_a^b f\P t \;\dd X\P t}$. And upon assuming the interchangeability of expecation and limit, we obtain

$\bspace \d{\Exp{\int_a^b f\P t \;\dd X\P t}} = 0$

$\bspace\begin{align}
\Var{\sum_{i=1}^n f\P{t_{i-1}} \SB{X\P{t_i} - X\P{t_{i-1}}}} &= \sum_{i=1}^n f^2\P{t_{i-1}} \Var{{X\P{t_i} - X\P{t_{i-1}}}}\\
&= \sum_{i=1}^n f^2\P{t_{i-1}} P{t_it_{i-1}}\\
\Rightarrow \Var{\int_a^b f\P t \;\dd X\P t} &= \int_a^b f^2\P t \;\dd t
\end{align}$

$Remark$

>The preceding operation, from $f$ to $\d{\int_a^b f\P t \;\dd X\P t}$, is called the ***white noise transformation***, or more loosely, $\CB{\dd X\P t, 0\leq t<\infty}$ is the ***white noise*** since it can be regarded as a time varying function $f$ travels through a white noise medium to yield the output at time $b$.

## Gaussian Processes

$Def.2$ 

>A stochastic process $X\P t$, $t\geq 0$ is called a ***Gaussian***, or ***Normal process***, if $X\P{t_1}$, $\dots$, $X\P{t_n}$ has a **multivariate normal distribution** for all $t_1$, $t_2$, $\dots$, $t_n$.

If $X\P t$ is already a **Brownian motion** Process, then because each of $X\P{t_1}$, $X\P{t_2}$, $\dots$, $X\P{t_n}$ can be expressed as a linear combination of the independent **normal** $r.v.$ $X\P{t_1}$, $X\P{t_2} - X\P{t_1}$, $\dots$, $X\P{t_n} - X\P{t_{n-1}}$ it follows that ***Brownian motion*** *is a* ***Gaussian process***. 


And **multivariate normal distribution** can be uniquelly defined by the marginal mean values and the covariance values. It follows that standard Brownian could also be defined as a Gaussian process having $\Exp{X\P t} = 0$ and, for $s\leq t$:

$\bspace\begin{align}
\Cov{X\P s,X\P t} &= \Cov{X\P s, X\P s + X\P t - X\P s} \\
&= \Cov{X\P s,X\P s} + \Cov{X\P s, X\P t - X\P s}\\
&= \Var{X\P s} = s
\end{align}$

Note that $\Cov{X\P s, X\P t - X\P s} = 0$ because of the independent increments.

Now another special stochastic process: ***Brownian bridge***, $\CB{X\P t, 0\leq t\leq 1 \mid X\P 1 = 0}$, as it tied down both at $0$ and at $1$. Since the conditional distribution of $X\P{t_1}$, $X\P{t_2}$, $\cdots$, $X\P{t_n}$  is multivariate normal it follows that **Brownian bridge** *is a* **Gaussian process**, with $\Exp{X\P s \mid X\P 1 = 0} = 0$, for $s< 1$; and for $s<t<1$, 

$\bspace \begin{align}\Cov{X\P s, X\P t \mid X\P 1 = 0} &= \Exp{X\P s X\P t \mid X\P 1 = 0} - \Exp{X\P s \mid X\P 1 = 0} \cdot \Exp{X\P t \mid X\P 1 = 0}\\
&= \Exp{\Exp{X\P s X\P t \mid X\P t, X\P 1 = 0}\mid X\P 1 = 0} - 0 - 0 \\
&= \Exp{X\P t \Exp{X\P s \mid X\P t}\mid X\P 1 = 0}\\
&= \Exp{X\P t \dfrac{s}{t}X\P t \mid X\P 1 = 0}\\
&= \dfrac{s}{t}\Exp{X^2\P t \mid X\P 1 = 0}\\
&= \dfrac{s}{t} t\P{1-t} = s\P{1-t}
\end{align}$

$Remark$

>In section [10.1](#Brownian-Motion) we've already proved that $\Exp{X\P s \mid X\P t = B} = \ffrac{s}{t}B$ and $\Var{X\P s \mid X\P t = B} = \ffrac{s}{t}\P{t-s}$

So, the **Brownian bridge** can be defined as a **Gaussian process** *with mean value $0$ and covariance function $s\P{1-t}$* for $s \leq t$.

$Proposition.1$

>If $\CB{X\P t,t\geq0}$ is **standard Brownian motion**, then $\CB{Z\P t, 0\leq t\leq 1}$ is a **Brownian bridge** process when $Z\P t = X\P t - tX\P 1$.

$Proof$

>As it is immediate that $\CB{Z\P t, t\geq 0}$ is a **Gaussian process**, all we need verify is that $\Exp{Z\P t} = 0$ and $\Cov{Z\P s,Z\P t} = s\P{1-t}$, when $s\leq t$. The former is obviously true and the latter follows from
>
>$\bspace\begin{align}
\Cov{Z\P s, Z\P t}&= \Cov{X\P s - sX\P 1, X\P t - tX\P 1}\\
&= \Cov{X\P s,X\P t} - t\cdot\Cov{X\P s , X\P 1}\\
&\bspace- s\cdot\Cov{X\P 1,X\P t}+st\cdot\Cov{X\P1,X\P1}\\
&= s - st - st + st = s\P{1-t}
\end{align}$

We also define the ***integrated Brownian motion***, as an illustration of how such a process may arice in practice. If $\CB{X\P t,t\geq0}$ is **Brownian motion**, then the **integrated Brownian motion** $\CB{Z\P t,t\geq0}$ is defined as

$\bspace \d{Z\P t = \int_0^t X\P s \;\dd s}$

thus the rate of change of $Z\P t$ follows a **Brownian motion**, $\dfrac{\dd Z\P t}{\dd t} = X\P t$ and thus

$\bspace \d{Z\P t = Z\P 0 + \int_0^t X\P s \;\dd s}$

and since **Brownian motion** is a **Gaussian motion**, **integrated Brownian motion** *is also a* **Gaussian motion**. To prove this, first recall that $W_1$, $W_2$, $\dots$, $W_n$ is said to have a multivariate normal distribution if they can be represented as

$\bspace \d{W_i = \sum\nolimits_{j=1}^m a_{ij} U_j}$, $i = 1,2,\dots,n$

where $U_j$, $j=1,2,\dots,m$ are independent normal random variables, and thus any set of partial sums of $W_1$, $W_2$, $\dots$, $W_n$ are also jointly normal.

Now $Z\P{t_1}$, $Z\P{t_2}$, $\dots$, $Z\P{t_n}$ is multivariate normal can now be shown by writing the integral form $Z\P t = \d{\int_0^t X\P s\;\dd s}$ as a limit of approximating sums. Thus it's a **Gaussian motion**.

Then, as $\CB{Z\P t, t\geq 0}$ is Gaussian it follows that its distribution is characterized by its mean value and covariance function. We now compute them under the condition that $\CB{X\P t, t\geq 0}$ is standard Brownian motion.

$\bspace\begin{align}
\Exp{Z\P t} &= \Exp{\int_0^t X\P s \; \dd s}\\
&= \int_0^t \Exp{X\P s}\;\dd s\\
&= 0
\end{align}$

and for $s\leq t$

$\bspace\begin{align}
\Cov{Z\P s,Z\P t}&= \Exp{Z\P s Z\P t} - \Exp{Z\P s}\Exp{Z\P t}\\
&= \Exp{\int_0^t X\P y \;\dd y \cdot \int_0^s X\P u \;\dd u} - 0\\
&= \int_0^s \int_0^t \Exp{X\P y X\P u}\; \dd y \dd u\\
&= \int_0^s \int_0^t \min\P{y,u}\; \dd y \dd u\\
&= \int_0^s \P{\int_0^u y\;\dd y+ \int_u^t u\;\dd y}\;\dd u = s^2\P{\dfrac{t}{2}- \dfrac{s}{6}}
\end{align}$

$Remark$

>$\Exp{X\P y X\P u} = \Cov{X\P y,X\P u} = \min\P{y,u}$ for whatever $y$ and $u$, and this shall be remembered.

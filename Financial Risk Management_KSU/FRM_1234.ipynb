## Typology of Risk Exposures
Risks Faced by a Financial Institution 
- Market Risk: changes in financial market prices and rates will reduce the value of the bank's positions.
- Credit Risk: a change in the credit quality of a counterparty will affect the value of a bank's position.
- Liquidity Risk流动资金: both "funding liquidity risk" and "trading-related liquidity risk"
  - Funding liquidity risk: a financial institution's ability to raise the necessary cash to roll over its debt, to meet the cash, margin, and collateral requirements of counterparties, and (in the case of funds) to satisfy capital withdrawals.
  - Trading-related liquidity risk: an institution will not be able to execute a transaction at the prevailing market price because there is, temporarily, no appetite for the deal on the "other side" of the market.
- Operational Risk: potential losses resulting from inadequate systems, management failure, faulty controls, fraud, and human errors.
- Legal Risk: when a counterparty, or an investor, loses money on a transaction and decides to sue the bank to avoid the obligations
- Human Factor Risk: a special form of operational risk. It relates to the losses that may result from human errors

## Historical Evolution
- 1988 Basle Accord established international minimum capital guidelines linking bank’s capital requirements to their credit exposures.
- 1996 Amendment to 1998 BIS Accord extended the Accord to include risk-based capital requirements for market risks.
- Currently banks are required to satisfy three capital adequacy standards:
    - A maximum assets to capital multiple of 20.
    - An 8 percent minimum ratio of eligible capital to risk-weighted assets.
    - A minimum capital charge to compensate for market risk of traded instruments on and off the balance sheet.
- 1996 Amendment officially consecrates the use of internal models based on the value-at-risk (VAR) methodology to assess market risk exposure.
- 1996 Amendment made it clear that banks must implement a risk management infrastructure that is fully integrated with their daily risk management, their settings of trading limits and their risk monitoring of operation
- The amendment requires banks to develop sophisticated analytical approaches to measure risk, and use them in practice to monitor their position and their risk limits.
- BIS 2000+ Accord. 
- Improvement of counterparty risk management practices in the wake of the LTCM debacle.

## Group of Thirty (G-30) Policy Recommendations
In 1993 the Group of Thirty (G-30) published a report that described 20 best-practice price risk management recommendations for dealers and end-users of derivatives, and four recommendations for legislators, regulators, and supervisors.

- Main focus was in managing derivatives.
- Policies governing derivatives should be clearly defined.
- Senior management should approve procedures and controls.
- Dealers should have a market risk management function with clear independence and authority.
- Dealers should use a consistent measure to calculate daily the market risk.
- VaR should be compared to market risk limits.
- Regular stress simulations should be performed.

## The 1988 BIS Accord: The “Accord”
initially developed by the Basle Committee on Banking Supervision, and later endorsed by central bank governors of the G10 countries.

first step in establishing a level playing field across member countries for internationally active banks.

Defines two minimum standards for meeting capital adequacy requirements: An asset-to-capital multiple, and a risk-based capital ratio.
- Asset-to-capital standard is an overall measure of the bank’s capital adequacy.
- Risk-based capital ratio focus on the credit risk associated with specific on- and off- balance sheet asset categories.
- Risk-based capital ratio takes the form of a solvency ratio (known also as the Cooke ratio).

Limitied in scope, because it ignores complex issues -- such as portfolio effects, as well as “netting”.

A well diversified portfolio is much less likely to suffer from massive credit losses

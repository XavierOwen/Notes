# Exotic Options

- ***Plain vanilla options***: standard call and put options.
- ***Exotic options***: options different from the standard call and put options.
    - Exotic options are *less liquid* than standard options, and often have discontinuous payoffs.
- ***Path-dependent options***: payoff depends on what the path of the stock price has been up to maturity (rather than only on the price at maturity).

And following the previous chapter, we assume the stock price following the Merton-Black-Scholes model, and the stock does not pay dividends.

>For the parameters in pricing an option, say, a European call, we'll write $c\P{t.S_t,T,K}$. And an intuitive proposition is that
>
>$$c\P{t,k\cdot S_t, T,k\cdot K} = k\cdot c\P{t,S_t,T,K}$$

## Forward Start Option
To buy a ***Forward start options*** requires the payment for today, but it starts to count at some future time. We'll foucus on one special case: A European call option with maturity $T$ and strike price $S_{t_{\void_1}}$, $t_1 < T$. The current time is zero.

The price of a forward start options is

$$\begin{align}
&\ExpH^\QQ\SB{e^{-r\P{T-0}}\P{S_T - S_{t_{\void_1}}}^+ \mid \FcF_0}\\
=\;& \ExpH^\QQ\SB{e^{-rt_{\void_1}} \ExpH^\QQ \SB{e^{-r\P{T-t_{\void_1}}}\P{S_T - S_{t_{\void_1}}}^+ \mid \FcF_{t_{\void_1}}} \mid \FcF_0}\\
=\;& \ExpH^\QQ_0\SB{c\P{t_1,S_{t_{\void_1}},T,S_{t_{\void_1}}} e^{-rt_{\void_1}}}\\
=\;& \ExpH^\QQ_0\SB{ \ffrac{S_{t_{\void_1}}}{e^{rt_{\void_1}}} c\P{t_1,\color{brown}{1},T,\color{brown}{1}}},\bspace \text{martingale}\\
=\;& S_{t_{\void_0}} \cdot c\P{t_1,1,T,1} = S_0c\P{t_1,1,T,1}\\
=\;& c\P{t_1,S_0,T,S_0}\\
=\;& S_0 \N{d_1'} - S_0 e^{-r\P{T-t_{\void_1}}}\N{d_2'}\\
d'_{1,2} = \;& \ffrac{1}{\sigma\sqrt{T-t_1}} \P{\log \ffrac{S_0}{S_0} + \P{r\pm\ffrac{\sigma^2}{2}}\P{T-t_1}} = \ffrac{\sqrt{T-t_1}}{\sigma} \P{r\pm \ffrac{\sigma^2}{2}}
\end{align}$$

$Remark$

>Must be at the money state. And this price, the price of the forward start at-the-money call option starting at the future date $t_1$, s the same as the price of the regular at-the-money call option with time to maturity $T-t_1$.

## Binary or Digital Options

A binary option (or digital option) pays $1$ if the asset price ends up above a given strike price $K$, and $0$ dollars otherwise. It's payoff is $\begin{cases}
1,&\text{if } S_T>K\\
0,&\ow
\end{cases}$

Now if the stock price follows the BS model, then the price of this binary options is

$$\ExpH^\QQ_t\SB{e^{-r\P{T-t}} \mathbf 1_{\CB{S_T>K}}} = e^{-r\P{T-t}}\N{d_2}$$

where $d_2$ has the same definition with that in BS model.

## Asian Options
The payoff of Asian options depends on the *average* of the asset price over the *lifetime* of the contract. The payoff has two definitions, that is

$$\P{S_{\text{avg}} - K}^+,\bspace S_{\text{avg}} = \begin{cases}
\ffrac{1}{T} \d{\int_0^T S_T \;\dd t},&\text{arithmetic Asian option}\\
\exp\CB{\ffrac{1}{T} \d{\int_0^T \log S_T \;\dd t}},&\text{geometric Asian option}
\end{cases}$$

Asian options can reduce the risk. And it's normally cheaper than European or American options. And about the two different calculation, for the geometric one, since the *geometric average* of a set of *lognormally distributed* $r.v.$ is also *lognormal*, we are able to find the analytical formulas for valuing geometric Asian Options.

## Barrier Options
A barrier option is one that is *activated* or *deactivated* if the asset price crosses a preset barrier. And the payoff is same with the vanillas'.

- **knock-in**s
    - The barrier is ***up-and-in*** if the option is only *activate* if the barrier is hit from below (the rising price)
    - The barrier is ***down-and-in*** if the option is only *activate* if the barrier is hit from above (the falling price)
- **knock-out**s
    - The barrier is ***up-and-out*** if the option will be *deactivated*  (become *worthless*) if the barrier is hit from below (the rising price)
    - The barrier is ***down-and-out*** if the option will be *deactivated* (become *worthless*) if the barrier is hit from above (the falling price)

**e.g.** **down-and-in call** option. 

It has the same payoff with the vanilla call option, but only if the option is *activated*, when *asset price goes below a certain level* $L$ at some time during the life of the contract. Otherwise the payoff would be $0$. Namely,

$$\begin{cases}
\P{S_T - K}^+, &\text{if }\d{\min_{0\leq t\leq T} S_t \leq L}\\
0,&\ow
\end{cases}$$

And its price can be valued as

$$\ExpH^\QQ_t\SB{e^{-r\P{T-t}} \P{S_T- K}^+ \mathbf 1_{\CB{\min\limits_{0\leq t\leq T}\;\:S_t \leq L}}}$$

A useful relation is that $c_{\text{d&i}} = c - c_{\text{d&o}}$, since

$$\ExpH^\QQ_t\SB{e^{-r\P{T-t}} \P{S_T- K}^+ \mathbf 1_{\CB{\min\limits_{0\leq t\leq T}\;\:S_t \leq L}}} + \ExpH^\QQ_t\SB{e^{-r\P{T-t}} \P{S_T- K}^+ \mathbf 1_{\CB{\min\limits_{0\leq t\leq T}\;\:S_t > L}}} = \ExpH^\QQ_t\SB{e^{-r\P{T-t}} \P{S_T- K}^+}$$

Besides, they cost less than their plain vanilla options.

## Lookback Options
They are those whose payoff is depended on the maximum or minimum asset price reached during the life of the contract.

- ***floating strike***: the *optimal value* of the underlying asset’s price during the option life.
    - ***floating lookback call option***: with payoff $S_T - \d{\min_{0 \leq t \leq T}}S_t$
    - ***floating lookback put option***: with payoff $\d{\max_{0 \leq t \leq T}}S_t - S_T$
- ***fixed strike***: $K$, while the price will be considered NOT floating
    - ***fixed lookback call option***: with payoff $\big(\d{\max_{0 \leq t \leq T}}S_t - K\big)^+$
    - ***fixed lookback put option***: with payoff $\big(K - \d{\min_{0 \leq t \leq T}}S_t\big)^+$

## Exchange options
The holder has the right to exchange asset $S_t^\P{1}$ for $S_t^\P{2}$ and thus its payoff is $\P{S_T^\P{2} - S_T^\P 1}^+$. There're two method to price this, and the first is baesd on the **Two Risky Assets model**

### Two Risky Assets model
Now $\dd S_t^\P 1 = S_t^\P 1 \SB{\mu_1 \dd t + \sigma_1 \dd W_t^\P 1}$ and $\dd S_t^\P 2 = S_t^\P 2 \SB{\mu_2 \dd t + \sigma_2 \dd W_t^\P 2}$ and suppose that $W_t^\P 1$ and $W_t^\P 2$ are Brownian motions with instantaneous correlation $\rho$ such that $\Exp{W_t^\P 1W_t^\P 2} = \rho t$, $\forall\: t> 0$.

### Two-Dimensional Itô's rule
The $\CB{X_t}$ and $\CB{Y_t}$ are two Itô processes:

$$\begin{align}
\dd X_t &= \mu_X \dd t + \sigma_{X,1}\dd W_t^\P 1 + \sigma_{X,2} \dd W_t^\P 2\\
\dd Y_t &= \mu_Y \dd t + \sigma_{Y,1}\dd W_t^\P 1 + \sigma_{Y,2} \dd W_t^\P 2
\end{align}$$

Then combine the two we have the Itô's formula like

$$\dd f\P{t,X_t,Y_t} = \ffrac{\partial f}{\partial t}\dd t +\ffrac{\partial f}{\partial X}\dd X_t +\ffrac{\partial f}{\partial Y}\dd Y_t + \ffrac{1}{2}\ffrac{\partial^2 f}{\partial X^2}\P{\dd X_t}^2 + \ffrac{1}{2}\ffrac{\partial^2 f}{\partial Y^2}\P{\dd Y_t}^2 + \color{brown}{\ffrac{1}{2}\ffrac{\partial^2 f}{\partial X\partial Y}\P{\dd X_t}\P{\dd Y_t}}$$

### Product Rule
$$\dd \P{X_t Y_t} = X_t \dd Y_t + Y_t \dd X_t + \color{brown}{\dd X_t \dd Y_t}$$

And here $W_t^\P 1$ and $W_t^\P 2$ are Brownian motions with correlation $\rho$, then

$$\dd W_t^\P 1 \dd W_t^\P 2 = \rho \dd t$$

### Dynamic Replication (No Arbitrage)
We now use dynamic replication to price the option. The portfolio process is now two-dimensional $\pi_t = \P{\pi_t^\P 1,\pi_t^\P2}$. 

And the wealth process satisfies

$$\dd V_t = \ffrac{V_t - \P{\pi_t^\P 1 + \pi_t^\P2}}{B_t} \dd B_t + \ffrac{\pi_t^\P1}{S_t^\P1}\dd S_t^\P1 + \ffrac{\pi_t^\P2}{S_t^\P2}\dd S_t^\P2$$

$$\dd V_t = \P{rV_t + \pi_t^\P1\P{\mu_1 - r} + \pi_t^\P2\P{\mu_2 - r}}\dd t + \pi_t^\P1\sigma_1 \dd W_t^\P1 + \pi_t^\P2\sigma_2 \dd W_t^\P2$$

Let $c\P{t, S_t^\P1,S_t^\P2}$ denote the price of the exchange option at time $t$. Using **two-dimensional Itô's rule** yields

$$\begin{align}\dd c &= \ffrac{\partial c}{\partial t}\dd t + \ffrac{\partial c}{\partial S_t^\P 1} \dd S_t^\P 1 + \ffrac{\partial c}{\partial S_t^\P 2} \dd S_t^\P 2 + \ffrac{1}{2}\ffrac{\partial^2 c}{\partial S_t^\P1\partial S_t^\P1} \P{\dd S_t^\P 1}^2 + \ffrac{1}{2}\ffrac{\partial^2 c}{\partial S_t^\P2\partial S_t^\P2} \P{\dd S_t^\P 2}^2 + \color{brown}{\ffrac{1}{2}\ffrac{\partial^2 c}{\partial S_t^\P1\partial S_t^\P2}\dd S_t^\P 1\dd S_t^\P 2}\\
&= \ffrac{\partial c}{\partial t}\dd t + \ffrac{\partial c}{\partial S_t^\P 1} S_t^\P 1 \SB{\mu_1 \dd t + \sigma_1 \dd W_t^\P 1} + \ffrac{\partial c}{\partial S_t^\P 2} S_t^\P 2 \SB{\mu_2 \dd t + \sigma_2 \dd W_t^\P 2}\\
&\bspace + \ffrac{1}{2}\ffrac{\partial^2 c}{\partial S_t^\P1\partial S_t^\P1} \P{S_t^\P 1}^2 \SB{\mu_1 \dd t + \sigma_1 \dd W_t^\P 1}^2 + \ffrac{1}{2}\ffrac{\partial^2 c}{\partial S_t^\P2\partial S_t^\P2} \P{S_t^\P 2}^2 \SB{\mu_2 \dd t + \sigma_2 \dd W_t^\P 2}^2\\
&\bspace + \color{brown}{\ffrac{1}{2}\ffrac{\partial^2 c}{\partial S_t^\P1\partial S_t^\P2}S_t^\P 1 \SB{\mu_1 \dd t + \sigma_1 \dd W_t^\P 1}S_t^\P 2 \SB{\mu_2 \dd t + \sigma_2 \dd W_t^\P 2}}\\
&\bbspace\text{since } \dd t\dd t = 0, \dd W_t^\P 1 \dd W_t^\P 1 = \dd t, \dd W_t^\P 1 \dd t = 0, \text{ and } \color{brown}{\dd W_t^\P 1 \dd W_t^\P 2 = \rho \dd t}\\[0.7em]
&= \Big[ \ffrac{\partial c}{\partial t} + \mu_1 S_t^\P 1 \ffrac{\partial c}{\partial S_t^\P 1} + \mu_2 S_t^\P 2 \ffrac{\partial c}{\partial S_t^\P 2} + \ffrac{1}{2}\P{\sigma_1S_t^\P1}^2 \ffrac{\partial^2 c}{\partial S_t^\P1\partial S_t^\P1} + \ffrac{1}{2}\P{\sigma_2S_t^\P2}^2 \ffrac{\partial^2 c}{\partial S_t^\P2\partial S_t^\P2}\\
&\bspace+ \color{brown}{\rho \sigma_1 \sigma_2 S_t^\P1 S_t^\P2 \ffrac{\partial^2 c}{\partial S_t^\P1\partial S_t^\P2}}\Big]\dd t + \sigma_1 S_t^\P1 \ffrac{\partial c}{\partial S_t^\P1} \dd W_t^\P 1 + \sigma_2 S_t^\P2 \ffrac{\partial c}{\partial S_t^\P2} \dd W_t^\P 2
\end{align}$$

Thus to replicate, $c\P{t,S_t^\P1,S_t^\P2} = V_t$. And similar to the one-asset case, comparing the $\dd t$ term and the $\dd W_t^\P 1$, $\dd W_t^\P 2$ terms, we have the ***Black-Scholes equation for two assets***:

$$\begin{align}
\ffrac{\partial c}{\partial t} + \ffrac{1}{2}\P{\sigma_1S_t^\P1}^2 \ffrac{\partial^2 c}{\partial S_t^\P1\partial S_t^\P1} + \ffrac{1}{2}\P{\sigma_2S_t^\P2}^2 \ffrac{\partial^2 c}{\partial S_t^\P2\partial S_t^\P2} + \rho \sigma_1 \sigma_2 S_t^\P1 S_t^\P2 \ffrac{\partial^2 c}{\partial S_t^\P1\partial S_t^\P2}&\\
+ rS_t^\P1 \ffrac{\partial c}{\partial S_t^\P1} + rS_t^\P2 \ffrac{\partial c}{\partial S_t^\P2} - rc &= 0
\end{align}$$

with the **replicating portfolio** $\ffrac{\pi_t^\P1}{S_t^\P 1} = \ffrac{\partial c}{\partial S_t^\P1}$ and $\ffrac{\pi_t^\P2}{S_t^\P 2} = \ffrac{\partial c}{\partial S_t^\P2}$ and **boundary condition**: $c\P{T,S_T^\P 1,S_T^\P2} = \P{S_T^\P 2 - S_T^\P1}^+ = S_T^\P1\P{\ffrac{S_T^\P 2}{S_T^\P1} - 1}^+$.

And to solve this, a little conjecture would be that the price may be of the form $c\P{t,S_t^\P 1,S_t^\P2} = S_t^\P 1\cdot h\P{t,Z_t}$ for some function $h$ and $Z_t = \ffrac{S_t^\P 2}{S_t^\P 1}$. Thus

$$\begin{align}
0=\;&S_t^\P 1 \ffrac{\partial h}{\partial t} + \ffrac{1}{2} \P{\sigma_1 S_t^\P1}^2 \ffrac{\partial^2 c}{\partial Z_t^2}\P{\ffrac{\partial Z_t}{\partial S_t^\P1}}^2 + \ffrac{1}{2} \P{\sigma_2 S_t^\P2}^2 \ffrac{\partial^2 c}{\partial Z_t^2}\P{\ffrac{\partial Z_t}{\partial S_t^\P2}}^2\\
&\bspace+\rho \sigma_1 \sigma_2 S_t^\P1 S_t^\P2 \ffrac{\partial^2 c}{\partial Z_t^2}\ffrac{\partial Z_t^2}{\partial S_t^\P1\partial S_t^\P2} + rS_t^\P1 \ffrac{\partial c}{\partial Z_t}\ffrac{\partial Z_t}{\partial S_t^\P1} + rS_t^\P2 \ffrac{\partial c}{\partial Z_t}\ffrac{\partial Z_t}{\partial S_t^\P2} - rS_t^\P1 h\P{t,Z_t}\\
\Rightarrow 0=\;& h_t + \ffrac{1}{2} \P{\sigma_1^2 + \sigma_2^2 - 2 \rho\sigma_1\sigma_2}Z_t^2\ffrac{\partial^2 h}{\partial Z_t^2}
\end{align}$$

That's just the form of **Feynman–Kac Formula**, with its **boundary condition**: $h\P{T,Z_T} = \P{Z_T-1}^+$ (with "strike price" $K=1$). Comparing with the standard **Feynman–Kac Formula** we see that $r = 0$, $\mu=0$, and $\sigma = \sqrt{\sigma_1^2 + \sigma_2^2-2\rho\sigma_1\sigma_2}$. So the preceding *PDE and its boundary condition* is equivalent to the following *SDE* (stochastic differential equation)

$$\dd Z_t = \sqrt{\sigma_1^2 + \sigma_2^2-2\rho\sigma_1\sigma_2} \cdot Z_t \dd \hat{W}_t$$

where $\hat{W}_t$ is the standard brownian motion in the measure where $Z_t$ has no drift (which is hard to find and so omitted here). Then we can use the regular BS formula for $h$ and then multiply $S_t^\P1$:

$$c\P{t,S_t^{\P1},S_t^{\P2}} = S_t^{\P1}\P{Z_t\N{\hat d_1} - 1\cot e^0 N\P{\hat d_2}} = S_t^{\P2} \N{\hat d_1} - S_t^{\P1} \N{\hat d_2}$$

$$\begin{align}
\hat d_1 &= \ffrac{1}{\sigma\sqrt{T-t}} \SB{\log\P{Z_t}+ \ffrac{1}{2}\sigma^2\P{T-t}} = \ffrac{1}{\sigma\sqrt{T-t}} \SB{\log\P{\ffrac{S_t^{\P2}}{S_t^{\P1}}}+ \ffrac{1}{2}\sigma^2\P{T-t}}\\
\hat d_2 &= \ffrac{1}{\sigma\sqrt{T-t}} \SB{\log\P{\ffrac{S_t^{\P2}}{S_t^{\P1}}} - \ffrac{1}{2}\sigma^2\P{T-t}}
\end{align}$$

>Obtaining the formula is not the end. Here's a little extension. To price options on a *maximum* or a *minimum* of two assets, remember that
>
>$$\min\P{S_T^\P1,S_T^\P2} = S_T^\P2 - \P{S_T^\P2 - S_T^\P1}^+,\max\P{S_T^\P1,S_T^\P2} = S_T^\P2 + \P{S_T^\P1 - S_T^\P2}^+$$

And I really want to introduce another method to price the exchange option whose payoff is $\P{S_T^\P{2} - S_T^\P 1}^+$. First we write

$$\ExpH^\QQ_t\SB{e^{-r\P{T-t}}\P{S_T^\P2 - S_T^\P1}^+} = \ExpH^\QQ_t\SB{e^{-r\P{T-t}}S_T^\P1\P{\ffrac{S_T^\P2}{S_T^\P1} -1}^+}$$

Out purpose is to find a new measure that could bring $S_T^\P1$ out from the expectation. How? See a remark first

$Remark$

>How should we understand the conclusion from Girsanov's Theorem, that if $\left.\ffrac{\dd \tilde P}{\dd \PP}\right|_{\FcF_t} = L_t$, then $\ExpH^{\tilde P}\SB{\Phi\P t\mid \FcF_s} = \ExpH_s^{\tilde P}\SB{\Phi\P t} = \ExpH_s^{\PP}\SB{\Phi\P t \ffrac{L_t}{L_s}}$? Actually the $L_s$ in the *denominator* implies the "measure transform" of the given information, $\FcF_s$; and the $L_t$ in the *numerator* implies the "measure transform" of the calculated events, $\Phi\P t$. And after this, here's a little generalization
>
>$$\ExpH^\QQ_t\SB{g\P{S_T}} = \ExpH^{\QQ^*}_t\SB{g\P{S_T}\cdot \ffrac{\left.\ffrac{\dd \QQ}{\dd\QQ^*} \right|_{\FcF_T}}{\left.\ffrac{\dd \QQ}{\dd\QQ^*} \right|_{\FcF_t}}}$$
>
>Especially, when $t=0$, $\left.\ffrac{\dd \QQ}{\dd\QQ^*} \right|_{\FcF_0} = 1$ and thus
>
>$$\ExpH^\QQ_0\SB{g\P{S_T}} = \ExpH^{\QQ^*}_0\SB{g\P{S_T}\cdot \left.\ffrac{\dd \QQ}{\dd\QQ^*} \right|_{\FcF_T}}$$

Thinking inversely, we now define $L_t = \left.\ffrac{\dd Q^*}{\dd Q}\right|_{\FcF_t} = \ffrac{e^{-rt}S_t^\P 1}{S_0^\P1}$. We divide $S_0^\P1$ here to gurantee that $L_0 = 1$. Then

$$\begin{align}
\ExpH^\QQ_t\SB{e^{-r\P{T-t}}\P{S_T^\P2 - S_T^\P1}^+} &= \ExpH^\QQ_t\SB{e^{-r\P{T-t}}S_T^\P1\P{\ffrac{S_T^\P2}{S_T^\P1} -1}^+}\\
&= \ExpH^{\QQ^*}_t\SB{e^{-r\P{T-t}}S_T^\P1\P{\ffrac{S_T^\P2}{S_T^\P1} -1}^+ \cdot \ffrac{L_t}{L_T}}\\
&= \ExpH^{\QQ^*}_t\SB{e^{-r\P{T-t}}S_T^\P1\P{\ffrac{S_T^\P2}{S_T^\P1} -1}^+ \cdot e^{-r\P{T-t}}\ffrac{S_t^\P1}{S_T^\P1}}\\
&= S_t^\P1\ExpH^{\QQ^*}_t\SB{\P{\ffrac{S_T^\P2}{S_T^\P1} -1}^+}\\
\end{align}$$

## Alteration to the BS Model
### Merton-Black-Scholes Model with Deterministic Parameters
If the *volatility* is a deterministic function of time only, $\sigma = \sigma_t$; and the interest rate is also a  deterministic function of time, $r = r_t$. Then the BS formula at time $t$ is then modifies by

$$\sigma^2\P{T-t} \longrightarrow \int_t^T \sigma^2\P u\;\dd u,\bspace r\P{T-t}\longrightarrow \int_t^T r\P u\;\dd u\\
\Rightarrow S_T = S_0 \exp\CB{\int_t^T \P{r_s - \ffrac{1}{2}\sigma_s^2}\;\dd s + \int_t^T \sigma_s\;\dd \widetilde W_s}$$

Given the $r.v.$, we define a probability triple, $\P{\Omega, \FcF, \mathbb{P}}$, here $\Omega$ is the ***sample space***, $\FcF$ is a *collection* of subset of $\Omega$, the ***events***, $\Omega \in \FcF$. And $\mathbb{P}$ is the probability of each event, say $A \in \FcF$.

$Def$ *Measurable*

A real-valued $r.v.$ $X$ is a real-valued function on $\Omega$ that is $\FcF$***-measureable***, meaning that $\forall x \in \mathbb{R}, \CB{\omega \in \Omega: X\P{\omega} \leq x }\in \FcF$. The notation is $X \in \FcF$. The distribution can be defined $F\P{x} = \mathbb{P}\SB{X \leq x}$

## Stochastic Processes

$Def$ *Stochastic Processes*

A collection of $r.v.$ that are all defined on a common sample space $\Omega$ and indexed by $t \in I$:

$$\CB{X_t}_{t\in I}, \forall t \in I, X_t : \Omega \to \mathbb{R}$$

- Discrete time process: $I$ discrete
- Continuous time process: $I$ continuous
- Continuous variable process: the underlying variable $X$ can take any value within a certain range.
- Discrete variable process: certain discrete values are possible.

### Discrete Time Stochastic Processes

**e.g.** *Random Walk*

$\bspace$Let $X_1,X_2,\dots$ be $i.i.d.$ distributed $r.v.$ and let $S_n = \sum\limits_{i=1}^n X_i$. Here $S_n$ is the ***random walk***, and the special case when $\mathbb{P}\SB{X_i = 1} = \mathbb{P}\SB{X_i = -1} = 0.5$ is called ***simple random walk***. 

$Def$ *filtration*

To specify a discrete time stochastic process, we require not just a single $\sigma$-filde, but an increasing sequence of them, $\FcF_n \subset \FcF_{n+1} \subset \cdots \subset \FcF$. Then the collection $\CB{\FcF_n}_{n\geq 0}$ is the ***filtration***.

$Def$ *adapted*

We call the **discrete time stochastic process** $\CB{X_n}_{n\geq 0}$ ***adapted*** to the **filtration** $\CB{\FcF_n}_{n\geq 0}$ if $X_n$ is $\FcF_n$-measurable for each $n$.

$Remark$

> $\FcF_n$ can be regarded as encoding all the information about the evolution of the stochastic process up until time $n$. That is, if we know whether each event in $\FcF_n$ happens or not then we can infer the path followed by the stochastic process up until time $n$.

$Def$ *conditional expectation*

The ***conditional expectation*** of $X$ is $\Exp\SB{X\mid \FcF_n}$, given the information available at time $n$.

- Linearity: $\Exp\SB{c_1 X + c_2 Y \mid \FcF_n} = c_1 \Exp\SB{X \mid \FcF_n} + c_2 \Exp\SB{Y \mid \FcF_n}$
- We can taking out what we've known: if $Y$ is $\FcF_n$-measurable, then $\Exp\SB{YX \mid \FcF_n} = Y\Exp\SB{X \mid \FcF_n}$
- Law of iterated expectations: for $0 \leq n \leq m, \Exp\SB{\Exp\SB{X\mid\FcF_m}\mid \FcF_n} = \Exp\SB{X\mid \FcF_n}$
- Independence: if $X$ is independent of $\FcF_n$, then $\Exp\SB{X \mid \FcF_n} = \Exp\SB{X}$

$Remark$

>["The smaller field wins"](https://ocw.mit.edu/courses/sloan-school-of-management/15-070j-advanced-stochastic-processes-fall-2013/lecture-notes/MIT15_070JF13_Lec9.pdf). Actually, for $0 \leq n \leq m$, or $\FcF_n \subset \FcF_m$, we have $\Exp\SB{\Exp\SB{X\mid\FcF_m}\mid \FcF_n} = \Exp\SB{X\mid \FcF_n}$ and $\Exp\SB{\Exp\SB{X\mid\FcF_n}\mid \FcF_m} = \Exp\SB{X\mid \FcF_n}$. The complete proof is not as simple as you may think, so go to see the ref link.

## Martingale

$Def$ *MartingaleÈûÖ*

A discrete time stochastic process $\CB{X_n, n = 0,1,2,\dots}$ is a $P$***-martingale***, if 

1. $\Exp^P\SB{|X_n|} < \infty, \forall n$
2. $\Exp^P\SB{X_{n+1} \mid \FcF_n} = X_n, \forall n$
3. $\CB{X_n}_{n\geq 0}$ is **adapted** to the **filtration** $\CB{\FcF_n}_{n\geq 0}$, or, $X_n$ is $\FcF_n$-measurable for each $n$

**e.g.**  Discounted stock process in Binomial model

We define $\bar S_n = \ffrac{S_n} { e^{rn}}$, then $\Exp^Q\SB{\bar S_{n+1} \mid \FcF_n} = \bar{S}_n$. Thus $\bar S_n$ is a ***Q-martingale***.

$Proof$

> From the definition of $Q$***-martingale***, we first have $ S_0 = e^{-rT}\Exp^Q\SB{S_T}$. While in this chapter, we have $ \ffrac{S_0} {e^{r\cdot 0}} = \Exp^Q\SB{\big.\ffrac{S_T} {e^{rT}}\big| \FcF_T}$. And thus, for any $t$, we have $\bar S_0 = \Exp^Q\SB{\bar S_t \mid \FcF_t}$
>
>And by the property of Stochastic Process, we have $\Exp^Q\SB{\bar S_{n+1} \mid \FcF_n} = \bar{S}_n$, that's it.

$Remark$

>For the binomial model, the discounted price process of a European option is also a $Q$**-martingale**, say $c\P{t}$ the value of a European call option at time maturity $T$, then $c\P{n} = e^{-r\P{T-n}}\EE[Q]{\P{S_T - K}^+ \mid \FcF_n}$ thus $\bar c\P{n} = \ffrac{c\P{n}}{e^{rn}} = e^{-rT}\EE[Q]{\P{S_T - K}^+ \mid \FcF_n}$ is a $Q$**-martingale**.

**e.g.(V)** Simple random walk

This simple random walk $S_n$ is a $\mathbb{P}\newcommand{\abs}[1]{\left| #1 \right|}$-martingale.

>$$\Exp^P\SB{\abs{S_n}} = \Exp^P\SB{\abs{\sum X_n}} = \sum \Exp^P\SB{\abs{X_n}} = n < \infty$$
>
>$$\Exp^P\SB{S_{n+1}\mid\FcF_n} = \ffrac{1} {2}\P{S_n+1} + \ffrac{1} {2}\P{S_n-1} = S_n$$
>
>Also, $X_n$ is $\FcF_n$-measuable then so is $S_n$, for each $n$. Then all three conditions are satisfied! Done!

$Def$

***Supermartingale***: $\forall n, \Exp\SB{|X_n|} < \infty$ and $\forall n, \Exp\SB{X_{n+1}\mid\FcF_n} \leq X_n$

***Submartingale***: $\forall n, \Exp\SB{|X_n|} < \infty$ and $\forall n, \Exp\SB{X_{n+1}\mid\FcF_n} \geq X_n$

## Stopping time

$Def$

Given a sample space $\Omega$ equipped with a filtration $\CB{\FcF_n}_{n \geq 0}$, a ***stopping time*** is a random variable $\tau: \Omega \to \mathbb{Z}_+$ with the property that 

$$\CB{\tau \leq n} \in \FcF_n, \forall n\geq 0$$

So we can decide whether or not $\tau\leq n$ based on the information available at time $n$.

**e.g.** Simple random walk

>We can define $\tau_1$ to be the first time that the random walk takes the value $1$, that is $\tau_1 = \inf \CB{i \geq 0: S_i = 1}$. It's a **stopping time** because
>
>$$\CB{\tau_1 \leq n} = \bigcup_{i=1}^{n}\CB{S_i = 1} \in \FcF_n$$

### American options revisited

Consider an American option that pays $g\P{\tau}$ dollars if exercised at $\tau \leq T$, that is $g\P{\tau} = \P{S_\tau - K}^+$.

The value of American options at time $t$ is given by: $V\P{t} = \d{\max_{t \leq u \leq T}}\Exp^Q\SB{\ffrac{g\P{u}} {e^{r\P{u-t}}}\mid \FcF_t}$

And the Optimal exercise time: $\tau = \min\CB{u \mid V\P{u} = g\P{u}}$ and it's obviously a **stopping time**.

For the multiperiod binomial model, the American option pricing formula is given by the dynamic programming algorithm (see the example in last chapter for more details)

$$\begin{align}
V\P{T} &= g\P{T}\\
V\P{t} &= \max\CB{g\P{t}, e^{-r}\SB{qV^{u}\P{t+1}+ \P{1-q}V^{d}\P{t+1}}}
\end{align}$$

$Lemma.3\newcommand{\AcA}{\mathscr{A}}
\newcommand{\FcF}{\mathscr{F}}$

The discounted price process of American options is a **Q-supermartingale**.

$Proof$

> Let $\tau_{t+1}$ be the optimal exercise time for price $V\P{t+1}$ and write $\Exp^Q\SB{\cdot\mid\FcF_t}$ by $\Exp^Q_t\SB{\cdot}$ for notational convenience, then
>
>$$V\P{t+1} = \max_{t+1 \leq u \leq T} \Exp^Q_{t+1}\SB{\ffrac{g\P{u}} {e^{r\P{u-\P{t+1}}}}} = \Exp^Q_{t+1}\SB{\ffrac{g\P{\tau_{t+1}}} {e^{r\P{\tau_{\void_{t+1}}\:-\P{t+1}}}}}$$
>
>Then we have the inequity for the discounted price:

>$$\begin{align}
\Exp^Q_t\SB{\ffrac{V\P{t+1}} {e^{r\P{t+1}}}} &= \Exp^Q_t\SB{ \Exp^Q_{t+1} \SB{\ffrac{g\P{\tau_{t+1}}} {e^{r\cdot\tau_{\void_{t+1}}}}}} \\
&= \Exp^Q_{t}\SB{\ffrac{g\P{\tau_{t+1}}} {e^{r\cdot\tau_{\void_{t+1}}}}} \leq \Exp^Q_{t}\SB{\ffrac{g\P{\tau_{t}}} {e^{r\cdot\tau_{\void_{t}}}}}\\
&= \max_{t\leq u\leq T} \Exp^Q_t\SB{\ffrac{g\P{u}} {e^{ru}}} = \ffrac{V\P{t}} {e^{rt}}
\end{align}$$

## Continuous Time Stochastic Processes

With tradings happen in such short time intervals, the market security prices can be approximately modeled as continuous time processes.

Basic process: Brownian motion (or the Wiener process).

$Def$

A **continuous time stochastic process** is a ***standard Brownian motion***, if 

- $\forall t >s$, the increment $W_t - W_s$ has the normal distribution $\N{0,t-s}$
- The process $W_t$ has independent increments: for any set of times $0 \leq t_1 < t_2 < \cdots < t_n$, the $r.v.$ $W_{t_{\void_2}} - W_{t_{\void_1}}$, $W_{t_{\void_3}} - W_{t_{\void_2}}$, $\dots$, $W_{t_{\void_{n}}} - W_{t_{\void_{n-1}}}$ are independent.
- $W_0 = 0$ 
- The sample paths: $\CB{W_t: t \geq 0}$ are continuous function of $t$.

### Brownian Motion
We can make a little change in the first condition: $\forall t >s$, the increment $\P{\widehat W_t - \widehat W_s}$ has the normal distribution $\mathcal{N}\P{\mu\P{t-s},\sigma^2\P{t-s}}$. Then we call it the ***Brownian motion with drift*** $\mu$ ***and volatility*** $\sigma$ (not $\sigma^2$). And thus

$\bspace$Any Brownian motion with drift $\mu$ and volatility $\sigma$ can be written as $\widehat W_t = \mu t + \sigma W_t$ where $W_t$ is the standard Brownian motion.

### Martingale in Continuous Time
Let $\renewcommand{\FcF}{\mathcal{F}}\FcF$ be a $\sigma$-field. $\CB{\FcF_t}_{t \geq 0}$ is called **filtration** if:

- $\FcF_t$ is a sub-$\sigma$-field of $\FcF$ for all $t$
- $\FcF_s \subset \FcF_t$ for $s < t$

And we can also define the ***natural filtration*** $\FcF_t^X$ which encodes the information generated by the stochastic process $X$ on the interval $\SB{0,t}$. $A \in \FcF_t^X$ if based upon observations of the ***trajectory*** $\CB{X_s, 0 \leq s \leq t}$, it is possible to decide whether the event $A$ has occurred or not.

Here the notation for **trajectory** is: if the value of a random variable $Z$ can be completely determined given observations of the trajectory $\CB{X_s, 0 \leq s \leq t}$, then we write $Z \in \FcF_t^X$. What's more, roughly speaking, $Z \in \FcF_t^X$ $iff$ $Z$ is a "function" of $\CB{X_s, 0 \leq s \leq t}$. Namely, there is a "function" $g$ such that $Z = g\P{X_s, 0 \leq s \leq t}$

And we call a stochastic process $\CB{Y_t}_{t\geq 0}$ **adapted to** a filtration $\CB{\FcF_t^X}_{t \geq 0}$ if 

$$Y_t \in \FcF_t^X, \bspace\forall t \geq 0$$

### Conditional Expectation

For notational simplicity, denote the Brownian filtration by $\CB{\FcF_t}_{t \geq 0}$. Let $\Exp\SB{X\mid \FcF_t}$ denotes the **conditional expectation** of $X$, given the information available at time $t$.

- Linearity: $\Exp\SB{c_1 X + c_2 Y \mid \FcF_t} = c_1 \Exp\SB{X\mid \FcF_t} + c_2\Exp\SB{X\mid \FcF_t}$
- Taking out: if $Y$ is $\FcF_t$-measurable, $\Exp\SB{YX\mid \FcF_t} = Y\Exp\SB{X \mid \FcF_t}$
- Law of iterated expectations: if $0 \leq s \leq t$, then $\Exp\SB{\Exp\SB{X\mid\FcF_t}\mid \FcF_s} = \Exp\SB{X \mid\FcF_s}$
- Independence: if $X$ is independent of $\FcF_t$, then $\Exp\SB{X \mid\FcF_t} = \Exp\SB{X}$

Let $\P{\Omega, \FcF, \mathbb{P}}$ be a probability space with filtration $\CB{\FcF_t}_{t\geq 0}$. Suppose 

1. $\CB{M_t}_{t \geq 0}$ is a stochastic process adapted to $\CB{\FcF_t}_{t\geq 0}$
2. $\Exp\SB{\abs{M_t}}<\infty, \forall t \geq 0$
3. $\Exp^{\mathbb{P}}\SB{M_t\mid \FcF_s} = M_s$ for any $t\geq s\newcommand{\PP}{\mathbb{P}}$

then we call $\CB{M_t}_{t \geq 0}$ is a $\PP$**-martingale**.

$\PP$**-supermartingale** will be obtained by changing the last condition to $\Exp^{\mathbb{P}}\SB{M_t\mid \FcF_s} \leq M_s$ for any $t\geq s$, and $\PP$**-submartingale** can be obtained by changing the last condition to $\Exp^{\mathbb{P}}\SB{M_t\mid \FcF_s} \geq M_s$ for any $t\geq s$.

$Remark$

>It is possible that a process could be a martingale with respect to ($w.r.t.$) one measure but not another one.

Also, if $X$ is a $\PP$**-supermartingale**, then $\Exp^{\PP}\SB{M_t} \leq \Exp^{\PP}\SB{M_s}$ for all $t \geq s$. This can be obtained by taking the expectation on both sides of $\Exp^{\mathbb{P}}\SB{M_t\mid \FcF_s} = M_s$ and then use the law of iterated expectation.

$Proposition.7$

If $\Exp\SB{\abs{\xi}} < \infty$, let $M_t = \Exp\SB{\xi \mid \FcF_t}$, $t \geq 0$, then $\CB{M_t}_{t \geq 0}$ is a martingale.

$Proof$

>- $M_t = \Exp\SB{\xi \mid \FcF_t} \in \FcF_t$, the first condition satisfied
>- $\Exp\SB{\abs{M_t}} =\Exp\SB{\abs{\Exp\SB{\xi \mid \FcF_t}}} < \Exp\SB{\Exp\SB{\abs{\xi} \mid \FcF_t}} = \Exp\SB{\abs{\xi}} < \infty$, second done
>- $\forall t \geq s$, $\Exp\SB{M_t \mid \FcF_s} = \Exp\SB{\Exp\SB{\xi \mid \FcF_t} \mid \FcF_s}\using{\text{Law of iterated expectations}} \Exp\SB{\xi\mid \FcF_s} = M_s$, third done

$Lemma.4$

## Stochastic Calculus

Suppose the stock price is $S_t = f\P{t,W_t}$. Using **Taylor's Theorem** we have

$$\begin{align}
f\P{t + \Delta t,W_{t+\Delta t}} - f\P{t,W_t} &= \Delta_t \ffrac{\partial f\P{t,W_t}} {\partial t} + O\P{\Delta_t^2} + \P{W_{t+\Delta t}-W_t}\ffrac{\partial f\P{t,W_t}} {\partial W} \\
&\bspace+ \ffrac{1} {2}\P{W_{t+\Delta t}-W_t}^2 \ffrac{\partial^2 f\P{t,W_t}} {\partial W^2} +\cdots \\
&= \Delta_t f_t\P{t,W_t} + O\P{\Delta_t^2} + \P{W_{t+\Delta t}-W_t} f_W\P{t,W_t} + \ffrac{1} {2}\P{W_{t+\Delta t}-W_t}^2 f_{WW}\P{t,W_t}
\end{align}$$

$W_t$ is the Brownian motion so that $\Exp\SB{\P{W_{t+\Delta t} - W_t}^2} = \Delta_t$. And as for the first derivative, if $\P{W_{t+\Delta t}-W_t}^2$ is $O\P{\Delta_t}$ then $W_{t+\Delta t}-W_t$ should be $O\P{\sqrt{\Delta t}}$ which could lead to **unbounded changes** in $S_t$ over a bounded time interval. So we conclude that

$$\dd{S_t} = f_t \P{t,W_t} \dd{t} +f_W\P{t,W_t} \dd{W_t} + \ffrac{1} {2}\: f_{WW} \P{t,W_t} \dd{t}\\
\Downarrow \\
S_t = S_0 + \int_{0}^{t} f_s\P{s,W_s}\dd{s} + \boxed{\int_{0}^{t} f_W\P{s,W_s}\dd{W_s}} + \int_{0}^{t} \ffrac{1} {2} \: f_{WW}\P{s,W_s}\dd{s}$$

Notice that the boxed term is a stochastic integral. Now we seek to make sense of that.

### It√¥ Integral for Simple Processes

We define the ***It√¥ Integral*** for the ***simple integrands*** $h\P{s,W_s}$ and then extend it to nonsimple integrands. This happens because the **Brownian motion** is *nowhere differentiable*.

First we let $\Pi = \CB{t_0,t_1,\dots,t_n}$ denote the ***partition*** of $\SB{0,T}$ and assume that $h\P{s,W_s}$ is constant in $s$ on each interval. For this process $h\P{s,W_s}$ we call it the ***simple process*** (I think this might be the same with **simple integrands**). 

Let $t_0, t_1,\dots,$ be the trading dates and $h\P{s,W_s}$ can be regarded as the position taken in the asset at each trading date and held to the next trading date. Then the gains are

$$
h\P{s,W_s} = \sum_{j=0}^{n-1} \newcommand{\II}{\mathbb{I}}\mathbf{1}_{[t_{\void_j},t_{\void_{j+1}}\:)}\P{s}\cdot h\P{t_j,W_{t_{\void_j}}}
\Rightarrow
\begin{cases}
I\P{t} = h\P{t_0,W_{t_{\void_0}}} \SB{W_t - W_0} & 0 \leq t \leq t_1\\
I\P{t} = h\P{t_0,W_{t_{\void_0}}} \SB{W_{t_{\void_1}} - W_0} + h\P{t_1,W_{t_{\void_1}}} \SB{W_t - W_{t_{\void_1}}} & t_1 \leq t \leq t_2\\
\bspace \cdots & \space\\
I\P{t} = \d{\sum_{j=0}^{k-1}h\P{t_j, W_{t_j}}\SB{W_{t_{\void_{j+1}}} - W_{t_{\void_j}}}} + h\P{t_k,W_{t_{\void_k}}} \SB{W_t - W_{t_{\void_k}}} &  t_k \leq t \leq t_{\void_{k+1}}
\end{cases}$$

Thus for this simple integrand $h\P{t,W_t}$ we have its It√¥ integral is just this process $I\P{t}$:

$$I\P{t} = \sum_{j=0}^{k-1}h\P{t_j, W_{t_j}}\SB{W_{t_{\void_{j+1}}} - W_{t_{\void_j}}} + h\P{t_k,W_{t_{\void_k}}} \SB{W_t - W_{t_{\void_k}}} = \int_{0}^{t} h\P{s, W_s}\dd{W_s}$$

### It√¥ Integral for General Processes

$Assumption$

1. $h\P{t,W_t}$ is adapted to the filtration $\FcF_t$
2. $\Exp\SB{\d{\int_{0}^{T} h^2\P{t,W_t}\dd{t}}}<\infty$

then we'll construct the It√¥ Integral the same way, spliting the time interval and consider

$$I_n\P{t} = \sum_{j=0}^{n-1} h\P{t_j,W_{t_{\void_{j}}}}\SB{W_{t_{\void_{j+1}}} - W_{t_{\void_{j}}}}$$

It√¥, this person showed that there is a limit $I(t)$ such that $\Exp\SB{\P{I_n\P{t} - I\P{t}}^2} \to\infty$, and called the limit: the ***It√¥ integral***

$$I\P{t} = \int_{0}^{t} h\P{s,W_s}\dd{W_s}$$

$Remark$

>A kindly notice: $W_t$ is a **Brownian motion**.

### Properties of It√¥ Integral

- **Continuity**: The paths of the It√¥ Integral $\d{I\P{t} = \int_{0}^{t} h\P{s,W_s}\dd{W_s}}$ (it do have a path! like a Brownian motion) are continuous functions of $t$.
- **Adaptivity**: For each $t$, $I\P{t}$ is $\FcF_t$ measurable.
- **Linearity**: $\d{\int_{0}^{t} \P{h_1\P{s,W_s} + h_2\P{s,W_s}}\dd{W_s} = \int_{0}^{t} h_1\P{s,W_s}\dd{W_s} + \int_{0}^{t} h_2\P{s,W_s}\dd{W_s}}$
- **Martingale**: Yeah, face the truth: $I\P{t}$ is a martingale!
- **It√¥ isometry**: $\d{\Exp\SB{I^2\P{t}} = \int_{0}^{t} \Exp\SB{h^2\P{s,W_s}}\dd{s}}$

$Proof$ Martingale

> Firstly, it's adapted to $\FcF_t$. And since we've assumed that $\Exp\SB{\int_{0}^{T} h^2\P{t,W_t}\dd{t}}<\infty$, we have 
>
>$$\Exp\SB{\abs{I\P{t}}} \leq \Exp\SB{\ffrac{I^2\P{t}+1} {2}}<\infty$$

>Then we write $\Exp_s\SB{I\P{t}} = I\P{s} + \Exp_s\SB{I\P{t} - I\P{s}}$. Now we prove that $\Exp_s\SB{I\P{t} - I\P{s}} = 0$ by partition the interval $\SB{s,t}$ in $n$ subintervals
>
>$$\begin{align}
\Exp_s\SB{I\P{t} - I\P{s}}=& \Exp_s\SB{ \lim_{n\to\infty} \sum_{j=0}^{n-1} h\P{t_j, W_{t_{\void_{j}}}}\P{W_{t_{\void_{j+1}}} - W_{t_{\void_{j}}}} }\\[0.6em]
&\bspace \text{assume the summation is exchangable with the expectation } \\[0.5em]
=& \lim_{n\to\infty} \sum_{j=0}^{n-1} \Exp_s\SB{ h\P{t_j, W_{t_{\void_{j}}}}\P{W_{t_{\void_{j+1}}} - W_{t_{\void_{j}}}} }\\[0.6em]
&\bspace \text{condition on }\FcF_{t_{\void_{j}}}, t_j>s, \Exp_s\SB{X} = \Exp_s\SB{\Exp_t\SB{X}} \\[0.5em]
=& \lim_{n\to\infty} \sum_{j=0}^{n-1} \Exp_s\SB{ \Exp_{t_{\void_{j}}}\SB{ h\P{t_j, W_{t_{\void_{j}}}} \P{W_{t_{\void_{j+1}}} - W_{t_{\void_{j}}}} } }\\
=& \lim_{n\to\infty} \sum_{j=0}^{n-1} \Exp_s\SB{  h\P{t_j, W_{t_{\void_{j}}}} } \Exp_{t_{\void_{j}}} \SB{{W_{t_{\void_{j+1}}} - W_{t_{\void_{j}}}} } \\
=& \lim_{n\to\infty} \sum \Exp_s\SB{  h\P{t_j, W_{t_{\void_{j}}}} } \cdot 0 = 0
\end{align}$$

Now we need to introduce a new concept: ***Quadratic Variation***. But first we need to define the ***mesh***: for the partition of interval $\SB{0,t}$ into $n$ subintervals $\left[t_j,t_{j+1}\right)$ where $0 = t_0 \leq t_1 \leq \cdots \leq t_n = t\newcommand{\norm}[1]{\left\| #1 \right\|}$, its **mesh** is

$$\norm{\Pi} := \max_{1\,\leq \, j \, \leq \, n} \abs{\,t_j - t_{j+1}}$$

with this we have the **sampled quadratic variation** of $\CB{W_t}$ corresponding to this partition $\Pi$:

$$Q_\Pi = \sum_{j=1}^{n} \P{W_{t_{\void_{j}}} - W_{t_{\void_{j-1}}}}^2$$

then if the *limit* of $Q_\Pi$ when the **mesh** goes to $0$ exists, we call it the **quadratic variation** of $\CB{W_t}$ up to time $t$:

$$\SB{W}_t = \lim_{\norm{\Pi}\to 0}Q_{\Pi}$$

$Theorem.5$

Let $W_t$ be a standard Brownian motion, then $\SB{W}_t = t$.

$Proof$

>We are going to prove that $\d{\lim_{\norm{\Pi}\to 0} \sum_{j=1}^{n} \P{W_{t_{\void_{j}}} - W_{t_{\void_{j-1}}}}^2 } = t $. First notice that:
>
>$$\Exp\SB{\sum_{j=1}^{n}\P{W_{t_{\void_j}} - W_{t_{\void_{j-1}}}}^2} = \sum_{j=1}^{n}\Exp\SB{\P{W_{t_{\void_j}} - W_{t_{\void_{j-1}}}}^2} = \sum_{j=1}^{n} \P{t_j - t_{j-1}} = t$$
>
>Then the only thing left is to check its variance:

>$$\begin{align}
\Exp\SB{\P{\sum_{j=1}^{n}\P{W_{t_{\void_j}} - W_{t_{\void_{j-1}}}}^2 - t }^2} &=  \Var{\sum_{j=1}^{n} \P{W_{t_{\void_j}} - W_{t_{\void_{j-1}}}}^2}\\
&= \sum_{j=1}^{n} \Var{\P{W_{t_{\void_j}} - W_{t_{\void_{j-1}}}}^2} \\
&= \sum_{j=1}^{n} 2\P{t_j - t_{j-1}}^2 \to 0
\end{align}$$

$Remark$

>More on the distribution related to $W_t$
>
>$W_t - W_s \sim \N{0,t-s}$, then $\P{t-s}^{-1}\P{W_t - W_s}^2 \sim \chi^2\P{1}$ and what's more, $\Exp\SB{\chi^2\P{k}} = k $ and $\Var{\chi^2\P{k}} = 2k $
>
>And the integral and differential form of this $Theorem$
>
>$$t=\SB{W}_t = \lim_{\norm{\Pi}\to 0} \sum_{j=1}^{n} \P{W_{t_{\void_{j}}} - W_{t_{\void_{j-1}}}}^2 = \int_{0}^{t} \P{\dd{W_s}}^2 \iff \P{\dd{W_t}}^2 = \dd{t}$$

The notation of $\dd{W_t}$ is something not exactly the same with the similar one on a normal independent variable, say $\dd t$. This little $\dd{}$ means something else:

Consider the quadratic variation of function $t$.

$$\begin{align}
&\sum_{j=1}^{n} \P{t_j - t_{j-1}}^2 \leq \norm \Pi \cdot t \sum_{j=1}^{n} \P{t_j - t_{j-1}} =  \norm \Pi \cdot t \to 0 \space \text{ as } \norm \Pi \to 0\\
\Longrightarrow 0 =& \lim_{\norm{\Pi}\to 0}\sum_{j=1}^{n} \P{t_j - t_{j-1}}^2 = \int_{0}^{t}\P{\dd s}^2 \\
\Longrightarrow 0 =& \;\P{\dd t}^2
\end{align}$$

Also, by Cauchy-Schwartz inequality,

$$\begin{align}
&\abs{\sum_{j=1}^{n} \P{W_{t_{\void_j}} - W_{t_{\void_{j-1}}}}\P{t_j - t_{j-1}}}\\
\leq\space & \mathstrut{\sqrt{\sum_{j=1}^{n} \P{W_{t_{\void_j}} - W_{t_{\void_{j-1}}}}^2}} \mathstrut{ \sqrt{ \sum_{j=1}^{n} \P{t_j - t_{j-1}}^2}} \\
&\text{same derivation like the preceding one}\\
\Longrightarrow 0 =& \lim_{\norm{\Pi}\to 0}\sum_{j=1}^{n} \P{W_{t_{\void_j}} - W_{t_{\void_{j-1}}}}\P{t_j - t_{j-1}} =\int_{0}^{s} \dd{W_s}\;\dd{s}\\
\Longrightarrow 0 =& \;\dd{W_t}\dd{t}
\end{align}$$

Compute $\d{\int_{0}^{T} W_t\;\dd{W_t}}$

## It√¥ processes

Suppose we have two *adapted processes*, $\CB{\mu\P{t,\cdot}}$ and $\CB{\sigma\P{t,\cdot}}$ and they satisfy some conditions, then the process $\CB{X_t}$ defined by 

$$X_t = X_0 + \int_0^{t}\mu\P{s,X_s}\;\dd{s} + \int_0^{t}\sigma\P{s,X_s}\;\dd{W_s},\bspace\forall\; t>0$$

this is called a ***It√¥ process***. And its corresponding differential form is:

$$\dd{X_t} = \mu\P{t,X_t}\;\dd{t} + \sigma\P{t,X_t}\;\dd{W_t}$$

- $\mu\P{t,X_t}\;\dd t$ can be interpreted as the expected change in $X$ in an instant $\dd t$.
- $\mu\P{t,X_t}$ is called the ***drift*** of the process $X$ at time $t$. Also!!! If $X_t$ is a **martingale**, this **drift** must be $0$ since $\Exp\SB{\dd X_t} = 0$.
- $\sigma\P{t,X_t}$ is called the ***"diffusion" coefficient*** of $X$ at time $t$.

$Remark$ 

>here is that $\d{\int_{0}^{t} h\P{s,W_s}}\;\dd{W_s}$ is a **Martingale**. We can also see that its expectation is always $0$, so
>
>$$\begin{align}
\Exp\SB{\int_{0}^{t} h\;\dd W_s} &= \Exp\SB{\lim_{\norm{\Pi}\to0}\sum_{j=0}^{n-1}h\P{t_j,W_{t_{\void_{j}}}} \P{W_{t_{\void_{j+1}}}-W_{t_{\void_j}}}}\\
&= \lim_{\norm{\Pi}\to0}\sum_{j=0}^{n-1} \Exp\SB{h\P{t_j,W_{t_{\void_{j}}}} \P{W_{t_{\void_{j+1}}}-W_{t_{\void_j}}}}\\
&= \lim_{\norm{\Pi}\to0}\sum_{j=0}^{n-1} \Exp\bigg[\Exp\SB{h\P{t_j,W_{t_{\void_{j}}}} \P{W_{t_{\void_{j+1}}}-W_{t_{\void_j}}}\mid \FcF_{t_{\void_j}}}\bigg]\\
&= \lim_{\norm{\Pi}\to0}\sum_{j=0}^{n-1} \Exp\bigg[h\P{t_j,W_{t_{\void_{j}}}} \cdot \Exp\SB{ \P{W_{t_{\void_{j+1}}}-W_{t_{\void_j}}}\mid \FcF_{t_{\void_j}}}\bigg]\\
&\bspace\text{by the independency}\\
&= \lim_{\norm{\Pi}\to0}\sum_{j=0}^{n-1} \Exp\SB{h\P{t_j,W_{t_{\void_{j}}}}}\cdot \Exp\SB{W_{t_{\void_{j+1}}}-W_{t_{\void_j}}}\\
&= 0 = \int_{0}^{\mathbf{0}} h\;\dd W_s
\end{align}$$
>
>And with this we have the conclusion that for a It√¥ process $X_t$, given that the first two conditions of being a martingale already satisfied, to say that is is a martingale is to say its drift is $0$

### It√¥'s formula

Assume that the process $X$, $\dd{X_t} = \mu\P{t,X_t}\;\dd{t} + \sigma\P{t,X_t} \;\dd{W_t}$.

Consider a function $f$, twice differentiable and then, define $Y_t = f\P{t,X_t}$, then $Y$ has a stochastic differential given by 

$$\dd{f\P{t,X_t}} = \ffrac{\partial f} {\partial t} \dd{t} + \ffrac{\partial f} {\partial X} \dd{X_t} + \ffrac{1}{2}\ffrac{\partial^2 f} {\partial X^2} \P{\dd X_t}^2$$

or equivalently,

$$\dd{f\P{t,X_t}} = \P{\ffrac{\partial f} {\partial t} + \mu\P{t,X_t} \ffrac{\partial f} {\partial X} +\ffrac{1} {2}\sigma^2\P{t,X_t}\ffrac{\partial^2 f} {\partial X^2} }\dd{t} + \sigma\P{t,X_t}\ffrac{\partial f} {\partial X} \dd{W_t}$$

$Proof$

> First, by **Taylor expansion**
>
>$$\begin{align}
&f\P{t+\Delta_t} - f\P{t,X_t}\\
=\;& \Delta_t f_t\P{t,X_t} + O\P{\Delta_t^2} + \P{X_{t+\Delta_t} -X_t}f_X\P{t,X_t}\\
& \bspace + \ffrac{1}{2}\P{X_{t+\Delta_t} -X_t}^2 f_{XX}\P{t,X_t} + \text{higher order terms}
\end{align}$$

>$\P{X_{t+\Delta_t} -X_t}^2$, this term will not disappear so we need to further simplify this term:
>
>$$\begin{align}
&\P{X_{t+\Delta_t} -X_t}^2\\
=\;& \SB{\mu\P{t,X_t}\Delta_t + \sigma\P{t,X_t} \P{W_{t+\Delta_t} -W_t}}^2 \\
=\;& \mu^2 \P{t,X_t} \Delta_t^2 + 2\mu \P{t,X_t} \sigma\P{t,X_t} \Delta_t \P{W_{t+\Delta_t} -W_t}\\
&\bspace +\sigma^2 \P{t,X_t} \P{W_{t+\Delta_t} -W_t}^2
\end{align}$$
>

>
>$$\P{\dd X_t}^2 = \sigma^2\P{t,X_t} \dd{t}\\
\Longrightarrow \dd f\P{t,X_t} = f_t\dd{t} + f_X \dd X_t + \ffrac{1}{2} \sigma^2 \P{t,X_t} \dd{t}$$

## Feynman-Kac Formula

$Theorem.7$

Assume the function $f$ solves the boundary value problem:

$$\begin{cases}
-rf\P{t,x} +\ffrac{\partial f} {\partial t}\P{t,x} + \mu\P{t,x} \ffrac{\partial f} {\partial x}\P{t,x} + \ffrac{1}{2}\sigma^2 \P{t,x} \ffrac{\partial^2 f} {\partial x^2}\P{t,x} = 0\\
f\P{T,x} = \Phi\P{x} 
\end{cases}$$

And define $\CB{X_t}_{0\leq t\leq T}$ to be the solution of the ***stochastic differential equation***: $\dd{X_t} = \mu\P{t,X_t} \dd{t}+\sigma\P{t,X_t} \dd{W_t}$. Then under some regularity condition, we have

$$f\P{t,x} = \Exp\SB{e^{-r\P{T-t}}f\P{T,X_T}\mid X_t = x} = \Exp\SB{e^{-r\P{T-t}}\Phi\P{X_T}\mid X_t = x}$$

The two equation, are equivalent.

$Remark$

>This $Theorem$ makes sense in that we can consider $f\P{t,X}$ to be an option price and $\Phi\P{t}$ is its payoff, if under risk-neutral measure $Q$. $r$ could be consider the interest rate. Also $\sigma$ is the volatility, $\mu$ is the drift.
>
>And the one with the form of expectation looks somehow like to be convertable to be a **martingale**.

$Proof$

>We apply the It√¥'s formula to $e^{-rt}f\P{t,X_t}$ which gives 
>
>$$\begin{align}
\dd{\P{e^{-rt}f\P{t,X_t}}} &= e^{-rt}\P{-rf +\ffrac{\partial f} {\partial t} +\mu\P{t,X_t}\ffrac{\partial f} {\partial X} + \ffrac{1} {2}\sigma^2 \P{t,X_t} \ffrac{\partial^2 f} {\partial X^2}}\dd{t}\\
&\bspace +e^{-rt} \sigma\P{t,X_t} \ffrac{\partial f} {\partial X} \dd{W_t} \\
&= e^{-rt} \sigma\P{t,X_t} \ffrac{\partial f} {\partial X} \dd{W_t}\\
\Longrightarrow e^{-rT}f\P{T,X_T} &= e^{-rt}f\P{t,X_t} + \int_{t}^{T} e^{-rs} \sigma\P{s,X_s} \ffrac{\partial f} {\partial X} \dd{W_s}
\end{align}$$
>
>The expecation of last term is $0$. Therefore, $e^{-rt}f\P{t,X_t}$ is a martingale and we have
>
>$$\begin{align}
e^{-rt} &= \Exp\SB{e^{-rT} f\P{T,X_T} \mid \FcF_t} = \Exp\SB{e^{-rT}\Phi\P{X_T}\mid \FcF_t}\\
&\text{and actually the left side use the informtion only in $t$}\\
&\text{however more information, before $t$ is given}\\
\Longrightarrow f\P{t,X_t} &= \Exp\SB{f\P{t,X_t}\mid\FcF_t} = \Exp\SB{\Exp\SB{e^{-rT}\Phi\P{X_T}\mid \FcF_t}\mid X_t} \\
&= \Exp\SB{e^{-r\P{T-t}}\Phi\P{X_T}\mid X_t}
\end{align}$$
>
>And it's obvious to see from the expectation form to the PDE form.

To find the $Q$ measure at this time, we introduce the ***Girsanov's Theorem***, with which we can derive a risk neutral measure for pricing contingent claims.

$Theorem.8$ **Girsanov's Theorem**

Suppose that $\CB{W_t}_{t \geq0}$ is a $\PP$-Brownian motion with the natural filtration $\CB{\FcF_t}_{t\geq 0}$ and $\CB{\theta_t}_{t\geq0}$ is an $\CB{\FcF_t}_{t\geq0}$-adapted process such that

$$\Exp\SB{\exp\CB{\ffrac{1}{2}\int_{0}^{T} \theta_t^2 \;\dd{t}}}<\infty$$

Then we define $L_t = \exp\CB{\d{\mathbf-\int_{0}^{t} \theta_s \;\dd{W_s} \mathbf-\ffrac{1}{2}\int_{0}^{t}\theta_s^2 \;\dd{s}}}$ and let $\tilde P$ be the probability measure defined by 

$$\tilde P\SB{A} = \int_A L_t\P{\omega} \PP\P{\dd{\omega}}$$

Then under the new probability measure $\tilde P$, the process defined by

$$\widetilde W_t = W_t \mathbf +\int_{0}^{t}\theta_s\;\dd{s}$$

is a standard Brownian motion.

$Remark$

>$L_t$ is a martingale. This conclusion is guranteed by that $\Exp\SB{\exp\CB{\ffrac{1}{2}\int_{0}^{T} \theta_t^2 \;\dd{t}}}<\infty$, the **Novikov‚Äôs condition**. 
>
>$P$ is the objective probability, $\tilde P$ is the risk-neutral probability in continuous time. And the definition can be rewritten as
>
>$$\left.\ffrac{\dd \tilde P}{\dd \PP}\right|_{\FcF_t} = L_t$$
>
>This $L_t$ is the Radon-Nikodym derivative of $\tilde P$ with respect to $\PP$.
>***
>We have the distribution of $\d{\int_{0}^{t} \theta_s\;\dd{W_s}}$. It's normally distributed with mean $0$ by spliting it apart, and variance $\d{\int_{0}^{t} \Exp\SB{\theta_s^2}\;\dd{s}}$ by the It√¥'s isometry.

The expectation under the new measure $\tilde P$: $\Exp^{\tilde P} \SB{\Phi\P{t}} = \Exp^{\tilde P} \SB{\Phi\P{t}\mid \FcF_0}  = \Exp^\PP\SB{\Phi\P{t}L_t}$, or more generally 

$$\Exp^{\tilde P} \SB{\Phi\P{t}\mid \FcF_s} = \Exp^\PP\SB{\Phi\P{t}\ffrac{L_t}{L_s}\mid\FcF_s}$$

**e.g.**

The basic reference model for stock price in continuous time is

$$\dd S_t = \mu S_t \dd t + \sigma S_t \dd W_t$$

where $\CB{W_t}_{t\geq 0}$ is a $\PP$-Brownian motion. Find a measure under which the discounted stock price $\CB{\tilde S_t}_{t\geq 0}$ is a martingale. Here, $\tilde S_t = e^{-rt}S_t$.

>By It√¥'s formula, under $\PP$ measure, we have
>
>$$\dd \tilde S_t =\tilde S_t \P{-r + \mu} \dd t + \sigma \tilde S_t \dd W_t$$
>
>With the intention to *zero* the **drift** $\mu_t$, we let $\tilde W_t = W_t + \ffrac{\mu - r}{\sigma}t$:
>
>$$\dd \tilde S_t =  \sigma \tilde S_t \dd \tilde W_t$$
>
>According to Girsanov‚Äôs Theorem, we can define, by letting $\theta = \ffrac{\mu-r}{\sigma}$
>
>$$\newcommand{\QQ}{\mathbb{Q}}
\left.\ffrac{\dd\QQ}{\dd\PP}\right|_{\FcF_t} = \exp\CB{-\theta W_t - \ffrac{1}{2}\theta^2 t}$$
>
>Then $\CB{\tilde W_t}_{t\geq0}$ is a $\QQ$-Brownian motion and the discounted stock price $\CB{\tilde S_t}_{t\geq 0 }$ is a $\QQ$-martingale.

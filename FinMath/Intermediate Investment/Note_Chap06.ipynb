# Efficient Diversification
## Diversification and Portfolio Risk
Sources of risk
1. general economic conditions: business cycles, inflation, interest rates, exchange rates, ... **None** of these *macroeconomic* factors can be predicted with certainty.
2. firm-specific influences: management style and philosophy, ... those affect the stock without noticeably affecting other firms.
***
- non-diversifiable risk (market risk, systematic risk): Risk factors common to the whole economy.
    - Ex: business cycle, inflation rate, interest rates, and exchange rates
- diversifiable risk (firm-specific risk, nonsystematic risk, Unique risk): Risk that can be eliminated by *diversification*.
    - Ex: Dell’s success in research and development, its management style and philosophy.

##  Asset Allocation with Two Risky Assets
And with a simple scenario analysis we can find that investing two risky assets will have a lower standard deviation of return than investing only one of them.

For risky assets $\newcommand{\ffrac}{\displaystyle \frac}r_1$ and $r_2$, we have:

$$\begin{array}{c:cc} \hline
 & \text{Use future probability distribution} & {\text{Use historical data}\\ \text{(equal probability)}} \\ \hline
{\text{Expected}\\ \text{return}} & E(r) = \bar{r} = \sum p_i \, r_i & E(r) = \bar{r} = \ffrac{1} {n}\sum r_i\\
\text{Variance} & Var(r) = \sum p_i\,\sideset{}{^2}{(r_i - \bar{r})} & Var(r) =\ffrac{1} {n-1} \sum \sideset{}{^2}{(r_i - \bar{r})} \\
{\text{Standard}\\ \text{Deviation}} & \sigma = \sqrt{ Var(r) } & \sigma = \sqrt{ Var(r) } \\
\text{Covariance} & Cov(r_1,r_2) = \sum p_i \, (\sideset{} {_{1,i}}{r} - \sideset{} {_{1}} {\bar{r}}) \, (\sideset{} {_{2,i}}{r} - \sideset{} {_{2}} {\bar{r}}) & Cov(r_1,r_2) = \ffrac{1} {n-1}\sum (\sideset{} {_{1,i}}{r} - \sideset{} {_{1}} {\bar{r}}) \, (\sideset{} {_{2,i}}{r} - \sideset{} {_{2}} {\bar{r}}) \\
\text{Correlation} & \sideset{} {_{1,2}} {\rho} = \ffrac{Cov(r_1,r_2)} {\sigma_1 \sigma_2} & \sideset{} {_{1,2}} {\rho} = \ffrac{Cov(r_1,r_2)} {\sigma_1 \sigma_2} \\ \hline
\end{array}$$

### The Mean-Variance Criterion
Portfolio A is said to dominate portfolio B if all investors prefer A over B, meaning that A has a higher mean return and lower variance or standard deviation than B:

$$E(r_A) \geq E(r_B) \text{ AND } \sigma_A \leq \sigma_B$$

## The Optimal Risky Portfolio with a Risk-Free Asset
![](./Raw/Efficient Frontier.png)

| Global minimum-variance portfolio (G)| Optimal risky portfolio (P) |
| :----------------------------------: | :-------------------------: |
|   ![](./Raw/GPortfolio.png)  |![](./Raw/OPortfolio.png)|

For the global minimum-variance portfolio, let $\displaystyle \sigma_G = \min_{w_1,w_2} \{ \sigma_P \}$ with constraints $w_1 + w_2 = 1$.

$$\begin{align}
\sigma_P^2 =& w_1\sigma_1^2 + (1 - w_1)^2 \sigma_2^2 + 2w_1(1-w_1) \sigma_{1,2} \\
=& \sigma_1^2 w_1^2 + \sigma_2^2 - 2w_1\sigma_2^2 + \sigma_2^2 w_1^2 + 2\sigma_{1,2} - 2w_1^2 \sigma_{1,2} \\
=& (\sigma_1^2 + \sigma_2^2 - 2\sigma_{1,2})w_1^2 - 2(\sigma_2^2 - \sigma_{1,2}) w_1 + \sigma_2^2
\end{align}$$

So we reach the lowest when $w_1 = \ffrac{\sigma_2^2 - \sigma_{1,2}} {\sigma_1^2 + \sigma_2^2 -2\sigma_{1,2}}$.
***
Then for the Optimal risky portfolio, let $\displaystyle \sigma_O = \max_{w_1,w_2}\{\frac{E(r_P) - r_f} {\sigma_P} \}$ with constraints $w_1 + w_2 = 1$.

The result (let the derivative of sharpe ratio with respect to $w_1$ be $0$) is 

$$
w_1 = \frac{[E(r_1)-r_f]\sigma_2^2 - [E(r_2)-r_f]\sigma_{1,2}}{[E(r_1)-r_f]\sigma_2^2 + [E(r_2)-r_f]\sigma_1^2 - [E(r_1)-r_f + E(r_2)-r_f]\sigma_{1,2}}, w_2 = 1 - w_1 $$

### Optimization for 3-asset risky portfolio
- Minimum variance frontier: $\displaystyle \min_{w_1,w_2,w_3} \{ SD(r_p)\}$, with constraints
    1. $w_1 + w_2 + w_3 = 1$
    2. $E(r_p) = const$

- global minimum variance portfolio $\displaystyle \min_{w_1,w_2,w_3} \{ SD(r_p)\}$ with constraints
    1. only $w_1 + w_2 + w_3 = 1$

- optimal risky portfolio: $\displaystyle \max_{w_1,w_2,w_3} \left \{ \ffrac{E(r_p) - r_f} {\sigma_p} \right\}$
    1. still $w_1 + w_2 + w_3 = 1$
    2. connect it to the risk-free asset to form the optimal CAL

## A Single-Index Stock Market
**Index models**: Model that relates stock returns to returns on both a broad market index and firm-specific factors.

$Def$

**excess return**: for stock $i$, its excess return is denoted as $R_i = r_i - r_f$, rate of return in excess of risk-free return

Then we decompose that into three parts

$$\boxed{ R_i = \alpha + \beta R_m + e_i}, R_m = r_m - r_f$$

1. **alpha**: A stock’s expected return *beyond* that induced by the market index; it's also the expected excess return when the market’s excess return is zero.
2. **beta**: The sensitivity of a security’s excess return to the market index's excess return.
3. **market return**: $r_m$, and **market excess return**: $R_m = r_m - r_f$
4. **Residual (firm-specific or residual risk)**: $e_i$, component of return variance that is independent of the market factor.

$\odot$
So it's assumed that the variance of the firm-specific return, $e_i$ is independent of market performance $r_m$ and $\alpha$, and that $Var(\alpha)$ is $0$.

We can also decompose its variance

$$\begin{align}
Var(R_i) &= Var(\alpha + \beta R_m + e_i) \\
&= Var(\alpha) + Var(\beta R_m) + Var(e_i) + Corr(\dots) \\
&= 0 +  \beta^2 Var(R_m) + Var(e_i) + 0 \\
&= \boxed{ \beta^2 \sigma_m ^2 + \sigma_e^2} \\
&= \boxed{ \text{Systematic risk} + \text{Firm-Specific risk} }
\end{align}$$

### Statistical and Graphical Representation of the Single-Index Model
Use regression model to estimate the parameter $\alpha$ and $\beta$.

![](./Raw/Regression.png)

Excess return on security, $R_i$, as the dependent variable be explained by the independent (explanatory) variable $R_m$, market excess return.

The formula for $\beta$

$$\boxed{\hat{\beta} = \frac{Cov(R_i,R_m)} {Var(R_m)}}$$

$Proof$

We assume that the firm-specific shocks and market excess return is uncorrelated, $i.e.$,

$$\boxed{Cov(e_i, R_m) = 0}$$

so that $Cov(R_i, R_m) = Cov(\alpha + \beta R_m + e_i, R_m) = Cov(\beta R_m, R_m) = \beta Var(R_m)$, which leads to the final result. **Aggressive** for those with $\beta >1$, and **defensive** for those with $\beta$ under $1$. Some are even lower than $0$, which provides a hedge against systematic risk. Because it profits when market goes down.

And the actual algorithm is to minimize the sum of the squared deviations of each point to the regression line, which is also called the **security characteristic line (SCL)**, the plot of a security’s expected excess return over the risk-free rate as a function of the excess return on the market. Algebraically, it can be represented as 

$$\boxed{ E[R_i \mid R_m] = \hat{\alpha}_i + \hat{\beta}_i R_m }$$

which reads: *The expectation of $R_i$ given a value of $R_M$ equals the intercept plus the slope coefficient times the value of $R_m$.* Also, $\beta$ is called the **regression coefficient**.

The regression intercept ($\alpha_D$) also gives us the expected excess return on stock $i$ when market return is *neutral*, that is, equal to the T-bill return.

We also measure the relative importance of systematic risk using $rho$, the correlation coefficient between $R_i$ and $R_m$

$$\begin{align}
\rho^2 &= \frac{\text{Systematic (explained) variance}} {\text{Total variance}} \\
&= \frac{\beta_i^2 \sigma_m^2} {\sigma_i^2} \\
&= \frac{\beta_i^2 \sigma_m^2} {\beta_i^2 \sigma_M^2 + \sigma^2 (e_i)}
\end{align}$$

**perfect correlation**, when $\rho^2 = 1$, no firm-specific effects
### Diversification in a Single-Index Security Market (skipped)
Now consider a portfolio that is equally divided to those whose return follows the single-index model $R_i = \alpha + \beta (r_m - r_f) + e_i$. Now what's the systematic and nonsystematic variances of the portfolio?

The beta of the portfolio is a simple average of the individual security betas, or the sum product of each beta and weight. Anyway that term keep its individual form $\beta R_m$, so that the systematic variance is $\beta^2 \sigma_m^2$. 
***
For the firm-specific risk, consider $n$ stocks with weights $w_i$, where $\sum w_i = 1$. So first the nonsystematic portfolio return is $e_p = \sum (w_i e_i)$. And assume they are all uncorrelated, so that the portfolio nonsystematic variance is

$$\sideset{}{_{e_p}^2}{\sigma} = \sum_{i=1} ^n {w_i^2\sideset{}{_{e_i}^2}{\sigma}}$$

$\odot$So actually this term will become negligible as the number of securities grows and the portfolio becomes ever more diversified.$\triangleleft$

### Using Security Analysis with the Index Model
Now that we can predict the market excess return $R_i$, so that we also predict the $\alpha_i = R_i - \beta_i R_m$. Now following certain steps we can find the position of that stock in the optimal risky portfolio.

$$w_i^0 = \frac{\alpha_i / \sigma^2(e_i)} {R_m/\sigma_m^2}$$

Then adjust that value

$$w_i^* = \frac{w_i^0} {1+w_i^0(1-\beta_i)}, w_M^* = 1- w_i^* = \frac{1 - \beta_i w_i^0} {1+w_i^0(1-\beta_i)}$$

Here the $M$ is the passive portfolio, so that the Sharpe ratio of this optimal portfolio exceeds that of the passive portfolio comes with

$$S_O^2 - S_M^2 = \left( \frac{\sigma_i^2} {\sigma(e_i)} \right)^2 := \left(\textbf{information ratio}\right)^2$$

Then comes more stocks to analyse, we shall finanly go to the **active portfolio**:The portfolio formed by optimally combining analyzed stocks.

For stock $i,j$, with already estimated $\alpha_i, \alpha_j$, known $\sigma(e_i), \sigma(e_j)$, how to determine their weights? Anyway, the result is for this active portfolio $A$, 

$$w_{i,A} = \frac{\sigma_{i}/\sigma^2(e_i)} {\sigma_{i}/\sigma^2(e_i) + \sigma_{j}/\sigma^2(e_j)},w_{j,A} = \frac{\sigma_{j}/\sigma^2(e_j)} {\sigma_{i}/\sigma^2(e_i) + \sigma_{j}/\sigma^2(e_j)}$$

So that
1. $\alpha_A = w_{i,A}\alpha_i + w_{j,A}\alpha_j$
2. $\beta_A = w_{i,A}\beta_i + w_{j,A}\beta_j$
3. $\sigma^2(e_A) = w_{i,A}^2\sigma^2(e_i) + w_{j,A}^2\sigma^2(e_j)$

### Risk and Return with Alternative Long-Term Investments
An ***incorrect*** belief, **time diversification**: spreading your risky investments over many time periods offers a similar benefit in risk reduction as spreading an investment budget over many assets in a given period, which lead to the conclusion that stocks become less risky over longer horizons.

> **e.g.**
>
>$$\begin{array}{l:ccc}\hline
\text{Strategy} & {\text{Invest only risky portfolio} \\ 2 \text{ years}} & {\text{One year risky,}\\ \text{one year risk free}} & {\text{Invest half in risky portfolio}\\ 2 \text{ years}} \\ \hline
\text{Risk Premium} & R+R=2R & 0+R=R & 2\cdot 0.5R = R \\
\text{Variance} & \sigma^2+\sigma^2 = 2\sigma^2 & 0 + \sigma^2 = \sigma^2 & 2\cdot (0.5\sigma)^2 = 0.5\sigma^2 \\
\text{Sharpe Ratio} & \frac{2R} {\sqrt{2}\sigma} = \sqrt{2}S_1 & \frac{R} {\sigma} = S_1 & \frac{R} {\sigma \, /\sqrt{2}} = \sqrt{2}S_1 \\
\text{Price of Risk} & \frac{2R} {2\sideset{}{^{^2}}\sigma} = P_1 & \frac{R} {\sideset{}{^{^2}}\sigma} = P_1 & \frac{R} {\sideset{}{^{^2}}\sigma /\, 2} = 2P_1 \\[2em] \hline
\end{array}
$$
>
>Just a friendly reminder:
>1. $\text{Sharpe Ratio} = \ffrac{\text{Risk Premium}} {SD}$
>2. $\text{Price of Risk} = \ffrac{\text{Risk Premium}} {Var}$

So actually the risk grows propotionally over time. The result is $SD \propto \sqrt{t}$, $i.e.$ the standard deviation of the return over $t$ years is $\sigma \sqrt{t}$.

## Summary
- The expected rate of return of a portfolio is the weighted average of the component asset expected returns with the investment proportions as weights.
- The variance of a portfolio is a sum of the contributions of the component-security variances *plus* terms involving the covariance among assets.
- Even if correlations are positive, the portfolio standard deviation will be less than the weighted average of the component standard deviations, as long as the assets are not perfectly positively correlated. Thus, portfolio diversification is of value as long as assets are less than perfectly correlated.
- The contribution of an asset to portfolio variance depends on its correlation with the other assets in the portfolio, as well as on its own variance. An asset that is perfectly negatively correlated with a portfolio can be used to reduce the portfolio variance to zero. Thus, it can serve as a perfect hedge.
- A portfolio manager identifies the efficient frontier by first establishing estimates for the expected returns and standard deviations and determining the correlations among them. The input data are then fed into an optimization program that produces the investment proportions, expected returns, and standard deviations of the portfolios on the efficient frontier.
- In general, portfolio managers will identify different efficient portfolios because of differences in the methods and quality of security analysis. Managers compete on the quality of their security analysis relative to their management fees.
- If a risk-free asset is available and input data are identical, all investors will choose the same portfolio on the efficient frontier, the one that is tangent to the CAL. All investors with identical input data will hold the identical risky portfolio, differing only in how much each allocates to this optimal portfolio and to the risk-free asset. This result is characterized as the separation principle of portfolio selection.
- The single-index model expresses the excess return on a security as a function of the market excess return: $R_i = \alpha_i + \beta_iR_m + e_i$. This equation also can be interpreted as a regression of the security excess return on the market-index excess return. The regression line has intercept $\alpha_i$ and slope $\beta_i$ and is called the security characteristic line (CAL).
- In a single-index model, the variance of the rate of return on a security or portfolio can be decomposed into systematic and firm-specific risk. The systematic component of variance equals $\beta_i^2$ times the variance of the market excess return. The firm-specific component is the variance of the residual term in the index-model equation.
- The beta of a portfolio is the weighted average of the betas of the component securities. A security with negative beta reduces the portfolio beta, thereby reducing exposure to market volatility. The unique risk of a portfolio approaches zero as the portfolio becomes more highly diversified.

## Exercise
### Q1
An investor can design a risky porgolio based on two stocks, $A$ and $B$. The standard deviation of return on stock $A$ is $24\%$ while the standard deviation on stock $B$ is $14\%$. The correlation coefficient between the return on $A$ and $B$ is $0.35$. The expected return on stock $A$ is $25\%$ while on stock $B$ it is $11\%$. The proportion of the minimum variance portfolio that would be invested in stock $B$ is approximately?

$Answer$

Since $\rho = 0.35$, then we have $Cov(A,B) = \rho \cdot \sigma_A \cdot \sigma_B = 117.6\%\%$. So that to minimum variance, for stock $B$, its weight should be 

$$w_B = \frac{\sigma_A ^2 - Cov(A,B)} {\sigma_A^2 + \sigma_B^2 - 2Cov(A,B)} = \frac{24^2-117.6} {24^2 + 14^2 - 2 \times 117.6} \approx 0.85$$

### Q2
Now for stock $A$ and $B$, their expected return are $18\%$ and $14\%$, and standard deviation are $20\%$ and $5\%$. The correlation coefficient is $0.5$. The risk-free rate of return is $10\%$. So in the optimal risky portfolio what proportion of stock $A$ should take?

$Answer$

Just follow the formula before, we can see that

$$w_A = \frac{[E(r_A)-r_f]\sigma_B^2 - [E(r_B)-r_f]\sigma_{A,B}}{[E(r_A)-r_f]\sigma_B^2 + [E(r_B)-r_f]\sigma_A^2 - [E(r_A)-r_f + E(r_B)-r_f]\sigma_{A,B}}$$

Here $\sigma_{A,B} = \sigma_A \cdot \sigma_B \cdot \rho_{A,B}$

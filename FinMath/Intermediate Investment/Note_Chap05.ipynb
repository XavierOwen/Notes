### Conventions for Annualizing Rates of Return
**Annual percentage return (APR)**: $\text{Per-period rate} \times \text{Periods per year}$

**Effective annual rate (EAR)**: defined as following

$$1 + \text{EAR} = \left( 1 + \frac{\text{APR}} {n}\right)^n$$

So that when $n$ becomes very large, we are actually assume that a **continuously compounded rates**, we have $1 + \text{EAR} = e^{\text{APR}} \Rightarrow \text{APR} = \ln (1 + \text{EAR}):= r_{cc}$

## Inflation and The Real Rate of Interest
**Consumer Price Index (CPI)**: measures the cost of purchasing a representative bundle of goods, the “consumption basket” of a typical urban family of four.

**Inflation rate**: $i$, the rate at which prices are rising, measured as the rate of increase of the CPI. 

Then if we let $r$ be the real interest rate, we have 

$$1 + r = \frac{1 + R} {1 + i} \Rightarrow r \approx R - i$$

here $R$ is the nominal interest rate, the growth rate of money.

**nominal interest rate**: The interest rate in terms of nominal (not adjusted for purchasing power) dollars.

**real interest rate**: The excess of the interest rate over the inflation rate. The growth rate of purchasing power derived from an investment.

### The Equilibrium Nominal Rate of Interest
**Fisher equation**: the nominal rate increases one-for-one with increases in the expected inflation rate, $E(i)$.

$$R = i + E(i)$$

### Risk and Risk Premium
What HPRs are possible, and how likely are they? 

**scenario analysis**: Process of devising a list of possible economic scenarios and specifying the likelihood of each one, as well as the HPR that will be realized in each case.

**probability distribution**: List of possible outcomes with associated probabilities.

**expected return**: The mean value of the distribution of HPR. $E(r)$

$$E(r) = \sum _{s = 1} ^S p(s)r(s)$$

here $s$ stand for each scenario.

**variance**: The expected value of the squared deviation from the mean.

$$\text{Var}(r) \equiv \sigma^2 = \sum _{s = 1} ^S p(s)[r(s) - E(r)]^2$$

**standard deviation**: The square root of the variance.

$$\text{SD}(r) \equiv \sigma = \sqrt{\text{Var}(r)}$$

### The Normal Distribution
We can transform any normally distributed return, $r_i$, into a “standard deviation score,” with $0$ mean and $1$ variance.

$$sr_i = \frac{r_i - E(r_i)} {\sigma_i}$$

$Conclusion$
1. The return on a portfolio comprising two or more assets whose returns are normally distributed also will be normally distributed.
2. The normal distribution is completely described by its mean and standard deviation. No other statistic is needed to learn about the behavior of normally distributed returns.
3. The standard deviation is the appropriate measure of risk for a portfolio of assets with normally distributed returns. In this case, no other statistic can improve the risk assessment conveyed by the standard deviation of a portfolio.

***
**value at risk (VaR)**: Measure of downside risk. The worst loss that will be suffered with a given probability, often 5%: *How much would I lose in a fairly extreme outcome*

$\odot$

### Deviation from Normality and Value at Risk
When returns are normal, knowing just the mean and standard deviation allows us to fully describe the entire distribution. In that case, we do not need to estimate VaR explicitly—we can calculate it exactly from the properties of the normal distribution. But when returns are not normal, the VaR conveys important additional information beyond mean and standard deviation, specifically, the probability of severe losses arising from deviations from normality.

**Kurtosis**: Measure of the fatness of the tails of a probability distribution relative to that of a normal distribution. Indicates likelihood of extreme outcomes. $\odot$ Positive values indicate higher frequency of extreme values than this benchmark.$\square$

**Skew**: Measure of the asymmetry of a probability distribution.$\odot$ Negative skew suggests that extreme negative values are more frequent than extreme positive ones.$\square$

When the two additional statistics are nonzero, be careful using the $\text{VaR}$

### Risk Premiums and Risk Aversion
**risk-free rate**: The rate of return that can be earned with certainty. (like on treasury bills)

**risk premium**: An expected return in excess of that on risk-free securities. (expected HPR - risk-free rate)

**excess return**: Rate of return in excess of the risk-free rate.

**risk aversion**: Reluctance to accept risk.

Investors taking on risk to earn a risk premium are speculating; gambling is the assumption of risk for no purpose beyond the enjoyment of the risk itself.

To determine an investor’s optimal portfolio strategy, now we only look at the complete portfolio and quantify his degree of risk aversion. Following the *principle of revealed preference: We infer preferences from the choices individuals are willing to make*, for portfolio $Q$ with a variance $\sigma_{Q}^2$, the degree of risk aversion is 

$$\boxed{A = \frac{E(r_Q) - r_f} {\sigma_{Q}^2}}$$

**price of risk**: $A$, the ratio of portfolio risk premium to variance.

And there's an estimate of average price of risk for all investors in a market. Suppose average net borrowing and lending are zero, in which case the average investor's complete portfolio can be represented by a stock-market index $M$. We use a small sample, such as the S&P 500 to do the calculation.

$$\bar{A} = \frac{\bar{r}_M - r_f} {\sigma_M^2}$$

This $\bar{A}$ is also called the **market price of risk**.

### The Sharpe Ratio
**Sharpe ratio**: Ratio of portfolio risk premium to standard deviation.

$$S = \frac{\text{Portfolio risk premium}} {\text{Standard deviation of portfolio excess return}} = \boxed{ \frac{E(\sigma_P) - r_f} {\sigma_P}}$$

$\odot$
A higher Sharpe ratio indicates a better reward per unit of SD, a stock with higher efficiency$\square$

**mean-variance analysis**: Ranking portfolios by their Sharpe ratios.

## Asset Allocation Across Risky and Risk-Free Portfolios
A simple strategy to control portfolio risk is to specify the fraction of the portfolio invested in broad asset classes such as stocks, bonds, and safe assets.

**capital allocation to risky assets**: The choice between risky and risk-free assets.

**complete portfolio**: The entire portfolio including risky and risk-free assets.

### The Risk-Free Asset
$\odot$
1. Commonly Treasury bills are viewed as the risk-free asset.
2. All the money market instruments are virtually immune to interest rate risk.
3. The risk-free asset has a standard deviation of zero but unknown return rate.
$\square$

### Portfolio Expected Return and Risk
Assume $y$ fraction will be allocated to the risky portfolio, and the left $(1-y)$ for the risk free asset. We use $r_C$ for the return of complete portfolio and $r_P$ for the risky asset return. Then we have

$$E[r_C] = y(E[r_P]) + (1-y)r_f \Rightarrow E[r_C] - r_f = y(E[r_P] - r_f)$$

And besides, since the standard deviation of risk-free asset is $0$, we have $\sigma_C = y \cdot \sigma_P$.

### The Capital Allocation Line
**capital allocation line (CAL)**: Plot of risk-return combinations available by varying portfolio allocation between a risk-free asset and a risky portfolio.

Its slope is the Sharpe Ratio, or sometimes called the *reward-to-volatility* ratio.

### Risk Aversion and Capital Allocation
More risk-averse, less of the risky asset, more of the risk-free asset, nearer to point $F$ in the CAL. Just don't like taking any possible risk. (The opposite is called the risk-tolerant investors)

We let $A$ be the degree of risk aversion.

$$y = \frac{\text{Available risk premium to variance ratio}} {\text{Required risk premium to variance ratio}} = \frac{[E(r_P) - r_f]/\sigma_P^2} {A} = \frac{E(r_P) - r_f} {A\sigma_P^2}$$

Here the *available risk premium* can be just as the *market price of risk*, $\bar{A}$.

## Passive Strategies and the Capital Market Line
**passive strategy**: Investment policy that avoids security analysis. $\odot$ A passive strategy is based on the premise that securities are fairly priced, and it avoids the costs involved in undertaking security analysis.$\triangleleft$

To avoid the costs of acquiring information on any individual stock or group of stocks, we may follow a "neutral" **diversification approach**. Select a diversified portfolio of common stocks that mirrors the corporate sector. This results in a *value-weighted* portfolio, which, for example, invests a proportion in GE stock that is equal to the ratio of GE’s market value to the market value of all listed stocks. And this strategy is called **indexing** (The rate of return on the portfolio replicates the return on the index).

**capital market line (CML)**: The capital allocation line using the market index portfolio as the risky asset.

Normally we can use S&P 500 index as the market index. It can be also seen as the CAL connecting 1-month T-bills and a broad index of common stocks (e.g., S&P 500 Index).

For those who want more excess return but less risk, it means higher **Sharpe ratio**, which is $\displaystyle \frac{E(P) - r_f} {\sigma}$.

## Summary
- Investors face a trade-off between risk and expected return. Historical data confirm our intuition that assets with low degrees of risk should provide lower returns on average than do those of higher risk.
- Shifting funds from the risky portfolio to the risk-free asset is the simplest way to reduce risk. Another method involves diversification of the risky portfolio. We take up diversification in later chapters.
- U.S. T-bills provide a perfectly risk-free asset in nominal terms only. Nevertheless, the standard deviation of real rates on short-term T-bills is small compared to that of assets such as long-term bonds and common stocks, so for the purpose of our analysis, we consider T-bills the risk-free asset. Besides T-bills, money market funds hold short-term, safe obligations such as commercial paper and CDs. These entail some default risk but relatively little compared to most other risky assets. For convenience, we often refer to money market funds as risk-free assets.
- A risky investment portfolio (referred to here as the risky asset) can be characterized by its Sharpe, or reward-to-volatility ratio. This ratio is the slope of the capital allocation line (CAL), the line connecting the risk-free asset to the risky asset. All combinations of the risky and risk-free asset lie on this line. Investors would prefer a steeper-sloping CAL, because that means higher expected returns for any level of risk.
- An investor’s preferred choice among the portfolios on the capital allocation line will depend on risk aversion. Risk-averse investors will weight their complete portfolios more heavily toward Treasury bills. Risk-tolerant investors will hold higher proportions of their complete portfolios in the risky asset.
- The capital market line is the capital allocation line that results from using a passive investment strategy that treats a market index portfolio, such as the Standard & Poor’s 500, as the risky asset. Passive strategies are low-cost ways of obtaining well-diversified portfolios with performance that will reflect that of the broad stock market.

## Key Formulae
- Arithmetic average of $n$ returns, $\displaystyle (r_1 + r_2 + \cdots + r_n) \over n$
- Geometric average of $n$ returns, $(1 + r_1)(1 + r_2)\cdots (1 + r_n)^{1/n}-1$ or $\displaystyle \left( p_{\text{final}} \over p_{\text{start}}\right)^{1/n} - 1$
- Continuously compounded rate of return, $\text{APR} = r_{cc}= \ln(1 + \text{Effective annual rate})$
- Real rate of return: $\displaystyle \left( 1 + {\text{Nominal Return Rate}} \over 1 + {\text{Inflation Rate}} \right)^{1/n} - 1$
- Real rate of return (continuous compounding): $r_{\text{nominal}} - \text{Inflation Rate}$
- Expected return: $\sum [p(\text{Scenario}) \times \text{ Return in scenario}]$
- Variance: $\sum [p(\text{Scenario}) \times \text{ Deviation from mean in scenario}^2]$
- Standard Deviation: $\sqrt{\text{Variance}}$
- Sharpe Ratio: $\displaystyle {\text{Portfolio risk premium} \over \text{Standard Deviation of excess return}} = \frac{E(r_p) - r_f} {\sigma_p}$
- Optimal capital allocation to the risky asset: $y = \displaystyle \frac{E(r_p) - r_f} {A\sigma_p^2}$

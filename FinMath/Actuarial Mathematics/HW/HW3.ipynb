## Question 3
($\text{a}$)

$$\begin{align}

($\text{b}$)

Since the payment could only happen in two times, that is when she is $85$ years old, or $100$ years old, there're only three possible value for $Z$.
1. Dies before age $85$, $Z = \$0$
2. Dies after age $85$ before $100$, $Z = \ffrac{\$500000} {(1.08)^{20}} = \$107274.10$
3. Dies after age $100$, $Z = \ffrac{\$500000} {(1.08)^{20}} + \ffrac{\$1000000} {(1.08)^{35}} = \$174908.65$

And for each of $Z$ value we have$$\DeclareMathOperator*{\Pr}{Pr}$$
1. $\Pr[Z=0] = {_{20}}q_{65} = 1 - {_{20}}p_{65} = 1 - \sqrt{\ffrac{1} {20}}\approx 0.7764$
2. $\Pr[Z = 107274.10] = {_{20|15}}q_{65} = {_{20}}p_{65} - {_{35}}p_{65} = \sqrt{\ffrac{1} {20}} - \sqrt{\ffrac{1} {35}} \approx 0.0546$
3. $\Pr[Z = 174908.65] = {_{35}}p_{65} = \sqrt{\ffrac{1} {35}} \approx 0.1690$

So to sum up, we have the following table

$$\begin{array}{c:ccc}\hline
z & 0 & 107274.10 & 174908.65 \\ \hline
\Pr[Z = z] & 0.7764 & 0.0546 & 0.1690\\ \hline
\end{array}$$


($\text{c}$)

Easily, we can find the probability that the PV of the benefits is less than $\$100000$ from the table is under the condition that $Z = \$0$, so that the value is $0.7764$.

($\text{d}$)

$$\begin{align}
\text{EPV} &= \sum_{z} z \cdot \Pr[Z = z] \\
&= 0 \times 0.7764 + 107274.10 \times 0.0546 + 174908.65 \times 0.1690 \\
&= 35416.72602
\end{align}$$

($\text{e}$)

$$\begin{align}

&= \sqrt{ 0.7764 \times \left( 0 - 35416.72 \right)^2 + 0.0546 \times \left( 107274.10 - 35416.72 \right)^2 + 0.1690 \times \left( 174908.65 - 35416.72 \right)^2}\\

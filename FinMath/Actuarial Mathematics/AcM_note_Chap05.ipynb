## Annual life annuities
Annuities whose payments are contingent on the survival of the annuitant, once each year.

### Whole life annuity-due

Standard whole-life annuity-due: pay $1$ dollar in advance until the death of the annuitant, meaning that the first payment occurs immediately, the second in one year from now, provided that $(x)$ is alive then, and payments follow at annual intervals with each payment conditional on the survival of $(x)$ to the payment date.

So there're totally $K_x +1$ times payments if the death occurs in $[K_x, K_x+1)$. So the present value of the series of payment is $Y = \text{PV} = \ddot{a}_{\overline{k+1}\displaystyle{|}}$, where $k := K_x$.

$$
\begin{align}
\ddot{a}_{x} &= \EE{Y} \\
&= \EE{ \frac{1 - \upsilon^{k+1}} {d}} \\
&= \frac{1- \EE{\upsilon^{k+1}}} {d} \\
&= \frac{1- A_x} {d}
\end{align}$$

$$\begin{align}
V[Y] &= V \left[ \frac{1 - \upsilon^{k+1}} {d} \right] \\
&= \frac{1} {d^2} V\left[ \upsilon^{k+1} \right] \\
&= \frac{{^{2}}A_{x} - A_{x}^2} {d^2}
\end{align}$$

Then we apply the idea of indicator variable to derive a more specific formula for $\ddot{a}_{x}$.

$$Y = I\left( T_x > 0 \right) + \upsilon \I{ T_x > 1} + \upsilon^2 \I{ T_x > 2} + \upsilon^3 \I{ T_x > 3} + \cdots $$

And since $\EE{\I{ T_x > 3 }} = \Pr \left[ T_x > t \right] = {_{t}}p_{x}$, so that we find the expected value of each side:

$$\EE {Y} = \ddot{a}_x = 1 + \upsilon p_x + \upsilon^2 {_{2}}p_{x} + \upsilon^3 {_{3}}p_{x} + \cdots = \sum_{k = 0}^{\infty} \upsilon^k {_{k}}p_{x}$$

And since we also have $\Pr\left[ K_x = k\right] = {_{k|}}q_{x}$, or we can just consider break them into parts and recombine

$$\ddot{a}_x = \sum_{k = 0}^{\infty} \upsilon^k {_{k}}p_{x} = \sum_{k=0}^{\infty} \ddot{a}_{\overline{k+1}\displaystyle{|}} \cdot {_{k|}}q_{x}$$

### Term annuity-due
Let the present value of this annuity is $Y$, so that actually we have

$$Y = \begin{cases}
\ddot{a}_{\overline{K_x+1}\d{|}} & \text{ if } K_x=0,1,\dots,n-1 \\ 
\ddot{a}_{\overline{n}|} & \text{ if } K_x \geq n 
\end{cases}$$

that is $Y=\ddot{a}_{\overline{\min{\left( K_x+1,n \right)}}\d{|}} = \ffrac{1 - \upsilon^{\min{\left( K_x+1,n \right)}}} {d}$ and then the EPV would be $\text{EPV} := \ddot{a}_{x:\overline{n}|} = \ffrac{1 - A_{x:\overline{n}|}} {d}$.

Another way is to directly summing the EPVs of individual payments:

$$\ddot{a}_{x:\overline{n}|} = \sum_{t=0}^{n-1} \upsilon^{t} \cdot {_{t}}p_{x}  = 1 + \upsilon\cdot p_{x} + \upsilon^2 \cdot {_{2}}p_{x} + \upsilon^3 \cdot {_{3}}p_{x} + \cdots$$

We can also rewrite that as

$$\ddot{a}_{x:\overline{n}|} = \sum_{k=0}^{n-1}\ddot{a}_{\overline{k+1}\d{|}}\cdot {_{k|}}q_{x}+{_{n}}p_{x} \cdot \ddot{a}_{\overline{n}|} $$

### Whole life immediate annuity
This time we use the $Y^*$ to denote the present value for this whole life immediate annuity, using the indicator

$$Y^* = \upsilon \I{T_x > 1} + \upsilon^2 \I{T_x >2} + \upsilon^3 \I{T_x > 3} + \cdots$$

So that the expected present value of this would be $\EE{Y^*} :=a_x = \EE{Y}-1=\ddot{a}_{x}-1$; also we have the variation: $\Var{Y^*} = \Var{Y} = \ffrac{{^{2}}A_{x} - A_{x}^2} {d^2}$

### Term immediate annuity
The ramdom variable for this is $Y = a_{\overline{\min\left(K_x, n\right)}\d{|}}$, and we can summing the EPVs of the individual payments, 

$$a_{x:\overline{n}|} = \sum_{t=0}^{n-1}\upsilon^{t+1}{_{t+1}}p_{x}=\sum_{t=1}^{n}\upsilon^{t}{_{t}}p_{x} = \upsilon \cdot p_{x} + \upsilon^2 \cdot {_{2}}p_{x} + \upsilon^3 \cdot {_{3}}p_{x} + \cdots$$

And we can difference this equation and that one above that then we can see

$$\ddot{a}_{x:\overline{n}|} - a_{x:\overline{n}|}=1- \upsilon^n \cdot {_{n}}p_{x} \Rightarrow a_{x:\overline{n}|} = \ddot{a}_{x:\overline{n}|} + \upsilon^n \cdot {_{n}}p_{x} - 1$$

## Annuities payable continuously
### Whole life continuous annuity
Now for a infinitesimal interval $(t + \mathrm{d}t)$ the payment under the annuity is $\mathrm{d}t$ provided $(x)$ is alive through the interval.

We would write $Y = \bar{a}_{\overline{T_x}\d{|}}:=\bar{a}_x$. And analogous to the annual annuity-due, we can also get the formula for that, using three different way:

$$\bar{a}_{x} = \EE{Y} = \EE{\frac{1 - \upsilon^{T_x}} {\delta}} = \frac{1 - \EE{\upsilon^{T_x}}} {\delta} = \frac{1 - \bar{A}_x} {\delta}$$

And we have the variance as $\Var{Y} = \Var{\ffrac{1 - \upsilon^{T_x}} {\delta}} = \ffrac{{^{2}}\bar{A}_{x} - \bar{A}_{x}^{2}} {\delta^2}$. Now the second approach is using the integral

$$\bar{a}_{x} = \int_{0}^{\infty} e^{-\delta t} {_{t}}p_{x} \;\mathrm{d}t$$

This can also be derived using the $I.r.v.$: $Y=\d{\int_{0}^{\infty}e^{-\delta t}\cdot\I{T_x >t}\; \mathrm{d}t}$. Finally, we can directly write down the EPV from the distribution of $T_x$:

$$\bar{a}_{x} = \int_{0}^{\infty} \bar{a}_{\overline{t}\d{|}} \cdot {_{t}}p_{x} \cdot \mu_{x+t} \; \mathrm{d}t$$

We can evaluate this formula first using the fact that $\ffrac{\mathrm{d}} {\mathrm{d}t} \bar{a} _{\overline{t}\d{|}} = \upsilon^t = e^{-\delta t}$ derived by differentiating the equation $\d{\bar{a}_n=\int_{0} ^{n} \upsilon^t \; \mathrm{d}t = \ffrac{1 - \upsilon^n} {\delta}}$, then we have.

$$
\begin{align}
\bar{a}_{x} &= \int_{0}^{\infty} \bar{a}_{\overline{t}\d{|}} \cdot {_{t}}p_{x} \cdot \mu_{x+t} \; \mathrm{d}t \\
&= \int_{0}^{\infty} \bar{a}_{\overline{t}\d{|}} \cdot \frac{\mathrm{d}} {\mathrm{d}t}\left( -{_{t}}p_{x} \right) \; \mathrm{d}t \\
&= - \left( \bar{a}_{\overline{t}\d{|}} \cdot {_{t}}p_{x} \big.\big|_{0}^{\infty} - \int_{0}^{\infty} {_{t}}p_{x} \cdot e^{- \delta t} \; \mathrm{d}t \right) \\
&= \int_{0}^{\infty} e^{- \delta t} \cdot {_{t}}p_{x} \; \mathrm{d}t
\end{align}$$

and especially when $\delta = 0$, we have $\bar{a}_{x} = \overset{\circ}{e}_{x}$, the complete expectation of life.

### Term continuous annuity
This time the present value $r.v.$ is $\bar{a}_{\overline{\min{\left(T_x\,,n\right)}}\d{|}}=\ffrac{1 - \upsilon^ {\min{\left(T_x\,,n\right)}}} {\delta}$ with its EPV denoted as $\bar{a}_{x: \overline{n}|}$. Analogous to the term annuity-due, we have three expressions for this EPV.

- using the result of $\EE{{\upsilon}^{\min{\left(T_x\,,n\right)}}} = \bar{A}_{x: \overline{n}|}\\[0.7em]$
$$
\bar{a}_{x: \overline{n}|} = \EE{\frac{1 - \upsilon^{\min{\left(T_x\,,n\right)}}} {\delta}} = \frac{1 - \EE{\upsilon^{\min{\left(T_x\,,n\right)}}}} {\delta} = \frac{1 - \bar{A}_{x: \overline{n}|}} {\delta}\\[0.7em]
$$

- using the $I.r.v.$ approach we have $\\[0.7em]$
$$
\bar{a}_{x: \overline{n}|} = \int_{0}^{n} e^{-\delta t} {_{t}}p_{x} \; \mathrm{d}t \\[0.7em]
$$

- taking the expected value of the present value random variable$\\[0.7em]$
$$\bar{a}_{x: \overline{n}|} = \int_{0}^{n} \bar{a}_{\overline{t}\d{|}} \cdot {_{t}}p_{x} \cdot \mu_{x+t} \; \mathrm{d}t + \bar{a}_{\overline{n}|} \cdot {_{n}}p_{x} \\[0.7em]
$$

## Annuities payable $m$ times per year
### Intro
Recall the definition of $K_{x}^{(m)} = \ffrac{1}{m}\left \lfloor mT_x \right \rfloor$ ,the complete future lifetime rounded down to the lower $1 / m\text{th}$ of a year. And here we call $\ddot{a}_{\overline{n}|}^{(m)}$ an **annual factor**. 

### Life annuities payable $m$ times a year
Consider first an annuity of total amount $1$ per year, payable *in advance* $m$ times per year throughout the lifetime of $( x )$ , with each payment being $1 / m$. Then the present value random variable for this annuity is

$$\ddot{a}^{(m)}_{\overline{K_{x}^{(m)}\,+\frac{1} {m}}\big.\big|} = \frac{1 - \upsilon^{K_{x}^{(m)}\,+\frac{1} {m}}} {d^{(m)}}$$

so that the EPV is

$$\ddot{a}^{(m)}_{x} = \frac{1 - \EE{\upsilon^{K_{x}^{(m)}\,+\frac{1} {m}}}} {d^{(m)}} = \frac{1 - A_{x}^{(m)}} {d^{(m)}}$$

But still we can apply the indicator random variable method then we can see

$$\ddot{a}^{(m)}_{x} = \sum_{r=0}^{\infty} \frac{1} {m} \upsilon^{\frac{r} {m}} \cdot {_{\frac{r} {m}}}p_{x}$$

If payable in arrear, similar to the idea of annuity-due, we have the formula: $a_{x}^{(m)} = \ddot{a}_{x}^{(m)} -\ffrac{1} {m}$

### Term annuities payable $m$ times a year
This time the present value $r.v.$ is expressed like $\ddot{a}^{(m)}_{\overline{\min\left(K_{x}^{(m)}\,+\frac{1} {m}\, , n\right)}\big.\big|} = \ffrac{1 - \upsilon^{\min\left(K_{x}^{(m)}\,+\frac{1} {m}\, , n\right)}} {d^{(m)}}$. And the formula for EPV is

$$\ddot{a}_{x:\overline{n}|}^{(m)} = \frac{1 - \EE{\upsilon^{\min\left(K_{x}^{(m)}\,+\frac{1} {m}\, , n\right)}}} {d^{(m)}} = \frac{1 - A_{x:\overline{n}|}^{(m)}} {d^{(m)}}$$

Or using the $I.r.v.$ approach we find that

$$\ddot{a}_{x:\overline{n}|}^{(m)} = \sum_{r=0}^{mn-1} \frac{1} {m} \upsilon^{\frac{r} {m}} \cdot {_{\frac{r} {m}}}p_{x}$$

At last, for the $1 / m\text{thly}$ term immediate annuity, we can easily find that by comparing with the annuity due, and we can see that

$$a_{x:\overline{n}|}^{(m)} = \ddot{a}_{x:\overline{n}|}^{(m)} - \ffrac{1} {m} \left( 1 - \upsilon^n \cdot {_{n}}p_{x} \right)$$

## Deferred annuities
A deferred annuity is an annuity under which the first payment occurs at some specified future time. Easily we can obtian

$${_{u|}}\ddot{a}_{x} + \ddot{a}_{x:\overline{u}|} = \ddot{a}_x \\
{_{u|}}\bar{a}_{x} + \bar{a}_{x:\overline{u}|} = \bar{a}_x
$$

And since actually ${_{u|}}\ddot{a}_{x} = \upsilon^u \cdot {_{u}}p_{x} + \upsilon^{u+1} \cdot {_{u+1}}p_{x} + \upsilon^{u+2} \cdot {_{u+2}}p_{x} + \cdots$, then we also have

$${_{u|}}\ddot{a}_{x} = \upsilon^u \cdot {_{u}}p_{x} \cdot \ddot{a}_{x+u}={_{u}}E_x \cdot \ddot{a}_{x+u}$$

And same for $1/m\text{thly}$ payment

$${_{u|}}\ddot{a}_{x}^{(m)} = {_{u}}E_x \cdot \ddot{a}_{x+u}^{(m)}$$

Then with these we can derive the formula of term annuity, whole life annuity and whole life annuity for an older

$$\ddot{a}_{x:\overline{n}|} = \ddot{a}_x - \upsilon^n \cdot {_{n}}p_{x} \cdot \ddot{a}_{x+n}\\
\ddot{a}_{x:\overline{n}|}^{(m)} = \ddot{a}_x^{(m)} - \upsilon^n \cdot {_{n}}p_{x} \cdot \ddot{a}_{x+n}^{(m)}$$

Finally we need the deferred term annuity

$$\begin{align}
{_{u|}}\ddot{a}_{x:\overline{n}|} &= \sum_{t=u}^{n+u-1} \upsilon^t \cdot {_{t}}p_{x} \\
&= \upsilon^u \cdot {_{u}}p_{x} \sum_{t=0}^{n-1} \upsilon^t\cdot {_{t}}p_{x+u} \\
&= {_{u}}E_x \cdot \ddot{a}_{x+u:\overline{n}|} \\
&= \ddot{a}_{x:\overline{u+n}\d{|}} - \ddot{a}_{x:\overline{u}|}
\end{align}
$$

So that we can do decomposiont: $\bar{a}_{x:\overline{n}|} = \sum\limits_{u=0}^{n-1} {_{u|}}\ddot{a}_{x:\overline{1}\d{|}}$

## Guaranteed annuities
In this case, pension annuity is guaranteed to be paid for some period even if the life dies before the end of the period. First we can write the present value random variable:

$$\begin{align}
Y &= \begin{cases}
\ddot{a}_{\overline{n}|} & \text{ if } K_x \leq n-1 \\ 
\ddot{a}_{\overline{K_x + 1}\d{|}} & \text{ if } K_x \geq n 
\end{cases} \\[0.7em]
&= \ddot{a}_{\overline{n}|} + \begin{cases}
0 & \text{ if } K_x \leq n-1 \\ 
\ddot{a}_{\overline{K_x + 1}\d{|}} - \ddot{a}_{\overline{n}|} & \text{ if } K_x \geq n 
\end{cases} \\[0.7em]
&= \ddot{a}_{\overline{n}|} + Y^*
\end{align}
$$

The the EPV is

$$\ddot{a}_{\overline{x:\overline{n}|}} = \EE{Y} = \ddot{a}_{\overline{n}|} + \EE{Y^*} = \ddot{a}_{\overline{n}|} + {_{n|}}\ddot{a}_x = \ddot{a}_{\overline{n}|} + {_{n}}E_{x} \cdot \ddot{a}_{x+n}$$

Similar for $1/m\text{thly}$ payment

$$\ddot{a}_{\overline{x:\overline{n}|}}^{(m)} = \ddot{a}_{\overline{n}|}^{(m)} + {_{n}}E_{x} \cdot \ddot{a}_{x+n}^{(m)} $$

## Increasing annuities
###  Arithmetically increasing annuities
This increasing annuity-due where the amount of the annuity is $t$ at time $t+1$, provided that $(x)$ survive time $t$. Then the EPV is 

$$\left( I \ddot{a} \right)_x = \sum_{t=0}^{\infty} \upsilon^t \cdot \left( t + 1 \right) \cdot {_{t}}p_{x}$$

Similarly, for the term 

$$\left( I \ddot{a} \right)_{x:\overline{n}|} = \sum_{t=0}^{n-1} \upsilon^t \cdot \left( t + 1 \right) \cdot {_{t}}p_{x}$$

Finally is the continous case, when the annuity is payable continuously with the payments increasing by $1$ at each end of the year. By analogy, we have the formula:

$$\left( I \bar{a} \right)_{x:\overline{n}|} = \sum_{m=0}^{n-1} \left( m + 1 \right) \cdot {_{m|}}\bar{a}_{x:\overline{1}\d{|}}$$

We also have the notation for the annuity continuousely payable with the rate of payment at time $t>0$ is $t$. We have its EPV is

$$\left( \bar{I} \bar{a} \right)_{x:\overline{n}|} = \int_{0}^{n} t\cdot e^{-\delta t} \cdot {_{t}}p_{x} \; \mathrm{d} t$$

### Geometrically increasing annuities
Consider an annuity-due with annual payments where the amount of the annuity is $( 1 + j )\times t$ at times $t = 0, 1, 2, \dots , n − 1$ provided that $( x )$ is alive at time $t$.

$$\sum_{t=0}^{n-1} (1+j)^t \cdot \upsilon^t \cdot {_{t}}p_{x}:= \ddot{a}_{x:\overline{n}|i^*}$$

where $\ddot{a}_{x:\overline{n}|i^*}$ is the EPV of a term annuity-due evaluated at interest rate $i^*$ where $i^* = \ffrac{1+i} {1+j} -1 = \ffrac{i-j} {1+j} $

## Evaluating annuity functions
### Recursions
We first need to assume a limiting age, $\omega$, so that $q_{\omega-1} = 1$, $\ddot{a}_{\omega-1} = 1$, then we have

$$\begin{align}
\ddot{a}_x &= 1 + \upsilon \cdot p_x + \upsilon^2 \cdot {_{2}}p_{x} + \upsilon^3 \cdot {_{3}}p_{x} + \cdots \\
&= 1 + \upsilon \cdot p_x \cdot \left( 1 + \upsilon \cdot p_{x+1} + \upsilon^2 \cdot {_{2}}p_{x+1} + \cdots \right) \\
&= 1 + \upsilon \cdot p_x \cdot \ddot{a}_{x+1}
\end{align}$$

and same for $1/m\text{thly}$, we first assume that $\ddot{a}_{\omega-1/m} = \ffrac{1} {m}$, then

$$\ddot{a}_{x}^{(m)} = \frac{1} {m} + \upsilon^{\frac{1} {m}} \cdot {_{\frac{1} {m}}}p_{x} \cdot \ddot{a}_{x + \frac{1} {m}}^{(m)}$$

### Applying the UDD assumption
Since

$$A_x^{(m)} = \frac{i} {i^{(m)}}A_x , \bar{A}_x = \frac{i} {\delta} A_x \;\; \bigg.\bigg\| \;\; \ddot{a}_x = \frac{1-A_x} {d}, \ddot{a}_{x}^{(m)} = \frac{1-A_{x}^{(m)}} {d^{(m)}}, \bar{a}_x = \frac{1-\bar{A}_x} {\delta}$$

We have

$$ \ddot{a}_{x}^{(m)} = \frac{1-\ffrac{i} {i^{(m)}}A_x} {d^{(m)}} = \frac{i^{(m)} - iA_x} {i^{(m)}d^{(m)}} = \frac{i^{(m)} - i(1-d\ddot{a}_x)} {i^{(m)}d^{(m)}} = \frac{id} {i^{(m)}d^{(m)}} \ddot{a}_x - \frac{i - i^{(m)}} {i^{(m)}d^{(m)}}:=\alpha(m)\ddot{a}_x - \beta(m)$$

and when $m \to \infty$

$$\bar{a}_x = \frac{id} {\delta^2} \ddot{a}_x - \frac{i-\delta} {\delta^2}$$

finally for term annuity

$$\begin{align}
\ddot{a}_{x:\overline{n}|}^{(m)} &= \ddot{a}_{x}^{(m)} - {_{n|}}\ddot{a}_{x+n}^{(m)}\\
&= \ddot{a}_{x}^{(m)} - \upsilon^n \cdot {_{n}}p_{x}\cdot\ddot{a}_{x+n}^{(m)}\\
&= \left[ \alpha(m) \ddot{a}_x - \beta(m)\right] - \upsilon^n \cdot {_{n}}p_{x} \left[ \alpha(m) - \ddot{a}_{x+n} - \beta(m) \right]\\
&= \alpha(m)\left[ \ddot{a}_x - \upsilon^n \cdot {_{n}}p_{x} \cdot \ddot{a}_{x+n} \right] - \beta(m) \left[ 1 - \upsilon^n \cdot {_{n}}p_{x} \right] \\
&= \alpha(m) \ddot{a}_{x:\overline{n}|} - \beta(m) \left[ 1 - \upsilon^n \cdot {_{n}}p_{x} \right]
\end{align}$$

And we can also **simplified** these formula to

$$
\ddot{a}_x^{(m)} \approx \ddot{a}_x - \frac{m-1} {2m}\\
\ddot{a}_{x:\overline{n}|}^{(m)} \approx \ddot{a}_{x:\overline{n}|} - \frac{m-1} {2m} \left[ 1 - \upsilon^n \cdot {_{n}}p_{x} \right]
$$

###  Woolhouse’s formula
As for function $g(t)$, the formula is written as

$$\int_{0}^{\infty}  g(t) \; \mathrm{d}t = h\sum_{k=0}^{\infty} g(kh) - \frac{h} {0!2!} g(0) + \frac{h^2} {2!3!} g'(0) - \frac{h^4} {3!5!} g''(0) + \cdots$$

And we apply the function $g(t) = {_{t}}p_{x} \cdot \upsilon^t$. Since $g(0) = 1$, $\lim\limits_{t\to\infty}g(t) = 0$ and for $g'(t)$ we have

$$\begin{align}
g'(t) &= {_{t}}p_{x} \cdot \frac{\mathrm{d}} {\mathrm{d}t} \upsilon^t + \upsilon^t \cdot \frac{\mathrm{d}} {\mathrm{d}t} {_{t}}p_{x} \\
&= -{_{t}}p_{x} \cdot \delta e^{-\delta t} - \upsilon^t \cdot {_{t}}p_{x} \cdot \mu_{x+t}\\[1em]
\Longrightarrow \; g'(0) &= -(\delta +\mu_{x})
\end{align}$$

First we have $h=1$, and we can get $\sum\limits_{k=0}^{\infty} g(k) -\ffrac{1} {2} + \ffrac{1} {12}g'(0) = \ddot{a}_x -\ffrac{1} {2} - \ffrac{1} {12}\left(\delta + \mu_{x}\right)$. Then we have $h=\ffrac{1} {m}$, still we ignore the higher order derivatives, then the result would be 

$$\begin{align}
\frac{1} {m} \sum_{k=0}^{\infty} g(k/m) -\frac{1} {2m} + \frac{1} {12m^2} g'(0) &= \frac{1} {m} \sum_{k=0}^{\infty} \upsilon^{k/m}\cdot {_{\frac{k} {m}}}p_{x} -\frac{1} {2m} + \frac{1} {12m^2} (\delta + \mu_x) \\
&= \ddot{a}_{x}^{(m)} - \frac{1} {2m} + \frac{1} {12m^2} (\delta + \mu_x)
\end{align}$$

We can see that these two will approach $\bar{a}_x$, so that actually

$$\ddot{a}_{x}^{(m)} - \frac{1} {2m} + \frac{1} {12m^2} (\delta + \mu_x) \approx \ddot{a}_x -\ffrac{1} {2} - \ffrac{1} {12}\left(\delta + \mu_{x}\right) \Rightarrow \boxed{ \ddot{a}_{x}^{(m)} \approx \ddot{a}_x - \ffrac{m-1} {2m} - \ffrac{m^2 - 1} {12m^2} (\delta + \mu_x)}$$

Similar for term annuity, we have

$$\begin{align}
\ddot{a}_{x:\overline{n}|}^{(m)} &= \ddot{a}_{x}^{(m)} - \upsilon^n \cdot {_{n}}p_{x} \cdot \ddot{a}_{x+n}^{(m)} \\
&\approx \ddot{a}_x - \ffrac{m-1} {2m} - \ffrac{m^2 - 1} {12m^2} (\delta + \mu_x) - \upsilon^n \cdot {_{n}}p_{x} \cdot \left( \ddot{a}_{x+n} - \ffrac{m-1} {2m} - \ffrac{m^2 - 1} {12m^2} (\delta + \mu_{x+n}) \right) \\
&= \ddot{a}_{x:\overline{n}|} - \frac{m-1} {2m} \left( 1 - \upsilon^n \cdot {_{n}}p_{x} \right) - \frac{m^2 - 1} {12m^2} \left( \delta + \mu_x - \upsilon^n \cdot {_{n}}p_{x} \cdot (\delta + \mu_{x+n}) \right)
\end{align}
$$

At last, is the continuous case, we have

$$
\bar{a}_x \approx \ddot{a}_x - \frac{1} {2} - \frac{1} {12} (\delta + \mu_{x+n}) \\
\bar{a}_{x:\overline{n}|} \approx \underset{\textbf{A}}{\underbrace{\ddot{a}_{x:\overline{n}|}}} - \underset{\textbf{B}}{\underbrace{\frac{1} {2} \left( 1 - \upsilon^n \cdot {_{n}}p_{x} \right)}} - \underset{\textbf{C}}{\underbrace{\frac{1} {12} \left( \delta + \mu_x - \upsilon^n \cdot {_{n}}p_{x} \cdot (\delta + \mu_{x+n}) \right)}}
$$

Sometimes we don't have $\mu_{x}$ and it's not wise to just omit the term $\textbf{C}$ for sometimes that will lead to poor approximation. And to solve this we can have another way to estimate that

$${_{2}}p_{x-1} = \exp\left\{ -\int_{x-1}^{x+1} \mu_s \; \mathrm{d}s \right\} \approx \exp\left\{ -2\mu_x \right\} \longrightarrow \mu_x \approx \frac{1} {2} \big( \log{\left(p_{x-1}\right)} + \log{\left(p_{x+1}\right)} \big)$$

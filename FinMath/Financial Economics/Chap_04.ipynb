### Redundant Security

When $X$ is not full rank, we might pick some securities out, say $j$, then we have

$$X_{\backslash\, j} = \begin{bmatrix}
x_{1,1} & \cdots & x_{1,j-1} & x_{1,j+1} & \cdots & x_{1,N} \\ 
\vdots & \ddots & \vdots & \vdots & \ddots & \vdots \\ 
x_{\omega,1} & \cdots & x_{\omega,j-1} & x_{\omega,j+1} & \cdots & x_{\omega,N} \\ 
\vdots & \ddots & \vdots & \vdots & \ddots & \vdots \\ 
x_{\Omega,1} & \cdots & x_{\Omega,j-1} & x_{\Omega,j+1} & \cdots & x_{\Omega,N} \\
\end{bmatrix}$$

here $x_j$ can be written as a linear combination of other payoff vectors. It's easy to verify that same payoff can be constructed without $x_j$ and that's why we call it ***redundant security***. In a **frictionless market**, its existence makes no sense to this market and we can equate $X$ and $X_{\backslash\, j}$.

However in real market there is friction. Eventhough some of the payoff vectors are linear dependent, to duplicate the payoff still require some extra cost. Thus we need to think about the allocation in market equilibrium均衡配置.

### Another Expression for Security Market
Now we suppose that $X$ is a full rank matrix. Since there're only $\Omega$ states, we have $N \leq \Omega$ and thus $\text{rank}\P{X} = \min\CB{N ,\Omega} = N$.

We then ues $X$ to construct $N$ independent portfolio, denoted as $\theta_1, \theta_2, \dots, \theta_N$. Let $H \equiv \SB{\theta_1, \theta_2,\dots,\theta_N}$, a $N \times N$ full rank matrix. Then

$$X_{\theta} = XH$$

Since $H$ is full rank and $\text{rank}\P{XHH^{-1}}\leq \text{rank}\P{XH} \leq \text{rank}\P{X}$ and $X$ is also full rank, we have $\text{rank}\P{XH} = \text{rank}\P{X} = N$. Thus $X_{\theta}$ is full rank.

Now we can use $X\theta_1, X\theta_2,\dots,X\theta_N$ to generate the original payoff vector by $X = X_\theta H^{-1}$. So actually the $n$-th column of $H^{-1}$, a combination of $\theta_1,\theta_2,\dots,\theta_N$ replicate the $n$-th security in the original market. More than that, we can prove that for any combination of securities in the original market $\theta$, it can be replicated by the combination of $\theta_1,\theta_2,\dots,\theta_N$, by

$$X\theta = XHH^{-1}\theta = X_\theta\P{H^{-1}\theta}$$

So here's the conclusion: In the frictionless market, $\theta_1,\theta_2,\dots,\theta_N$ can generate an equivalent market structure.

### Span

Now we consider the case when $N = \Omega = \text{rank}\P{X}$, when $X$ is an invertible matrix with rank $\Omega$. Thus we can replicate any securities in A-D market.

Consider a compound security $\theta$, with payoff vector $X\theta$. Let $1_\omega$ be the $\Omega\times1$ where all elements are $0$ except for a $1$ in the $\omega$ space. And they can generate any kind of paoffs, and the market build on these is **complete market**.

$Theorem.4$

The market is complete $iff$ the number of securities with independent payoff, is same with, the number of possible state in period of time $1$.

And we say, under this condition, the uncertainty in the economy can be ragarded as a ***span*** of the securities.

## Arbitrage

Denote the price vector for the securities, $S = \SB{S_1;S_2;\cdots;S_N}$, and payoff matrix $X$. How to find $S$? Or more specifically, the relationship between $S$ and $X$. We call the mapping from $X$ to $S$, the ***asset pricing relation*** or ***asset pricing model***.

Here, we consider a portfolio $\theta = \SB{\theta_1;\theta_2;\cdots;\theta_N}$. Its price at period of time $0$ is $\Tran S \theta$, and payoff at period of time $1$ is $X\theta$. Up to now, $\theta \in \mathbb{R}^{N}$. 

One more thing to mention here, we call a negative payoff $\theta_n$ or $X\theta$, a ***liability***. ANd the ***limited liability*** means that $X\theta \geq 0$.

$Def$ Arbitrage

An ***arbitrage*** is a portfolio that satisfies the following conditions:

- $\Tran S \theta \leq 0$
- $X\theta \geq 0$
- At least one of the above two inequalities hold true strictly.

Generally there're three categories.

1. $\Tran S \theta < 0$ and $X \theta = 0$
2. $\Tran S \theta = 0$ and $X \theta > 0$
3. $\Tran S \theta < 0$ and $X \theta > 0$

$Remark$

>$X \theta$ is a vector and we've defined the meaning of $=,>$ in that case! $X \theta > 0$ means the elements in $X \theta$ is always greater or equal to $0$, but at least one is always greater than $0$.

***

$Remark$

>Why there's the opportunity to arbitrage? From an example from textbook: the asset with same payoff are traded not in identical price! So to eliminate the chance to arbitrage, we should take good care of pricing given the payoff matrix.

## Principle of No-Arbitrage

$Theorem.2$ 

There's no chance to arbitrage in market equilibrium.

$Proof$

>It's true from the first glance because when there really are, agents will try to arbitrage and expand their wealth. Here's the formal proof, by contradictory. 
>
>Suppose in equilibrium market, the allocation is $\CB{c_k, k = 1,2,\dots,K}$, the security prices is $S$, and the payoff matrix is $X$. If there's a arbitrage $\theta$, for any agent $k$, he can have more consumption with no cost since $c_k + \SB{-\Tran S \theta; X\theta} > c_k$. Then by the **instatiability** $Axiom$, $c_k + \SB{-\Tran S \theta; X\theta} \succ c_k$ which break the market equilibrium.

$Remark$

>One agent is enough to prove it wrong that there're chances to arbitrage in market equilibrium.

$Def$ Principle of No-Arbitrage

To arbitrage, is not allowed to happen in the security market.

$Remark$

>This principle also depends on two assumptions we've mentioned before
>
>1. At least one agents in the market have the **instatiability**
>2. Frictionless market

## Fundamental Theorem of Asset Pricing

Now we denote the mapping from $X$ to $S$, the $V\P{x} = S$. Here the operator $V\P{\cdot}$ is called the ***pricing operator*** or the ***valuation operator***. Here're some of its properties.

$Theorem.3$ One price rule

Two securities (or portfolio) with identical payoff, share the same price: $x = y \Rightarrow V\P{x} = V\P{y}$

$Corollary$

$0 = V\P{0}$

$Theorem.4$

$x > 0 \Rightarrow V\P{x} > 0$

$Theorem.5$

$x_1 \geq x_2 \Rightarrow V\P{x_1} \geq V\P{x_2}$, or we say that $V\P{\cdot}$ is an increasing operator.

These three theorems can be easily obtained by transforming from the non-existence of three categories of arbitrage.

$Theorem.6$

In a frictionless market, the **pricing operator** is an *increasing*, *linear* operator meaning that $\forall a,b$, and three securities with payoff $x,y,z = ax+by$, we have $V\P{z} = aV\P{x} + bV\P{y}$. And since $V\P{0} = 0$, the **pricing operator** can be written in the form of $V\P{x} = \Tran\phi x$, here $\phi$ is a $\Omega \times 1$ positive vector, *NOT* the state price vector.

$Theorem.7$ Fundamental Theorem of Asset Pricing

There's no arbitrage in the security market $iff$ $\exists\,\phi \gg 0$ $s.t.$ $S = \Tran{\P{\Tran\phi X}}$

$Proof$

$\bspace$Sufficiency$\Leftarrow)$:

>For this part it's kinda easy. Suppose the equation is true in the market, then $V\P{0} = \Tran\phi 0 = 0$; also for $x > 0$, since the existence of a $\phi \gg 0$, we also have $V\P{x} > 0$, thus we have excluded the chance to arbitrage.

$\bspace$Necessity$\Rightarrow)$:

>For this part it's kinda hard (you can omit this part). First we need to introduce a $Lemma.1$, the **Stiemke** lemma:

$Lemma.1$ Stiemke

$m,n\in \mathbb{R}$, consider $X_{m \times n}$ and $\phi,\theta,S \in \mathbb{R}^n$. Then set $\CB{\theta: \SB{-\Tran S\theta;X\theta}>0}$ is an empty set $iff$ $\exists\,\phi \gg 0$ and $S = \Tran{\P{\Tran\phi X}}$.

>Then with his $Lemma$, we let $m = \Omega, n = N$, the conclusion is proved.

$Remark$

>$Theorem.7$ implies that there's a state price vector $\phi \gg 0$ that can be used for asset pricing for all securites. There might be more one possible here~ by the way.

$Theorem.8$

In a **complete security market** (If any consumption plan $c$ 都可以通过有限成本的可交易证券的组合来融资), there's only ONE possible state price vector.

$Proof$

> Let $S$ be the price vector of the $\Omega$ number of securities, and $\theta_\omega$ is the portfolio that can be used to replicate the state-$\omega$ contingent claim. Then, the state price of state $\omega$ is determined by $\phi_\omega = \Tran S \theta_\omega$.

## Risk-Neutral Pricing and Martingale

Suppose there's a ***riskless security*** with a certain payoff $1$ in period of time $1$, and thus its state price vector is $x_1 = \SB{1;1;\cdots;1} = \iota$ and price $S_1$. Easily seen that $S_1 = \Tran\phi \iota = \sum \phi_\omega$.

Define the ***rate of return*** as the pure payoff from one unit of **riskless security**, or we call it the ***riskless return rate***, denoted as $r_F$, so that

$$S_1 \P{1+ r_F} = 1, r_F = \ffrac{1-S_1} {S_1}$$

**Interest rate** is also known as the ***time value of money*** since it represent the payoff from consuming one unit of asset in exchange of some future asset. So here's anothre form for $S_1$: $S_1 = \ffrac{1} {1+r_F}$, a form that expressing the price using a linear function of of its payoff. Generally known as ***discounting***折现. And $r_F$ is the ***discounting rate***.

***

For other securities, say with payoff vector $x_n = \SB{x_{1,n}; x_{2,n}; \cdots; x_{\Omega, n}}$. From the pricing formula, we have

$$S_n = \Tran\phi x_n = \sum_{\omega \in \Omega} \phi_\omega x_{\omega,n}$$

Here $n = 2,3,\dots,N$. Then if we define $q_\omega \equiv \ffrac{\phi_\omega} {\d{\sum_{\omega'} \phi_{\omega'}}}$ which is *POSITIVE* obviously and $\sum q_\omega = 1$, here $\omega = 1,2,\dots,\Omega$.

It's a MEASURE! $Q \equiv \CB{q_\omega, \omega \in \Omega}$ can be regarded as a probability measure on $\Omega$. And thus we rewrite the pricing formula:

$$
\newcommand{\Exp} {\mathrm{E}}
S_n = S_1 \sum_{\omega} q_\omega x_{\omega,n} = \ffrac{\Exp^{Q}\SB{\tilde x_n}} {1+r_F},n=1,2,\dots,N$$

Here $\tilde x_n$ is the random variable for the payoff of security $n$ and $\Exp^{Q}\SB{\cdot}$ is the expectation under measure $Q$. More surprisingly, this formula can be interpreted as:

>The price of a security, $S_n$ is just its expectated payoff under measure $Q$ discouted by riskfree rate $r_F$. And it's true for not just the risk free security.

We call this formula the ***risk-neutral pricing*** formula and $Q$ is the risk-neutral measure.

$Remark$

$Q$ is not the probability of each state (which is just $1/\omega$ in the textbook), but defined by the state prices $\phi_\omega$.

**e.g.** Pricing securities

At period $1$, there're two state of same possibility $a$ and $b$. Also, two securities $1$ and $2$ with price and payoff:

$$1: 1\to\begin{cases}
1 \\
1
\end{cases}\bspace 2: 0.5 \to \begin{cases}
2 \\
0
\end{cases}$$

Pricing them!

>Security $1$ is the risk free security. So that the risk-free rate $r_F = \ffrac{1-1} {1} = 0$. Then the equation set for the state price vector:
>
>$$\begin{cases}
\phi_a + \phi_b = 1 \\
2 \times \phi_a = 0.5
\end{cases} \Longrightarrow \begin{cases}
\phi_a = 0.25 \\
\phi_b = 0.75
\end{cases}$$
>
>And the $Q$ measure now is $\CB{0.24,0.75}$. Thus:
>
>$$\ffrac{\Exp^Q\SB{\tilde x_1}} {1+0} = \ffrac{0.25 \times 1 + 0.75 \times 1} {1} = 1 \\
\ffrac{\Exp^Q\SB{\tilde x_2}} {1+0} = \ffrac{0.25 \times 2 + 0.75 \times 0} {1} = 0.5$$

***

Also, since in this market $S_1 = \ffrac{1} {1+r_F}$ and $\tilde x_1 = 1$, we can rewrite the pricing formula:

$$\ffrac{S_n} {S_1} = \Exp^Q\SB{\ffrac{\tilde x_n} {\tilde x_1}}$$

The left hand side, $\ffrac{S_n} {S_1}$ can be regarded as the price of $S_n$ using $S_1$ as measuring unit. So shorten the notation, we let $\hat S_{n,t} = \ffrac{S_{n,t}} {S_{1,t}}$, $t=0,1$ as the period. After that, we have

$$\hat S_{n,t}  = \Exp^Q\SB{\hat S_{n,t+1}}, t=0$$

What's more, if we write the price $S_t$ as a series, then it's a stochastic process.

$Def$

For a stochastic process $z_1,z_2,\dots$, if $z_t=\Exp_t\SB{z_{t+1}}$ always holds, we call this process a ***martingale***.

Then, let the riskless security price $S_1$ as the unit, the security price is a **martingale** under the measure of $Q$, the risk-neutral measure. And thus $Q$ is called the ***equivalent martingale measure***. Here the **equivalent** is to say that the probability measure, $Q$ and $P$, are equivalent if their set of events with $0$ probability are just the same. And here $p\gg 0$ and $q\gg 0$ which result in the empty $0$ probability set.

## Summary

- 无摩擦市场(Frictionless Market, Chap_02)中，$N$个线性独立的证券组合可以为市场结构(market structure)提供一个等价描述
- 当且仅当具有独立支付（不相关）的证券数等于状态数时证券市场是完全(complete)的
- 如果组合$\theta$满足以下条件：
    - $\Tran S\theta \leq 0$
    - $X\theta \geq 0$
    - 以上至少一个严格成立  
那么称$\theta$为套利(arbitrage)或者套利机会。套利**不依赖**于任何私有信息。在无摩擦的假设下，任何人都可以利用这个机会
- 无套利原理(Principle of No-Arbitrage)：市场均衡(Market Equilibrium)下无套利机会
- 资产定价基本定理：当且仅当存在状态价格向量(State Price Vector)$\phi\gg0$使得$S = \Tran X\phi$时，证券市场中不存在套利机会
- 在一个完全的证券市场中，状态价格向量是唯一的。
- 由资产定价基本定理(Fundamental Theorem of Asset Pricing)，资产定价公式为$S_n = \ffrac{\Exp^Q\SB{\tilde X_n}} {1+r_F}$，也称为风险定价公式。$Q$是依据状态价格向量定义的风险中性概率测度。
- 以债券（多用无风险债券）价格为计量单位，证券价格在风险中性测度(Risk-Neutral Measure)$Q$下是鞅(Martingale)，即有$\hat S_{n,t} = \Exp^Q\SB{\hat S_{n,t+1}}$。此时$Q$也称为等价鞅测度(Equivalent Martingale Measure)。

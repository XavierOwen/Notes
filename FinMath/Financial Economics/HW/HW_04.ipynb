## Question 13.1

$\P 1$

$\bspace$Suppose that the price vector of the stocks is $S$ and for agent $k$ with portfolio $\theta_k$, we have his demand for stocks:

$$\begin{array}{cl}

\end{align}
\end{array}$$

$\bspace$Then to maximize $\Exp{-e^{-\tilde w_{\void_k}}}$ is equivalent to minimize 

$$\begin{align}
&\ffrac{1}{2}\Var{\tilde w_k} - \Exp{\tilde w_k}\\
=\;& \ffrac{1}{2}\Var{ \P{\Tran{\theta_k} + y_k \Tran\iota}\tilde v + y_k\tilde\varepsilon} - \P{w_k + \Tran{\theta_k} \bar v + \Tran{\P{\bar \theta_k - \theta_k }}S + y_k \Tran\iota\bar v}\\
=\;& \ffrac{1}{2}\Tran{\P{\theta_k + y_k\iota}}\Sigma\P{\theta_k + y_k\iota} + \ffrac{\sigma^2y_k^2}{2} - w_k - \Tran{\theta_k} \bar v - \Tran{\P{\bar \theta_k - \theta_k }}S - y_k \Tran\iota\bar v
\end{align}$$

$\bspace$Take the derivative on $\theta_k$, since $\Sigma = \Sigma^{-1}$, we have

$$0 = \P{\Tran\theta_k + y_k \Tran\iota}\Sigma + \Tran{\P{S - \bar v}} \Rightarrow \theta_k = \Sigma^{-1}\P{\bar v - S} - y_k\iota$$

$\P 2$

$\bspace$To reach balance, $\sum\theta_k = \sum\bar\theta_k = \iota$ so that

$$K\cdot \Sigma^{-1}\P{\bar v - S} - \P{\sum y_k}\iota = \iota \Rightarrow S = \bar v - \ffrac{\P{\sum y_k +1}\Sigma \iota}{K}$$

$\P 3$

$\bspace$We now prove under this context $\bar r_n -r_F = \beta_{n,M}\P{\bar r_M - r_F}$ holds, where $\bar r_n$ is the mean value of $\tilde r_n = {\ffrac{\tilde v_n}{S_n}}-1$, $\bar r_M$ is the mean value of $\tilde r_M = {\ffrac{\Tran{\iota} \tilde v}{w_M}}-1$, for security $n$. $w_M = \Tran\iota S$. Then

$$\begin{align}
\bar r_n - r_F &= \bar r_n = \ffrac{\bar v_n - S_n}{S_n}\\
&= \ffrac{\P{\sum y_k +1}\P{\Sigma \iota}_n}{K S_n}\\
\bar r_M - r_F &= \bar r_M\\
&= \ffrac{\Tran\iota \bar v}{w_M} - 1
\end{align}$$

$\bspace$Then the $\beta_{n,M} = \ffrac{\Cov{\tilde r_M, \tilde r_n}}{\Var{\tilde r_M}} = \Big.\ffrac{\Cov{\Tran\iota \tilde v,{\tilde v}_n}}{w_M S_n}\Big/\ffrac{\Var{\Tran\iota\tilde v}}{w_M^2} = \ffrac{w_M\P{\Sigma\iota}_n}{S_n \Tran\iota\Sigma\iota}$. Now we check

$$\begin{align}

&= \ffrac{1}{w_M}\P{\Tran\iota \bar v - \Tran\iota S} = \ffrac{\Tran\iota \bar v}{w_M} - 1
\end{align}$$

$\P 4$

&= \ffrac{\Cov{\ffrac{\tilde v_n}{\P{p_a}_n}-1,\ffrac{\Tran\iota \P{\tilde v_a + \tilde v_b}}{\Tran\iota\P{p_a+p_b}} - 1}}{\Var{\ffrac{\Tran\iota \P{\tilde v_a + \tilde v_b}}{\Tran\iota\P{p_a+p_b}} - 1}}\\
&= \ffrac{\ffrac{1}{\P{p_a}_n\Tran\iota\P{p_a+p_b}}}{\ffrac{1}{\P{\Tran\iota\P{p_a+p_b}}^2}} \times \ffrac{\Cov{\tilde v_n,\Tran\iota\P{p_a+p_b}}}{\Var{\Tran\iota\P{p_a+p_b}}}\\
&= \ffrac{\Tran\iota\ffrac{1}{1+r_F}\P{ 2\bar v - \ffrac{\alpha}{K_a+K_b}\P{4\sigma_w^2I+\Sigma_a+\Sigma_b}\iota}}{\ffrac{1}{1+r_F}\P{\bar v_n - \ffrac{\alpha}{K_a + K_b}\P{\Sigma_a\iota}_n - \ffrac{2\alpha}{K_a + K_b}\sigma_w^2}}\cdot\ffrac{2\sigma_w^2+\P{\Sigma_a\iota}_n}{\Tran\iota\P{4\sigma_w^2I+\Sigma_a+\Sigma_b}\iota} \\
&= \ffrac{2\Tran\iota\bar v - \ffrac{\alpha}{K_a+K_b}\Tran\iota\P{4\sigma_w^2I+\Sigma_a+\Sigma_b}\iota}{\bar v_n - \ffrac{\alpha}{K_a + K_b}\P{\Sigma_a\iota}_n - \ffrac{2\alpha}{K_a + K_b}\sigma_w^2} \cdot \ffrac{\ffrac{\alpha}{K_a + K_b}\P{\Sigma_a\iota}_n + \ffrac{2\alpha}{K_a + K_b}\sigma_w^2}{\ffrac{\alpha}{K_a+K_b}\Tran\iota\P{4\sigma_w^2I+\Sigma_a+\Sigma_b}\iota}\\
&= \ffrac{\bar r_n - r_F}{\bar r_{wm} - r_F}
\end{align}$$

$\bspace$That's it. And similar for the $m$th asset in country $b$. Thus CAPM model holds. 委屈。。。

## Question 13.5
$\P 1$

$\bspace$To eliminate the chance to arbitrage, $\forall\: \omega$, $\phi_\omega > 0$ and $\sum\phi_\omega \leq 1$.

$\P 2$

$\bspace$For state $\omega$, the interest rate is $\tilde r = \ffrac{w_\omega}{\phi_\omega}-1$. Then the mean value and the variance are 

$$\bar r = \sum_{\omega}\ffrac{\pi_\omega w_\omega}{\phi_\omega}-1,\bspace\Var{\tilde r} = \sum_\omega \pi_\omega\P{\ffrac{w_\omega}{\phi_\omega}}^2 - \P{\sum_{\omega}\ffrac{\pi_\omega w_\omega}{\phi_\omega}}^2$$

$\bspace$And the covariance of portfolio $w$ and $w'$ is

$$\Cov{\tilde r,\tilde r'} = \Exp{\tilde r\cdot\tilde r'} - \bar r \cdot\bar r' = \sum_\omega \pi_\omega\ffrac{w_\omega w_\omega'}{\phi_\omega^2} - \P{\sum_\omega \pi_\omega\ffrac{w_\omega}{\phi_\omega}}\P{\sum_\omega \pi_\omega\ffrac{w_\omega'}{\phi_\omega}}$$

$\P3$

$\bspace$The payoff, for all states $\omega$, is $\ffrac{w_\omega^a}{\phi_\omega}\times1 = \ffrac{1}{\sum\phi_\omega}$. And the price of that state-$\omega$ security is $\sum w_\omega^a = 1$, thus, it's riskless portfolio.

$\bspace$To prove these two build up the frontier, consider the optimization problem

$$\begin{array}{cl}
\d{\min_{w_\omega}} & \ffrac{1}{2}\sum\nolimits_\omega {\pi_\omega} \P{\ffrac{w_\omega}{\phi_\omega}}^2\\
s.t. & \sum_\omega \pi_\omega \ffrac{w_\omega}{\phi_\omega}-1 = \bar r_p \\
& \sum_\omega w_\omega = 1
\end{array}$$

$\bspace$Then we write the corresponding Largrangian equation

$$L = \ffrac{1}{2}\sum\nolimits_\omega {\pi_\omega} \P{\ffrac{w_\omega}{\phi_\omega}}^2 + \lambda_r \P{\bar r_p +1 - \pi_\omega \ffrac{w_\omega}{\phi_\omega} } + \lambda_1 \P{1-\sum\nolimits_\omega w_\omega}$$

$\bspace$where the F.O.C. is $\pi_\omega \ffrac{w_\omega}{\phi_\omega^2} = \lambda_r \ffrac{\pi_\omega}{\phi_\omega} + \lambda_1$, which leads to

$$w_\omega = \lambda_r \phi_\omega + \lambda_1 \phi_\omega q_\omega \Rightarrow \left\{\begin{align}
\sum_{\omega}w_\omega = 1 &= \lambda_r\cdot\P{\sum\phi_\omega} + \lambda_1 \cdot\P{\sum\phi_\omega q_\omega}\\
\sum_\omega \ffrac{w_\omega}{q_\omega} = 1 + \bar r_p &= \lambda_r\cdot\P{\sum\ffrac{\phi_\omega}{q_\omega}} + \lambda_1\cdot\P{\sum\phi_\omega} = \lambda_r + \lambda_1\cdot\P{\sum\phi_\omega}
\end{align}\right.$$

$$\Rightarrow\begin{cases}
\lambda_r = \ffrac{\P{1+\bar r_p}\sum\phi_\omega q_\omega -\sum \phi_\omega}{\P{\sum\phi_\omega q_\omega} -\P{\sum \phi_\omega}^2}\\
\lambda_1 = \ffrac{\P{1+\bar r_p}\sum\phi_\omega -1}{\P{\sum\phi_\omega q_\omega} -\P{\sum \phi_\omega}^2}
\end{cases}$$

$$\begin{align}
w_\omega &= \lambda_r\phi_\omega + \ffrac{\lambda_1\phi_\omega^2}{\pi_\omega}\\
&= \lambda_r\phi_\omega +\lambda_1\phi_\omega q_\omega\\
&= \delta \cdot \ffrac{\phi_\omega}{\sum \phi_\omega} + \P{1-\delta}\cdot\ffrac{\phi_\omega q_\omega}{\sum\phi_\omega q_\omega}
\end{align}$$

$\bspace$where $\delta = \lambda_r \cdot \sum\phi_\omega = \ffrac{\P{1+\bar r_p}\sum\phi_\omega q_\omega\sum\phi_\omega -\P{\sum \phi_\omega}^2}{\P{\sum\phi_\omega q_\omega} -\P{\sum \phi_\omega}^2}$. That's it.

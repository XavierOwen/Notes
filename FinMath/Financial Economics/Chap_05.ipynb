## Properties and Boundaries of Option Price

$Theorem.1$

$c\P{S,K}$ and $p\P{S,K}$ are non-negative.

$Theorem.2$

$c$ is non-increasing on $K$ while $p$ is non-decreasing on $K$.

$Proof$

>$\forall \, K_1 > K_2, [\tilde X-K_1]_+ \leq [\tilde X-K_2]_+ \Longrightarrow c\P{S,K_1} \leq c\P{S,K_2}$

**(V)**$Theorem.3$

$c\P{S,K}$ and $p\P{S,K}$ is convex on $K$.

$Proof$

>$\forall \, K_1,K_2$ and $\alpha \in \P{0,1}$ we have
>
>$$\begin{align}
&\bspace[\tilde X -\P{\alpha K_1 +\P{1-\alpha}K_2}]_+\\
&= [\alpha\P{\tilde X - K_1} + \P{1-\alpha}\P{\tilde X - K_2}]_+ \\
&\leq \P{\alpha[\tilde X - K_1]_+} + \P{\P{1-\alpha} [\tilde X - K_2]_+}
\end{align}$$
>
>Then since $c\P{S,K} = V\P{[\tilde X - K]_+}$, a linear relation, thus $c\P{S,K}$ is a convex function on $K$.

$Theorem.4$

$\theta\geq 0$ is a portfolio from $N$ securities, the price vector $S = \SB{S_1;S_2;\cdots;S_N} >0$ and the exercise price vector: $K = \SB{K_1;K_2;\cdots;K_N} >0$. Then

$$c\P{\Tran S \theta, \Tran K\theta} \leq \sum_{i=1}^{N} \theta_i \cdot c\P{S_i,K_i},\bspace p\P{\Tran S \theta, \Tran K\theta} \leq \sum_{i=1}^{N} \theta_i \cdot p\P{S_i,K_i}$$

This is to say that the price of the option with the portfolio of assets as its underlying asset, is less than the price of the portfolio of options whose underlying asset is just one single security. (翻译的什么玩意。。。)

$Proof$

>$$\SB{\Tran{\P{\tilde X - K}}\theta}_+ = \SB{\sum_{i=1}^{N}\theta_i\cdot\P{\tilde X_i - K_i}}_+ \leq \sum_{i=1}^{N} \theta_i \cdot [\tilde X_i - K_i]_+$$

$Theorem.5$

$$S\geq c\P{S,K}$$

$Proof$

>$$\begin{cases}
\tilde X \geq [\tilde X - K]_+ \\
S = V\P{\tilde X}, c\P{S,K} = V\P{[\tilde X - K]_+}
\end{cases}\Longrightarrow S \geq c\P{S,K}$$

$Theorem.6$

If there's riskless security, with rate of return $r_F$, then

$$c\P{S,K} \geq \SB{S - \ffrac{K}{1+r_F}}_+$$

$Proof$

>Consider a portfolio like this: buy one share of stock and short sell $K$ amouts of riskless securities.
>
>Then it's current value at period $0$ is $S - \ffrac{K} {1+r_F}$ while at period $1$, the payoff is $\tilde X - K$. And since $[\tilde X - K]_+ \geq \tilde X - K$, thus $c\P{S,K} \geq S - \ffrac{K} {1+r_F}$. From $Theorem.1$, $c\P{S,K}\geq0$ thus $c\P{S,K} \geq \SB{S - \ffrac{K}{1+r_F}}_+$.

$Remark$

>Combine all theorems above we have the upper and lower boundary for $c$:
>
>$$\SB{S - \ffrac{K} {1+r_F}}_+ \leq c\P{S,K} \leq S$$

And to generate the similar result for put option, we can use the ***put-call parity平价关系***.

$Theorem.7$

Given the existence of a riskless security with interest rate $r_F$:

$$c\P{S,K} + \ffrac{K} {1+r_F} = p\P{S,K} + S$$

$Rroof$**(V)**

>Consider two possible portfolios:
>
>1. Buy one call option contract and one riskless security
>2. Buy one put option contract and one stock
>
>Comparing their payoff 
>
>$$\begin{array}{c||c|c}
& \tilde X \leq K & \tilde X > K \\ \hline
\text{Portfolio }1 & [\tilde X - K]_+ + K = K & [\tilde X - K]_+ + K = \tilde X \\ \hline
\text{Portfolio }2 & [K - \tilde X]_+ + \tilde X = K & [K - \tilde X]_+ + \tilde X = \tilde X 
\end{array}$$
>
>No difference! So their cost at peroid $0$ must be same as well, which is the preceding parity.

## American Option and Early Exercise

Since american options grant the holders more right than those from european options, thus $C\P{S,K} \geq c\P{S,K}$ and so is $P\P{S,K} \geq p\P{S,K}$. But is there the necessity to exercise early? Well, another factor, **dividend** should be taken into account, where extra payment will happen in period $0$.

### Early Exercise, no dividend payment

First notice that it's never optimal to early exercise an american call option. If so, the payoff would be $S-K$, however: $S - K \leq S - \ffrac{1} {1+r_F} \leq \SB{S- \ffrac{1} {1+r_F}}_+ \leq c\P{S,K}$. So $C = c$ always hold.

While speaking of American put option, the cose to early exercise, is that you are giving up the right to refuse to exercise in the future and the gain is to obtian $K$ right now instead of waiting longer.

$$\begin{align}
P\P{S,K} &= \max \P{K-S, p\P{S,K}} \\
&= \max\P{K-S, \ffrac{K} {1+r_F} - S + c\P{S,K}}
\end{align}$$

and there's a time when $K-S > \ffrac{K} {1+r_F} - S + c\P{S,K}$ so that it's optimal to early exercise. After a simplification, we have the condition: $\ffrac{r_F} {1+r_F} K > c\P{S,K}$

### Early Exercise, with dividend payment

Suppose that the stock paid $D$ dividend at period $0$ and the stock price after that is $S$. Then we have

$$\begin{align}
C\P{S,D,K} &= \max\P{S+D-K, c\P{S,K}} \\
C\P{S,D,K} &= \max\P{K-S-D, p\P{S,K}}
\end{align}$$

This is to say: the price of American options before the dividend payment is the max of

1. the payoff if exercise right after payment
2. the corresponding European options.

$Remark$

>Generally, the dividend will motivate the holder of American call option to early exercise and the holder of American put option to exercise later.

Besides, with its existence, the partity between the two also breaks, alright. It just changes to

$$c\P{S,K} + \ffrac{K} {1+r_F} + D = S + p\P{S,K}$$

consider two portfolio

1. hold $1$ call option with strike price $K$ and $\ffrac{K} {1+r_F} + D$ amount of riskless security
2. hold $1$ put option with strike price $K$ and $1$ share of stock

Then at period $1$, we have

$$\begin{array}{c||c|c}
& \tilde X \leq K & \tilde X > K \\ \hline
\text{Portfolio }1 & K + D\P{1+r_F} & \tilde X + D\P{1+r_F} \\ \hline
\text{Portfolio }2 & K + D\P{1+r_F} & \tilde X + D\P{1+r_F}
\end{array}$$

## Option Pricing in Complete Market

**(V)**

Undre the assumption of complete market, there's only one state price vector: $\phi_\omega, \forall \omega \in \Omega$. Let $S_\omega$ be the payoff of underlying asset under state $\omega$, then 

$$c\P{S,K} = \sum_{\omega} \phi_{\omega} \SB{S_\omega - K}_+ = \ffrac{1} {1+r_F} \Exp^Q\SB{\tilde S - K}_+$$

Here $Q$ is the risk-neutral measure. But we need further description on the stock price in period $1$. Let's have a look at the ***binomial tree model***.

$S$ is the current price. And with probability $p$, it rises to $uS$ otherwise shrinks to $dS$, with probability $1-p$. $B$ is the price of the riskless security (*Bond*, in short),then $B = \ffrac{1} {1+r_F}$.

$$S \to \begin{cases}
uS, &\text{with probability }p \\
dS, &\text{with probability }1-p 
\end{cases}, \bspace B \to \begin{cases}
1, &\text{with probability }p \\
1, &\text{with probability }1-p 
\end{cases}, \bspace$$

To eliminate the chance to arbitrage, we have $d < 1+r_F <u$. 

$Proof$

>If $d\geq 1+r_F$, we can build the portfolio: buy $1$ share using $S$, then sell $S\P{1+r_F}$ amount of riskless security and gain $\ffrac{1} {1+r_F} \times S\P{1+r_F} = S$ so that the payoff at period $0$ is $0$. Then if the stock price goes down, then we will still have a non-negative payoff in period $1$: $\SB{d-\P{1+r_F}}S \geq 0$ but if it goes up, then $\SB{u-\P{1+r_F}}S > 0$. 
>
>Similarly we can prove that $u > 1 +r_F$ must holds than finally $d < 1+r_F <u$.

Then we assume the state price vector to be $\phi = \SB{\phi_u; \phi_d} \gg 0$ and then

$$\begin{align}
S &= \phi_u \cdot u \cdot S + \phi_d \cdot d \cdot S \\
B &= \phi_u \cdot 1 + \phi_d \cdot 1 = \phi_u + \phi_d = \ffrac{1} {1+r_F}
\end{align}$$

And we can solve this equation: $\begin{cases}
\phi_u = \ffrac{1} {1+r_F}\cdot \ffrac{1+r_F-d} {u-d}\\
\phi_d = \ffrac{1} {1+r_F}\cdot \ffrac{u-\P{1+r_F}} {u-d}
\end{cases}$

Then we have the option price like

$$c = \phi_u \cdot \SB{uS-K}_+ + \phi_d \cdot \SB{dS-K}_+$$

And more than that, using the price of the stock and the security, we can price the options and other securities. This could happen because in **binomial tree model** there're only two different states and the market is complete by the stock and the security in it.

Other securities are redundant! From the principle of no arbitrage, their prices can be duplicated by that two. So for a European call option with strike price $K$. 

$$c_1 = \SB{S_1 - K}_+ = \begin{cases}
c_u \equiv \SB{uS - K}_+, &\text{if }S_1 = uS \\
c_d \equiv \SB{dS - K}_+, &\text{if }S_1 = dS
\end{cases}$$

Now consider the portfolio $\theta = \SB{\theta_S;\theta_B}$ for $S$ amount of stocks and $B$ amount of securities. And its payoff is

$$v_1 = \begin{cases}
v_u \equiv \theta_S \cdot uS + \theta_B, &\text{if } S_1 = uS \\
v_d \equiv \theta_S \cdot dS + \theta_B, &\text{if } S_1 = dS 
\end{cases}$$

Then to price the call option, we need to find the value of $\theta_S$ and $\theta_B$ so that $v_u=c_u$ and $v_d = c_d$. And the solution is 

$$\theta_S = \ffrac{c_u-c_d} {\P{u-d}S},\bspace \theta_B = \ffrac{uc_d - dc_u} {\P{c-u}}$$

And, that's same with the call option, I mean the payoff of this portfolio. For this portfolio, its price at period $0$ is $v = \theta_S S + \ffrac{\theta_B}{1+r_F}$. By the principle of no arbitrage, $v = c$:

$$c = \ffrac{1} {1+r_F} \SB{\ffrac{\P{1+r_F} -d} {u-d} c_u + \ffrac{u - \P{1+r_F}} {u-d}c_d}$$

Well, SAME with BEFORE! This method is valid!

Also we call $\theta_S$ the ***hedge ratio对冲比***, $\delta$.

And we can define the **equivalent martingale measure**, $Q$, here as

$$\begin{align}
q &= \ffrac{\phi_u} {\phi_u + \phi_d} = \ffrac{1+r_F - d} {u-d}\\
1-q &= \ffrac{\phi_d} {\phi_u + \phi_d} = \ffrac{u-\P{1+r_F}} {u-d}
\end{align}$$

and then rewrite the option pricing formula: $c = \ffrac{1} {1+r_F} \SB{q c_u + \P{1-q} c_d} = \ffrac{\Exp^Q\SB{c_1}} {1+r_F}$. We can further divide the both side with the price of riskless security $B$ and obtain the martingale:

$$\hat c_t \equiv \ffrac{c_t} {B} = \Exp^Q_t\SB{\hat c_{t+1}},\bspace \hat S_t \equiv\ffrac{S_t} {B} = \Exp^Q_t\SB{\hat S_{t+1}}$$

Here we only consider when $t=0$, cause there're only two period of time. And, also, $\hat c_t$ and $\hat S_t$ are $Q$-martingale when $Q$ is the **equivalent martingale measure**.

## Summary

- 欧式看涨（跌）期权(European Call (Put) Option)给予购买者一种权利：其可以在未来某一指定日期，以某一确定价格$K$从（向）期权出售者买入（卖出）单位股票。美式则允许期权购买者在到期日及之前随时行权
- 根据无套利原理(Principle of No-Arbitrage)，欧式看涨期权价格的上下界

$$\SB{S - \ffrac{K} {1+r_F}}_+ \leq c\P{S,K}\leq S$$

- 平价关系(Put-Call Parity)：若存在无风险证券(Riskless Security)，利率(interest rate)为$r_F$，则有

$$c\P{S,K} + \ffrac{K} {1+r_F} = p\P{S,K} + S$$

- 美式期权价不低于相应的欧式期权
    - 若无股利，美式看涨不会被提前执行，$c\P{S,K} = C\P{S,K}$
- 当有股利(Dividend)存在时
    - 股利会促使持有者提前执行看涨期权，推迟执行看跌期权
    - 平价关系变为：$c\P{S,K} + \ffrac{K} {1+r_F} + D = p\P{S,K} + S$
- 当证券市场完全时，期权可能冗余(redundant)。此时利用无套利原理，可以用二叉树模型对期权定价。当然也可能不冗余，此时根据没学的内容，期权增进市场完全性。

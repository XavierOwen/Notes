## Agents (resource and demand)

Suppose there're totally $K$ agents for $k = 1,2,\dots,K$. Two features: the resource he controls and his goals in this economy.

### Resources

1. Endowment 禀赋 (比如capital, assets and any other physical goods)
    - Denote as $e_{k,0}$ for the current period and $e_{k,1\omega}$ for the next future period with state $\omega \in \Omega$
    - Thus $e_k \equiv \SB{e_{k,0};e_{k,1}} = \SB{e_{k,0};\SB{e_{k,11},\dots, e_{k,1\Omega}}}$, $k = 1,2,\dots,K$
    - And actually for the endowment of each agent, it's an element from $\mathbb{R}^{1+\Omega}$
    - What's more, we assume it to be in $\mathbb{R}^{1+\Omega}_+$, positive quadrant.
2. Information (public but not private, as assumed, so that 信息对称，symmetric)
3. Production Technology生产技能, assumed to be $0$

### Demands

for agent $k$, denote his consumption at period $0$ (today, known) is $c_{k,0}$ and at period $1$ (future, unknown with multiple states) under state $k$, his consumption is $c_{k,1\omega}$. Thus the complete expression for consumption is:

$$c_k \equiv \SB{c_{k,0};c_{k,1}} = \SB{c_{k,0};\SB{c_{k,11},\dots, c_{k,1\Omega}}}$$

for $k = 1,2,\dots,K$

$\P{1}$ **Consumption Set**

可能的消费选择 $c = \SB{c_0,c_1}$, or a ***consumption plan***. It relies on he future state of the economy and it's end value is $\P{e_0, e_{1\omega}}$, or a ***consumption path***. All consumption plan in a set is called the ***consumption set***, $C$, normally as a subset of $\mathbb{R}^{1 + \Omega}$. Later without other explanation, $C = \mathbb{R}^{1 + \Omega}_+$.

$Def$ ***Convex set***

$A \subseteq \mathbb{R}^n$, $\forall a,b \in A$ and $\alpha \in \SB{0,1}$, if $\alpha a + \P{1-\alpha}b \in A$ we say that set $A$ is convex.

$Def$ ***Close set***

$A \subseteq \mathbb{R}^n$ and $\forall$ sequence $a_i \in A$ with limit $a$ and $a \in A$ as well, then $A$ is a close set.

$Assumption$

$C = \mathbb{R}^{1 + \Omega}_+$ is a close convex set in $\mathbb{R}^{1+\Omega}$

$\P{2}$ **preference**

For the agents, their preference on different consumption plans describe their consumption demand.

$Def$ ***preference***

**Preference** is a binary relation, $\succsim$: 

1. Completeness完备性: $\forall a,b \in C$, $a \succsim b$ or $b \succsim a$ or both are right when they are equivalent ($ a \sim b$)
    - and when $a \succsim b$ but not $b \succsim a$, we can use another denotation: $a \succ b$
2. Transtivity传递性: if $a \succsim b$ and $b \succsim c$, then $a \succsim c$. (Or rationality for all agents)

$\P{3}$ **Assumptions for preference**

$Axiom$

1. instatiability 不满足性: for two consumption plan $a$ and $b$, if $a > b$ then $a \succ b$.
2. continuity 连续性: $\forall c \in C$, set $\CB{a \in C: a \succsim c}$ and $\CB{b \in C: b \precsim c}$ are close.
3. convexity 凸性: $\forall a,b \in C$ and $\alpha \in \P{0,1}$, if $a \succ b$ then $\alpha a + \P{1-\alpha}b \succ b$
    - With this we can derive that $\CB{a \in C: a \succsim b}$ is a convex set.

$\P{4}$ **Utility Function**

$\bspace$We can assign a value (utility) to a consumption plan $c$. We call this mapping ***Utility Function***. And it satisfies that $\forall a,b \in C$, $a \succsim b$ is equivalent to $U\P{a} \geq U\P{b}$

$Def$

$\bspace$The utility function on consumption plan is a function from $C$ to $\mathbb{R}$, $U: C \to \mathbb{R}$, so that $\forall a,b \in C, U\P{a} \geq U\P{b}$ $iff$ $a \succsim b$ 

$Remark$

>The optimized choice is obviously the one with highest value of utility function.

$\bspace$We may face the question that whether there's always be a utility function well-defined, and here's a theorem for this confusion:

$Theorem.1$ (Debreu)

For a **preference** which satisfies the $Axiom.2$ defined on a **consumption set** $C$ that is *close* and *convex*, there exists a continuous utility function $U\P{\cdot}$ on $C$ so that $\forall a,b \in C, U\P{a} \geq U\P{b}$ $iff$ $a \succsim b$.

$Remark$

1. Here this preference actually satisfy: completeness, transtivity, continuity
2. The utility function obtained here is just **ordinal (序数的)**, meaning that it only gives the order of the consumption plans while the actual value of the function is insignificant since we can always make an monotonic transformation to change the value while the order remains the same.

And consequently, we rewrite the assumptions for **preference** with the form of utility function

1. instatiability 不满足性: for two consumption plan $a$ and $b$, if $a \geq b$ then $U\P{a} \geq U\P{b}$.
2. continuity 连续性: $\lim\limits_{a_n \to a}U\P{a_n} = U\P{a}$
3. convexity 凸性: if $U\P{a} > U\P{b}$, and $\alpha \in \P{0,1}$, then $U\P{\alpha \cdot a + \P{1-\alpha}\cdot b }> U\P{b}$

## Security Market

The allocation of assets is done by the trade in the financial market. The market allow the trader to allocate their assets in various periods of time and states.

### Security and its payoff

In this simple framework, the security market is consist of a group of securitis. A security is a **financial claims**. When bought at time $0$, it will give the holder a **payoff** at time $1$. Thus we can define a ***security*** by define all payoffs in all possible state at time $1$: $\vec x = \SB{x_1; x_2; \cdots; x_\Omega}$ (a *column vector*) in the $\mathbb{R}^\Omega$, the ***payoff space***.

$Remark$

>Sometimes we just ignore the ***dividend (股利)*** the payoff from security, because it might already be included in the price when trading.

***Financial asset*** is a kind of security that demand the financial claim for ***physical asset***. Its payoff comes from the ***underlying asset***. ***Stock*** is one of them.

- ***riskless bond***: those with a constant payoff in any state at time $1$
- ***risk bond***: stuffs like stocks, with various payoffs in different states.

$Remark$

1. The payoff is determined ONLY by the future economy
2. The payoff is exogenous, not affected by the behaviors of the agents
3. Every agents is facing the same probability distribution of different states at time $1$.

### Structure of The Market

Let there be $N$ tradable securities. Define the ***payoff vector*** (a *column vector*):

$\bspace \vec X_{\cdot , n} \equiv \SB{x_{1,n};x_{2,n}; \dots; x_{\omega, n}; x_{\Omega, n}}$

Then with this we define the ***market structure***, a matrix:

$\bspace X \equiv \SB{X_{\cdot, 1}, X_{\cdot, n}, \dots, X_{\cdot, N}} \equiv \left[\begin{array}{c}
X_{1,\cdot}\\
\cdots \\
X_{\omega,\cdot}\\
\cdots \\
X_{\Omega,\cdot}
\end{array}\right] = \SB{\begin{array}{ccccc}
x_{1,1} & \cdots & x_{1,n} & \cdots & x_{1,N} \\
\vdots  & \ddots & \vdots  & \ddots & \vdots  \\
x_{\omega,1} & \cdots & x_{\omega,n} & \cdots & x_{\omega, N} \\
\vdots  & \ddots & \vdots  & \ddots & \vdots  \\
x_{\Omega,1} & \cdots & x_{\Omega,n} & \cdots & x_{\Omega,N} \\
\end{array}}$

### Security portfolio

A ***security portfolio*** is a set of the numbers you hold for different securities, denoted as $\theta \equiv \SB{\theta_1;\theta_2;\dots;\theta_N}$, ***holding vector*** (a *column vector*).

For every agent $k$, we denote his holding vector at time $0$ as $\bar\theta_k \equiv \SB{\bar\theta_{k,1}; \bar\theta_{k,2}; \dots; \bar\theta_{k,N}}$ and with this we can define the ***market portfolio***: $\theta_M = \sum_{k=1}^{K} \bar\theta_k$, this is a set of all tradable securities.

### Marketed payoff

For any payoff $x$, if it can be duplicated or generated by a **security portfolio**, $i.e.$, $\exists \theta$ $s.t.$ $X\theta = x$, then we call this payoff ***marketed (市场化的)***, meaning that it's achievable by certain trading in the market.

Denote all possible marketed payoff as $M$: $M \equiv \CB{X\theta: \theta\in\mathbb{R}^N}$

Generally, $M$ is a subset of the payoff space $\mathbb{R}^{\Omega}$. Actually it's an $N\texttt{-dimension}$ subspace of $\mathbb{R}^{\Omega}$.

### Trading Process

- Centralized exchange集中化交易: the agents submit their request to buy or sell and we these are collected by the system, the buy or sell price is automatically given by the system and perform the trading.
- Over the counter柜台交易: The agents are connected by web so that they can directly find the opposite and finish the trade.

Other than these two trading class, the actual process is way more complex and is called the ***market micro-structure***.

Denote $S \equiv \SB{S_1;S_2;\dots;S_N}$ (a *column vector*) as the ***price vector*** for trading the securites. Also, denote $\theta_k\P{S} \equiv \SB{\theta_{k,1}\P{S};\dots;\theta_{k,N}\P{S}}$ as the ***demand vector*** for agent $k$ under **price vector** $S$. 

Then we assume the trading process in this book: among all possible $S$, the system will find one so that 

$$\sum_{k=1}^{K}\theta_k\P{S} = \sum_{k=1}^{K} \bar\theta_k = \theta_M$$

This $S$ now is called the ***market-clearing price***, with which are trades are finished (the amount of supply is equal to the amount of demand). This equation is also called the ***market-clearing condition***.

***

There're other factors affect the trading, called the ***market friction***. To ignore them when handle the real problem, we need some additional assumptions to make a ***frictionless market***:

1. 所有参与者可以无成本参与证券市场
2. 没有交易成本
3. 对于参与者的证券持有量没有头寸限制
4. 个体参与者的交易不会影响证券价格
5. 没有税收

## Basic Economy Model

$Def$ Economy

1. Two periods of time noted as $0$ then $1$ and there're $\Omega$ states at time $1$. There's a probability messure for $\Omega$. There's only one **Perishable** good.
2. There're $K$ agents
    1. They hold same information about the probability of any future states, denote as $P$
    2. They have **endowment** $e_k \in \mathbb{R}_{+}^{1+\Omega}$
    3. They have a **preference** $\succsim_k$ defined on the consumption set $C = \mathbb{R}_{+}^{1+\Omega}$ which also satisfy the $Axiom.1 \sim 3$
3. There's one frictionless market with **market structure** $X$.

$Def$ Security-Market Economy

Following the definition of **Economy**, if all $e_k$ can be expressed as the payoff of his initial security portfolio, $\bar\theta_k$, we call this **Economy** the ***ecurity-Market Economy***.

## Market Equilibrium

If, luckily, the price after being affected by the demand of agents of a security, made a perfect match of the supply and demand, we call the market has reached a eqilibrium.

$Remark$

>Two key factors in asset allocation: time and risk.

### Independent Optimizing

Think of a possible security portfolio $\theta = \SB{\theta_1;\dots;\theta_N}$ denote the number of securies $n$ holding. We call ***Long*** is $\theta_n > 0$ and ***Short*** if negative. Generally $\theta \in \mathbb{R}^N$. At this time, $\theta$的市场价值是$\Tran{S}\theta$

Consider an agent with endowment $e$. If he never trades, then his utility is $U\P{c} = U\P{e}$. He can use $\Tran{S} \theta$ to buy the portfolio $\theta$ at time $0$ then change his future consumption plan to $c_1 = e_1 + X\theta$ and the result is the change of current consumption plan: $c_0 = e_0 - \Tran{S}\theta$.

Here we call the cost to buy the portofolio at time $0$ ***savings (储蓄)*** (which means to borrow if it's negative). Then given $e$ the endowment, $X$ the payoffs and $S$ the price, all his possbile consumption plan construct a ***budget set***:

$$\text B\P{e, \CB{X,S}} = \CB{c \geq 0: c_0 = e_0 - \Tran{S} \theta, c_1 = e_1 + X\theta, \theta \in \mathbb{R}^{N}}$$

Given the set, the agent need to find the optimized one! An optimization problem:

$$\begin{array}{ll}
\max\limits_{\theta_k} & U_k\P{c_k} \\
s.t. & c_{k,0} = e_{k,0} -\Tran{S} \theta_k , c_{k,1} = e_{k,1} - X \theta_k \\
& c_0, c_1 \geq 0
\end{array}$$

for $k = 1,2,\dots,K$. Or equivalently written as $\d{\max_{c_{\left.\right._k} \in \text B \P{e_{\void_k}, \CB{X,S}}}U_k\P{c_k}}$.

The solution is $\theta_k \P{e_k,S}$.

### Market clearing

First we can assume that the initial number of portfolio each agent hold is $0$, then given the endowment of each agent $e_k$ we have the **market clearning condition**: $\sum\limits_{k=1}^{K} \theta_k \P{e_k,S} = 0$, which detremines the $S$.

Since $c_{k,0} = e_{k,0} -\Tran{S} \theta_k\P{e_k,S}$, thus $\sum c_{k,0} = \sum e_{k,0} - \Tran{S} \sum\theta_k \P{e_k,S} = \sum e_{k,0}$

And likewise we have $\sum c_{k,\omega} = \sum e_{k,\omega}, \omega = 1,2,\dots,\Omega$. We then rewrite it as $\sum c_k = \sum e_k$ called ***Walras's law***.

And to solve the eqilibirum, we first find the optimized **security portfolio** under *any* **price vector** $S$, or say, his demand $\theta_k \P{e_k,S}$. Then solve by let the sum of that equals to $0$. Generally, we have

$$S = S \P{P\;;\CB{U_k, e_k; k = 1,2,\dots,K};X}$$

implying that the price of securities are dependent on the ***fundamentals基本面 (primitives本源)*** of the economny: the risk; the preference and endowment, and the structure of the market.

## Optimality

How to decide an allocation is optimized or not? We first define the allocation $\CB{c_k,k = 1,2,\dots,K}$. And

$Def$ Pareto Dominant

An allocation $\CB{c_k, \forall k}$ is ***parato dominant*** to $\CB{c_k', \forall k}$ if $\forall k$: $U_k\P{c_k'} \geq U_k\P{c_k}$ and strictly true for at least one agent.

However dominancy can't ensure the feasibility so that the dominant allocation is applicable:

$Def$ Feasible

Given the total supply of an economy $\CB{e_k, \forall k}$, an allocation $\CB{c_k, \forall k}$ is ***feasible*** if: $\sum\limits_k c_k = \sum\limits_k e_k$

$Def$ Pareto Optimality

If allocation $\CB{c_k, \forall k}$ is **feasible** and NO other allocations dominant that, then it's of ***pareto optimality***.

$Remark$

An allocation of Pareto optimality is also called an ***efficient allocation***. Also, if an security market allows its agents reach that allocation, possibly, then it's a ***efficient market***.

## Summary

- 经济结构三个方面
    - 自然环境(Environment)
    - 经济参与者(Agents)
    - 金融市场(Market)
- 外部环境的两个要素
    - 时间：$0$ 与 $1$
    - 风险：即概率，在$1$期有$\Omega$种可能状态，并有对应概率测度$P$.
    - 假设经济中只有一种不可储存(perishable)的商品
- $K$个参与者，两个特征：经济**资源**和经济**需求**
    - 禀赋(Endowment)：$e_k\in\mathbb R_+^{1+\Omega}$
    - 每一个参与者对于未来状态的信息一致且完全由$P$描述
    - 假定参与者禀赋和信息并不会产生生产技术
    - 具有特定偏好(Preference)
        - 定义于$C=\mathbb R_+^{1+\Omega}$上，此集合**闭(CLose)**且**凸(Convex)**
        - 不满足性(instatiability)：$a>b\Longrightarrow a \succ b$
        - 连续性(continuity)：$\forall c \in C$, 集合 $\CB{a \in C: a \succsim c}$ 与 $\CB{b \in C: b \precsim c}$ 是**闭集**
        - 凸性(convexity)：$\forall a,b \in C$ and $\alpha \in \P{0,1}$, if $a \succ b$ then $\alpha a + \P{1-\alpha}b \succ b$
        - 以及完备性(Completeness)和传递性(Transtivity)
    - 效用函数(Utility Function)，$U$
- 金融市场由一组证券(Security)构成
    - 支付(Payoff)只取决于未来的实际经济状态
    - 支付可认为是外生(Exogenous)的，不受参与者影响
    - 参与者知道且一致认同此支付的概率分布
    - 市场结构(Market Structure)为矩阵$X$，无摩擦(frictionless)
- 市场均衡(Market Equilibrium)的两个条件
    - 每个参与者各自优化其证券需求量
    - 市场出清(Market Clearing)，交易的证券总需求等同于总供给，此时商品市场也出清
- 若配置(Allocation) $\CB{c_k,\forall\; k}$ 可行(feasible)，且无相对其占优(dominant)的可行配置，则称其为Pareto最优(Pareto Dominant)

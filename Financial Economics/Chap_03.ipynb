\newcommand{\ow}{\text{otherwise}}Def$ $\texttt{state-}\omega$ ***contingent claim***

Those security with payoff $1$ under state $\omega$ or $0$ payoff otherwise. The analytical form can be expressed as $x\P{\omega'} = 1_{\omega} \P{\omega'}$. Here $1_{\omega} \P{\cdot}$ is an ***indication function示性函数*** and will be assigned $1$ when $\omega' = \omega$ and $0$ otherwise.

$Remark$

This kind of security is also called the $\texttt{state-}\omega$ ***contingent security*** or just ***Arrow-Debreu security***. 状态 $\omega$ 或有要求权（证券）

And for the market that is *only* consisted of *all possible* **Arrow-Debreu security**, it've got a name of ***Arrow-Debreu security market***. And in this market the number of different securities is the same as the number of different states: $N = \Omega$.

If we order the security so that they match their corresponding state, we can find that it's **market structure** is a identity matrix.

$$X^{\text{A-D}} \equiv I_N$$

## State Price

Let $\phi_\omega$ be the price of $\texttt{state-}\omega$ **contingent claim** at time period $0$, with the same measure of the goods of the same time. It's price is called the ***state price*** of state $\omega$, since the only payoff happens under that state.

In the **Arrow-Debreu security market**, we have the ***state price vector*** of all **state price** of state $\omega$: $\phi \equiv \SB{\phi_1; \cdots; \phi_\Omega}$. 

In practice some of the numbers could be negative but here, all **state price** of state $\omega$ is *greater* than $0$. There's no free lunch. And thus: $\phi \gg 0$.

## Complete Market
$Def$

If any consumption plan $c$ 都可以通过有限成本的可交易证券的组合来融资, then we call this market a ***complete market***.

Consider any **consumption plan** in time period $1$, $x = \SB{c_{11}; c_{12}; \cdots; c_{1\Omega}}$. Then consider the **security portfolio**: $c_{11}$ units of $\texttt{state-}1$ **contingent security**, $c_{12}$ units of $\texttt{state-}2$ **contingent security** until $c_{1\Omega}$ units of $\texttt{state-}\Omega$ **contingent security**. That's to say $\theta = \SB{c_{11}; c_{12}; \cdots; c_{1\Omega}}$. 

It's obvious that this security portfolio has a payoff of $X\theta = X^{\text{A-D}}\SB{c_{11}; c_{12}; \cdots; c_{1\Omega}} = x$

And it's cost is $\Tran{\phi}\theta = \d{\sum_{\omega \in \Omega} \phi_\omega c_{1\omega}}$. And this market is a **complete market**.

## Optimizing the Agents

Given a **Arrow-Debreu security market** and its **state price vector**, consider an agent with $e$ endowment and utility function $U$. We can find the endowment in time period $1$ by considering the security portfolio: 

$$\d{\bar\theta = \SB{e_{11}; e_{12}; \cdots; e_{1\Omega}}} \Longrightarrow X^{\text{A-D}}\bar\theta = e_1 = \SB{\begin{array}{c}
e_{11}\\
e_{12} \\
\vdots\\
e_{1\Omega}
\end{array}}$$

and this will lead to the same payoff as the endowment so that we call this portfolio ***replicating portfolio复制组合***.

And since $\bar\theta$ is tradable and has a market value of $\Tran\phi\bar\theta$ we assign a market value to the endowment of that agent: $\omega = e_0 + \Tran\phi e_1$, which is also called the ***financial wealth*** (only defined in **complete market** or some of endowment is *NOT* replicatable).

Thus in **Arrow-Debreu security market**, we can think of that agent first transferring his endowment to cash: $\omega = e_0 + \phi\Tran e_1$. And use $c_0$ of them to do consumption and then buy portfolio $\theta$ to obtain a future comsumption $c_1 = \theta$. Here's the restriction for $c_{1,2}$:

$$c_0 + \Tran\phi c_1 = \omega = e_0 + \Tran\phi e_1$$

To shorten this expression, we first extend our **state price vector**: $\hat\phi = \SB{1;\phi}$. The first element is the price of one unit of good at time period $0$, $\phi_0$ in the unit of $\phi_0$, so the value there is still $1$ or to say, $\phi_0 = 1$. After that we rewrite the restriction as: $\Tran{\hat\phi} c = \Tran{ \hat \phi} e$ or $\Tran{\hat\phi}\P{e-c} = 0$

Now we choose the **consumption set** $C = \mathbb{R}_{+}^{1+\Omega}$. Thus the optimization problem is converted to:

$$\begin{array}{ll}
\d{\max_{c\,\in\, C}} & U\P{c}\\
s.t. & \Tran{\hat\phi}\P{e-c} = 0 \\
& c\geq 0
\end{array}$$

Can we solve this?

$Theroem.1$ 

If $C = \mathbb{R}_{+}^{1+\Omega}$ and $U\P{\cdot}$ is *continous* in $C$, then we can solve this optimization problem.

$Proof$

>The **consumption plan** in the **budget set**: $\text B\P{e, \CB{X^{\text{A-D}},s}} = B\P{e, \CB{I,\phi}} = \CB{c \geq 0: \Tran{\hat\phi}c = \omega = \Tran{\hat\phi}e }$. Since $\phi_0 =1$ and $\phi \gg 0$, we can say that $B\P{e, \CB{I,\phi}}$ is an **bounded close** set in $C$:
>
>Denote the smallest element in $\hat\phi$ as $\phi_m$ then a result is for any **consumption plan** $c$ in $B$, any of its element won't exceed $\omega \over \phi_\omega$. Thus $B$ is **bounded**.
>
>And in this budget plan, $\lim \Tran{\hat\phi}c_n = \lim \omega = \omega$, thus **closed**.
>
>Then by **Weierstress Theorem**, the continuous function $U\P{\cdot}$ defined on $\mathbb{R}^{1+\Omega}$ can find its maximum in a **bounded close** set.

What's the properties of the solution then, if we've already known its existence?

First by the **Insatiablity Axiom**, we have $\ffrac{\partial U} {\partial c_0} \,\hat = \, \partial_0 U >0$ and $\ffrac{\partial U} {\partial c_{1\omega}} \;\hat = \; \partial_\omega U > 0, \forall \omega \in \Omega$. That is to say in whatever states, the ***marginal utility*** is *greater* than $0$. Thus:

$$\text{D} U \equiv \SB{\begin{array}{c}
\partial_0 U \\
\partial_1 U \\
\vdots \\
\partial_\Omega U
\end{array}}\gg 0$$

Here we call the utility function $U$ which has a required maximum, the ***objective function***, under several restrictions. The standard approach to this kind of problem is using the ***Lagrangian multipliers***.

$Theorem$

Assuming $\text{D}U \gg 0$, then the solution of the optimization problem satisfies (or is the same with the problem below):

1. $\partial_i U = \lambda\phi_i + \mu_i$, for $i = 0,1,\dots,\Omega$
2. $\Tran{\hat\phi}\P{e-c} = 0$
3. $\mu_i c_i = 0$, for $i = 0,1,\dots,\Omega$

Here $\lambda>0$ is the buget set restriction and $\mu_i \geq 0$ is the nonnegative restrictions for consumption $c_i$.

And the preceding three conditions are collectively called ***Karush–Kuhn–Tucker (KKT) conditions*** or 一阶条件. In general cases, these three are just necessary but not sufficient. However since we also have assumed that $U$ is convex, the [second-order sufficient conditions] [2sc] are satisfied automatically ($\QQQ$need proof), thus the KKT conditions is sufficient for this problem.

[2sc]: https://en.wikipedia.org/wiki/Karush%E2%80%93Kuhn%E2%80%93Tucker_conditions#Second-order_sufficient_conditions

**e.g.**

Consider another **Lucas tree** with two states $a$ and $b$ at time period $1$ with equal probabilities. Suppose that the **state prices** are $\phi_a$ and $\phi_b$ respectively. There's an agent with **endowment** $\P{e_0; e_{1a}; e_{1b}}$ and **utility funciton** $U\P{c_0, c_{1a}, c_{1b}} = \log c_0 + 0.5\P{\log c_{1a} + \log c_{1b}}$. Then what's the optimized security portfolio?

> Given the state price and his endowment we can write his total wealth: $w=e_0 + \phi_a e_{1a} + \phi_b e_{1b}$. And the optimized problem can be solved using the lagrangian multiplier and was converted to:
>
>$\bspace \begin{array}{ll}
\d{\max_{c_{\void_0}, c_{\void_{1a}}, c_{\void_{1b}}}} & \log c_0 + \ffrac{1} {2} \P{\log c_{1a}} \\
s.t. & w - \P{c_0 + \phi_a c_{1a} + \phi_b c_{1b}} = 0 \\
& c_0, c_{1a}, c_{1b} \geq 0
\end{array} \Longrightarrow$
>
>1. $\partial_i U = \lambda\phi_i + \mu_i \Longrightarrow \begin{cases}
1/c_0 = \lambda + \mu_0 \\
0.5\P{1/c_{1a}} = \lambda \phi_a + \mu_a \\
0.5\P{1/c_{1b}} = \lambda \phi_b + \mu_b
\end{cases}$
>2. $\Tran{\hat\phi}\P{e-c} = 0 \Longrightarrow c_0 + \phi_a c_{1a} + \phi_b c_{1b} = w$
>3. $\mu_i c_i = 0$, for $i = 0,1,\dots,\Omega \Longrightarrow \mu_i c_i = 0$, for $i = 0,a,b$
> 
> We can assume that $\phi \gg 0$ thus $c \gg 0$ and then we can solve the problem:

$\space$

> $\bspace\begin{cases}
\mu_i = 0 \\
c_0 = \lambda^{-1} \\
c_{1a} = \P{2\lambda\phi_{1a}}^{-1} \\
c_{1b} = \P{2\lambda\phi_{1b}}^{-1}
\end{cases} \Longrightarrow \lambda = \ffrac{2} {w} \Longrightarrow \begin{cases}
c_0 = \ffrac{1} {2}w \\
c_{1a} = \ffrac{1} {4}\ffrac{w} {\phi_a}\\
c_{1b} = \ffrac{1} {4}\ffrac{w} {\phi_b}
\end{cases}$

$Remark$

>We can find that when this problem is optimized, the nonlinear restriction is not effective ($c \gg 0$ contains that $c_0, c_{1a}, c_{1b} \geq 0$). We call the optimized solution the ***interior solution***, ***corner solution*** otherwise, mostly when the onlinear restriction contains stuffs like $10 \geq c_0, c_{1a}, c_{1b} \geq 0$. To be concise, we assume that later those solutions are all **interior**.

***

Now we suppose the existence of the **interior solution** so that we can write the KKT conditions as: $\text{D}U_k\P{c_k} = \lambda_k \hat\phi$ and here $\lambda_k$ is determined by the budget restrictions. Thus, we can say $\forall k$ and $\omega \in \Omega$, $\partial_\omega U_k\P{c_k} = \lambda_k \phi_\omega$ with a special one when $k=0$ and $\phi_0 = 1$. Thus

$$\ffrac{\partial_\omega U_k\P{c_k}} {\partial_0U_k\P{c_k}} = \phi_\omega \iff \partial_0 U_k = \ffrac{1} {\phi_ \omega} \partial_\omega U_k$$

An intuitive interpretation is when reaching the optimization, the state price of $\omega$ is the ratio of the marginal effect under state $\omega$ at time period $1$ to the marginal effect at time period $0$.

## Market Equilibrium

After we find the optimized security portfolio for a single agent, now we change our view to the whole market equilibruim. In **Arrow-Debrue** economy, the equilibrium is a set of **state prices** $\CB{\phi_\omega:\omega\in\Omega}$ that satisfies:

1. Given the state price $\phi_\omega$, all agents have an opimization solution:
$$\begin{array}{lcl}
c_k \P{e_k, \phi} &\!\!=\!\!& \text{argmax}\, U_k\P{c_k} \\
s.t. && c_{k,0} + \Tran\phi c_{k,1} = e_{k,0} + \Tran\phi e_{k,1} \\
&& c_k \geq 0
\end{array}$$

2. Security market clearing:
$$\sum_k c_{k,0} \P{e_k, \phi} = \sum_k e_{k,0}\\
\sum_k c_{k,1} \P{e_k, \phi} = \sum_k e_{k,1}$$

Well, do they have a solution?

$Theorem.3$ 

For an economy with **Arrow-Debrue security market**, there's ALWAYS a equilibrium. 

***
Forget about it. We now focus on the properties of the solution. Consider two agents $k$ and $k'$. We have

$$\ffrac{\partial_\omega U_k\P{c_k}} {\partial_0 U_k\P{c_k}} = \phi_\omega = \ffrac{\partial_\omega U_{k'}\P{c_{k'}}} {\partial_0 U_{k'}\P{c_{k'}}}$$

Thus at whatever states, for whoever agents, they share the common marginal effect.

**e.g.** 

Consider an economy with two states of same probability, $a$ and $b$. There're two agents $1$ and $2$ with respective endowment:

$$e_1:100 \to \begin{cases}
0 \\
0
\end{cases}\bspace\bspace e_2:0 \to \begin{cases}
200 \\
50
\end{cases}$$

and a same utility function $U\P{c} = \log c_0 + \ffrac{1} {2} \P{\log c_a + \log c_b}$. Also, assume the ***A-D security market*** (only two securities available here). Find the equilibrium.

> Denote $\phi = \SB{\phi_a;\phi_b}$ as the state price. Then the wealth of agents: $w = \Tran{\hat\phi} e$, here $\Tran{\hat\phi} = \SB{1;\phi}$ and $e$ is his endowment.
>
>Now we can solve the optimization problem follow the same process as before and obtain that for $k=1,2$, $c_{k,0} = \ffrac{1} {2} w_k$; $c_{k,a} = \ffrac{1} {4} \ffrac{w_k} {\phi_a}$; $c_{k,b} = \ffrac{1} {4} \ffrac{w_k} {\phi_b}$. Here $w_1 = 100$ and $w_2 = 200\phi_a + 50 \phi_b$.
>
>Then the market clearing: 
>
>$$c_{1,a} + c_{2,a} = \ffrac{1} {4}\ffrac{100} {\phi_a} + \ffrac{1} {4} \ffrac{200\phi_a + 50\phi_b} {\phi_a} = 200 \\
c_{1,b} + c_{2,b} = \ffrac{1} {4}\ffrac{100} {\phi_b} + \ffrac{1} {4} \ffrac{200\phi_a + 50\phi_b} {\phi_b} = 50$$
>
>And the solution for the equation set is $\phi_a = 0.25$ and $\phi_b = 1$. Thus the wealth for agent $2$ is also $200\times  0.25 + 50 \times 1 = 100$, just like agent $1$. 
>
>The efficient allocation here is $c_1 = c_2 = \SB{50;\SB{100;25}}$. And we can also compute their marginal effect:
>
>$$\ffrac{\partial_\omega U_k\P{c_k}} {\partial_0 U\P{c_k}} = \ffrac{\d{\ffrac{1} {2} \ffrac{1} {c_{k,\omega}}}} {\d{\ffrac{1} {c_{k,0}}}} = \ffrac{c_{k,0}} {2 c_{k,\omega}} = \ffrac{\d{\ffrac{w_k} {2}}} {\d{\ffrac{2w_k} {4\phi_\omega}}} = \phi_\omega = \begin{cases}
0.25, &\text{if } \omega = a\\
1, &\text{if } \omega = b
\end{cases}$$

***

$Theorem.4$

For an economy with **A-D security market**, the allocation in equilibrium is **parato dominant**.

$Proof$

>Let $\CB{c_k,k=1,2,\dots,K}$ denote the allocation at equilibrium; $\phi$ the state price. The market clearing conditions require that $\d{\sum_{k=1}^{K} c_k = \sum_{k=1}^{K} e_k}$, or the ***feasibility condition***, which imply that

$\space$

>$$\Tran{\hat\phi}\sum_{k=1}^{K} c_k = \Tran{\hat\phi}\sum_{k=1}^{K} e_k$$
>
>because $\hat\phi \gg 0$. Now we give the proof by contradiction. Suppose that this allocation is not Pareto dominant, instead, there exists another feasible alloction $\CB{a_k,k=1,2,\dots,K}$ such that $a_{k'} \succsim c_{k'}$ for all agents $k'$ and at least for one agent $k$ $a_{k} \succ c_{k}$ (the textbook here use the sign $>$ not $\succ$).
>
>Then, $\forall k'$, if $a_{k'} \succsim c_{k'}$, then $\Tran{\hat\phi}a_{k'} \geq \Tran{\hat\phi}c_{k'}$. If not, then there exists $a_{k'}$ such that $a_{k'} \succsim c_{k'}$ while$\Tran{\hat\phi}a_{k'} < \Tran{\hat\phi}c_{k'}$ so that there can be a $\varepsilon = \Tran{\hat\phi}\P{c_{k'}-a_{k'}} >0$ which can be used to buy an extra consumption plan $d_{k'} \gg 0$ so that $a_{k'}+d_{k'} \gg a_{k'}$. Then from the **instatiability** of **preference**, we have:
>
>$$\Tran{\hat\phi}\P{d_{k'}+a_{k'}}$$ 
>
>and thus $a_{k'} + d_{k'} \succ a_{k'} \sim c_{k'}$ which is IMPOSSIBLE, since $c_{k'}$ is the optimized solution in the budget plan. So now we can draw the conclusion that $a_{k'} \succsim c_{k'} \Longrightarrow \Tran{\hat\phi}_a \geq \Tran{\hat\phi}_c$. 
>
>Next, we can also prove that $a_k \succ c_k$ leads to $\Tran{\hat\phi} a_k > \Tran{\hat\phi} c_k$ which means that $a_k$ is completely outside of the budget plan for agent $k$.
>
>Combine these two results, we have $\Tran{\hat\phi} \sum\limits_{k=1}^{K} a_k = \Tran{\hat\phi}a_k + \sum\limits_{k' \neq k}^{K} a_{k'} > \Tran{\hat\phi} \sum\limits_{k=1}^{K} c_k \using{\text{feasiblility condition}}\;\;\; \Tran{\hat\phi} \sum\limits_{k=1}^{K} e_k$. Thus $\CB{a_k}$ not feasible! A contradiction! So the allocation in equilibrium must be **parato dominant**.

## Summary

- 定义
    - 状态 $\omega$ 或有要求权($\texttt{state-}\omega$ contingent claim)表示状态 $\omega$ 时支付 $1$ 而其他状态下支付为 $0$ 的证券
    - 由所有可能的状态或有证券（即其完全集合）所构成的证券市场叫做 **Arrow-Debreu security market**
- **完全市场**(complete market)，如果市场中的任一有限消费计划都可以通过有限成本的可交易证券的组合来融资。**Arrow-Debreu证券市场**就是一个
- 对存在**Arrow-Debreu证券市场**的经济来说，均衡(Equilibrium)总是存在的，同时满足每一个参与者达到最优化和证券市场出清(Market Clearing)。均衡时，对于任何状态参与者的相对边际效应都相等
- ***福利经济学第一定理 (First Theorem of Welfare Economics)***：对存在**Arrow-Debreu证券市场**的经济来说，均衡达到的资源配置是Pareto最优。

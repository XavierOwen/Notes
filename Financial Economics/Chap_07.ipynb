## Definition of Risk Aversion

$Def$

$\tilde g$ is an uncertain payoff. If $\Exp\SB{\tilde g} = 0$, we call $\tilde g$ a ***fair game***.

$Def$

For an agent with utility function $u$ if $\forall\, \Exp\SB{\tilde g} = 0$, $\Exp\SB{u\P{w + \tilde g}} \leq \Exp\SB{u\P{w}}$, then we call this agent the type of ***risk averse***. Or, ***strict risk averse*** if the inequality holds strictly.

$Theroem.3$

The agent is (**strict**) **risk averse** $iff$ $u$ is a (**strictly**) **concave** function

$Proof$

$\bspace$Necessity$\Rightarrow)$:

>$\forall w_1 > w_2$ and $p\in\P{0,1}$, construct a **Bernoulli game** $\tilde g = \CB{g_1,g_2}$ with probability $\CB{p,1-p}$ and 
>
>$$g_1 = -\P{1-p}\P{w_2 - w_1},\bspace g_2 = p\P{w_2 - w_1}$$
>
>Then $\Exp\SB{\tilde g} = 0$. We define $w = pw_1 + \P{1-p}w_2$. Then $w_1 = w + g_1$ and $w_2 = w + g_2$. Since the agent is risk averse,
>
>$$pu\P{w + g_1} + \P{1-p}u\P{w+g_2} \leq u\P{w}\\
pu\P{w_1} + \P{1-p}u\P{w_2} \leq u\P{pw_1 +\P{1-p}w_2}, \bspace\forall\, w_1,w_2$$

$\bspace$Sufficiency$\Leftarrow)$:

>Since $u$ is convex utility function, by **Jensen** inequality, we have
>
>$$\Exp\SB{u\P{w + \tilde g}} \leq u\P{w + \Exp\SB{\tilde g}} = u\P{w}, \bspace \Exp\SB{\tilde g} =0$$
>
>Thus he's **risk averse**.

$Remark$

>This theorem proves that when using the utility function to describe the preference, convexity means risk aversion.

## Measure of Risk Aversion

### Absolute Risk Aversion

$Def$ Risk Premium

How the risk averse agent prefer the certain payoff to the uncertain? We define the ***risk premium风险溢价***, $\pi$ as the wealth an agent would like to give up in exchange of a lower risk in a fair game.

$$\Exp\SB{u\P{w+\tilde g}} = u\P{w - \pi}$$

And $-\pi$ is called the ***certainty equivalence确定性等值***.

An alternative defination could be: $$\Exp\SB{u\P{w + \tilde g + \hat\pi}} = u\P{w}$$

$\hat\pi$ is the minimum wealth an agent demand to take the risk. $\hat\pi$ might be different from $\pi$ but not when it's some kind of "small" risk.

$Def$ Game with small risk

When the $r.v.$ $\tilde g$ has a small range, $\tilde g$ is called a game with small risk.

***

Then there's taylor's expansion. 

$$\begin{cases}
\Exp\SB{u\P{w + \tilde g}}& = u\P{w} + u'\P{w}\Exp\SB{\tilde g} + \ffrac{1} {2} u''\P{w} \Exp\SB{\tilde g^2} + o\P{\tilde g^2} = u\P{w} + \ffrac{1} {2} u''\P{w} \Exp\SB{\tilde g^2} + o\P{\tilde g^2} \\
u\P{w - \pi} &= u\P{w} -u'\P{w}\pi + o\P{\pi^2}
\end{cases}\\[1.5em]
\Rightarrow \pi = \ffrac{1} {2} \SB{-\ffrac{u''\P{w}} {u'\P{w}}}\Exp\SB{\tilde g^2} = \ffrac{1} {2} \SB{-\ffrac{u''\P{w}} {u'\P{w}}}\Var{\tilde g}
$$

$Remark$

>Seeing from the derivation we can know that the second definition leads to the same result. For the game with small risk, the Variance is the measure of it. Also notice that $u'>0$ and $u'' \leq0$ so that actually $\pi$ is positively proportional to $\Var{\tilde g}$.

The ratio is called the ***Arrow-Pratt*** measure, $A\P{w} \equiv -\ffrac{u''\P{w}} {u'\P{w}} \geq 0$. And this term is also called the ***Absolute Risk Aversion***, which is related to the utility function $u$ and the wealth $w$.

More than that, we define the ***risk tolerance***: $T\P{w} \equiv \ffrac{1} {A\P{w}} = - \ffrac{u'\P{w}} {u''\P{w}}$

### Relative Risk Aversion

Now we take his total wealth as the base.

$$\Exp\SB{u\P{w\P{1+\tilde g}}} = u\P{w\P{1-\pi_R}}$$

Here the profit in the game is $w\tilde g$, propotion to the total wealth $w$. We can derive the similar formula of $\pi_R$ for small risk game:

$$\pi_R = \ffrac{1} {2}\SB{-\ffrac{wu''\P{w}} {u'\P{w}}} \Var{\tilde g}$$

The ***Relative Risk Aversion*** then is the ratio: $R\P{w} \equiv -\ffrac{wu''\P{w}} {u'\P{w}}$

## Examples

1. Linear of risk-neutral utility function: $u\P{w} = w$ where $A\P{w} = R\P{w} = 0$
2. Negative exponential utility function: $-e^{-aw}$ with $a > 0$ where $A\P{w} = a, R\P{w} = aw$. We also call this the ***constant absolute risk aversion*** (***CARA***).
3. Quadratic utility function: $u\P{w} = w -\ffrac{1} {2} aw^2$ with $a>0, w \in [0,1/a)$ where $A\P{w} = \ffrac{a} {1-aw}$ and $R\P{w} = \ffrac{aw} {1-aw}$
4. Power exponent utility function: $u\P{w} = \ffrac{1} {1-\gamma} w^{1-\gamma}$ with $\gamma > 0,\gamma \neq 1$ where $A\P{w} = \ffrac{\gamma} {w}$ and $R\P{w} = \gamma$. This one is also called the ***constant relative risk aversion*** (***CRRA***). 
5. Logarithmic utility function: $u\P{w} = \log w$ where $A\P{w} = 1/w$ and $R\P{w} = 1$, also a kind of ***CRRA***. 
6. Hyperbolic absolute risk aversion (HARA) utility function: directly defined by the **Arrow-Pratt** measure where $A\P{w} = \ffrac{1} {d+w/\gamma},\gamma \geq -1$. And actually it contains all types above
    - Linear of risk-neutral utility function: $d = \infty$
    - Negative exponential utility function: $r = -1, d = 1/a$
    - Quadratic utility function: $r \to \infty, d=1/a$
    - Power exponent utility function: $d = 0, \gamma>0, r \neq 1$
    - Logarithmic utility function: $d=0, r \to 1$

Given a reference, if the absolute risk aversion increases with the wealth, we can it a ***increasing absolute risk aversion*** (***IARA***), or equivalently, when $A'\P{w} > 0$. And if the relative risk aversion increases with the wealth, we call it a ***increasing relative risk aversion*** (***IRRA***), or equivalently, when $R'\P{w} >0$.　And if the two are decreasing, then we have the ***DARA*** and ***DRRA***.

## Comparison between Risk Aversion

$Theorem.4$ ***Pratt***

The following statements are equivalent, given two increasing, second order differentiable utility functions $u_1\P{\cdot}$ and $u_2\P{\cdot}$ with their **absolute risk aversions**, $A_1\P{\cdot}$ and $A_2\P{\cdot}$ respectively.

- $A_1\P{w} \geq A_2\P{w}, \forall \, w $
- $u_1\P{u_2^{-1}\P{z}}$ is **concave**
- $\exists \, f\P{\cdot}$, $f'\P{\cdot} >0$ and $f''\P{\cdot} \leq 0$ such that $u_1\P{w} = f\SB{u_2\P{w}}$
- $\pi_1 \geq \pi_2$, this holds for all possible $w$ and fair game.

$Proof$

$\bspace 1 \Rightarrow 2$

> Let $f\P{z} = u_1\P{u_2^{-1}\P{z}}$ and assign $u_2^{-1}\P{z}$ to $w$. We first compute the derivatives of $f$:
>
>$$f'\P{z} = \ffrac{\dd{u_1\P{w}}} {\dd{w}}\cdot\ffrac{\dd{w}} {\dd{z}} = u_1'\P{w}\cdot\ffrac{1} {u_2'\P{w}}>0$$
>
>$Remark$
>
>>The derivative of inverse  function can be obtained by
>>
>>$$w = u_2^{-1}\P{z} \Rightarrow z = u_2\P{w} \Rightarrow 1 = u_2'\P{w}\ffrac{\dd w} {\dd z}$$
>
>$$\begin{align}
f''\P{z} &= \ffrac{u_1''\P{w}u_2'\P{w} - u_1'\P{w}u_2''\P{w}} {\SB{u_2'\P{w}}^3}\\
&= -\SB{-\ffrac{u_1''\P{w}} {u_1'\P{w}} + \ffrac{u_2''\P{w}} {u_2'\P{w}}} \cdot\ffrac{u_1'\P{w}} {\SB{u_2'\P{w}}^2}\\
&= -\P{A_1\P{w} - A_2\P{w}} \ffrac{\text{positive}} {\text{positive}} \leq 0\\
\end{align}$$

$\bspace 2 \Rightarrow 3$

>Now since $u_2$ is incresing, $u_2^{-1}\P{\cdot}$ exists, we can define $f\P{z}=u_1\P{u_2^{-1}\P{z}}$. Then 
>
>$$f\P{u_2\P{w}} = u_1\P{w}$$
>
>Lucky we are! This $f$ can exactly satisfy the conditions in statement $3$ since it's just the same with what we defined in the preceding proof.

$\bspace 3 \Rightarrow 4$

> $\forall \tilde g$, we have
>
>$$u_1\P{w - \pi_1} =\Exp\SB{u_1\P{w + \tilde g}} = \Exp\SB{f\P{u_2\P{w + \tilde g}}}$$
>
>By **Jensen**, $\Exp\SB{f\P{u_2\P{w + \tilde g}}} \leq f\P{\Exp\SB{u_2\P{w + \tilde g}}} = f\P{u_2\P{w-\pi_2}} = u_1\P{w-\pi_2}$. By 不满足性, $w-\pi_1 \leq w-\pi_2 \Rightarrow \pi_1 \geq \pi_2$.

## First Order Risk Aversion

What if $u\P{\cdot}$ is indifferentiable? To find the fact we first consider the utility function

$$u\P{w} = \begin{cases}
a_+ \P{w - \bar w}, &\text{if } w \geq \bar w \\
a_- \P{w - \bar w}, &\text{if } w < \bar w 
\end{cases},\bspace a_->a_+ \geq 0$$

His wealth is $\bar w$. Now consider the fail Bernoulli game with equal possibility to win or loss $\delta$, its risk premium is, by the definition of risk premium: $\Exp\SB{u\P{\bar w + \tilde g}} = u\P{\bar w - \pi}$

$$ \ffrac{1} {2}a_+ \P{\bar w + \delta - \bar w} + \ffrac{1} {2}a_-\P{\bar w - \delta - \bar w} = a_- \P{\bar w - \pi - \bar w}\\
\Downarrow \\
\pi = \ffrac{1} {2} \ffrac{a_- - a_+ }{a_-}\delta$$

while the game has the variance $\delta^2$ and we've already proved that when differentiable, $\pi \propto A\P{w} \delta^2$ while linear related when indifferentiable. Thus when $\delta$ is small, its risk premium is one order greater than that with differentiable utility function, meaning a greater corresponding Risk Aversion.

## Summary

First let's find all that are related to "conVex" or "conCave".

- **Convex Set**: $A\subseteq \RR^n$, $\forall a,b \in A$ and $\alpha \in \SB{0,1}$, if $\alpha a + \P{1-\alpha}b \in A$ we say that set $A$ is a conVex set.
- **Strong Convexity** (preference), $\forall a,b \in C$ and $\alpha \in \P{0,1}$, if $a \succ b$, then $\alpha a+\P{1-\alpha}b \succ b$
- **Convexity** (utility), $\forall a,b \in C$ and $\alpha \in \P{0,1}$, if $U\P{a} > U\P{b}$, then $U\P{\alpha a+\P{1-\alpha}b} > U\P{b}$
- **Concavity** (function), $u\P{\cdot}$ as the utility of the agent by consumption path from initial to a certain state, $\forall x,y$ and $\alpha \in \SB{0,1}$, if $u\P{\alpha x + \P{1-\alpha }y} \geq \alpha u\P{x} + \P{1-\alpha} u\P{y}$ then we call it $u\P{\cdot}$ is conCave.
- **Weak Convexity** (preference): $\forall a,b \in C$ and $\alpha \in \SB{0,1}$, if $a \succsim b$, then $\alpha a + \P{1-\alpha}b \succsim b$
- ***Quasiconvex/Quasiconcave*** (function): A function ${\displaystyle f:S\to \mathbb {R}}$ defined on a convex subset $S$ of a real vector space is **quasiconvex** if for all ${\displaystyle x,y\in S}$ and ${\displaystyle \lambda \in [0,1]}$ we have: ${\displaystyle f(\lambda x+(1-\lambda )y)\leq \max {\big \{}f(x),f(y){\big \}}} $, and as for **quasiconcave**, $f(\lambda x+(1-\lambda )y)\geq \min {\big \{}f(x),f(y){\big \}}$.

## Question 2.1

## Question 2.2

$\P{1}$ instatiability

$\bspace$Suppose there're two possible consumption plan $c_i = \SB{c_{0,i};c_{1a,i};c_{1b,i}}, i=1,2$ and let $c_1 \geq c_2$ meaning that $c_{0,1} \geq c_{0,2}$ and $c_{1a,1} \geq c_{1a,2}$ and $c_{1b,1} \geq c_{1b,2}$. Then for the first Utility function,

$$U\P{c_1} = \log c_{0,1} + \ffrac{1} {2} \P{\log c_{1a,1} + \log c_{1b,1}} \geq \log c_{0,12} + \ffrac{1} {2} \P{\log c_{1a,2} + \log c_{1b,2}} = U\P{c_2} $$

$\bspace$And for the second one, when $1-\gamma > 0$, $\ffrac{1} {1-\gamma}x^{1-\gamma}$ is an increasing function on $x \geq 0$ and when $1-\gamma < 0$, $\ffrac{1} {1-\gamma}x^{1-\gamma}$ is still an increasing function on $x \geq 0$. Thus

$$U\P{c_1} = \ffrac{1} {1-\gamma}c_{0,1}^{1-\gamma} + \ffrac{1} {2} \P{\ffrac{1} {1-\gamma}c_{1a,1}^{1-\gamma} + \ffrac{1} {1-\gamma}c_{1b,1}^{1-\gamma}} \geq \ffrac{1} {1-\gamma}c_{0,2}^{1-\gamma} + \ffrac{1} {2} \P{\ffrac{1} {1-\gamma}c_{1a,2}^{1-\gamma} + \ffrac{1} {1-\gamma}c_{1b,2}^{1-\gamma}} = U\P{c_2} $$

$\bspace$For the third one, when $\alpha > 0$, $-e^{-\alpha x}$ is an increasing function on $x \geq 0$, however it turns to decreasing when $\alpha < 0$ which fails the instatiability. So one possible solution is to let $U\P{c} = -\alpha e^{-\alpha c_{\void_0}}-\ffrac{\alpha} {2} \P{e^{-\alpha c_{\void_{1a}}} + e^{-\alpha c_{\void_{1b}}}}$. Then we can easily prove that $U\P{c_1} \geq U\P{c_2}$. But to prove the convexity, we'd better assume that $\alpha > 0$.

$\P{2}$ continuity

$\bspace$Consider a series of consumption plan $\CB{c_{i}}, i=1,2,\cdots$ with a limit $\lim\limits_{n \to \infty} c_n= c_{\cdot}$ meaning that $\lim\limits_{n \to \infty} c_{j,n}= c_{j}$ for $j = 0,1a,1b$. For the first utility function, by the property of $\log$ function, we have $\lim\limits_{n \to \infty} \log c_{j,n} = \log \P{ \lim\limits_{n \to \infty} c_{j,n}} = \log c_{j}$ and thus we have their linear combination $\lim\limits_{n \to \infty}U\P{c_n} = U\P{c}$

$\bspace$For the second and third one, since $\ffrac{1} {1-\gamma}x^{1-\gamma}$ and $-e^{-\alpha x}$ are continuous on the domain, we can use the same method to draw the conclusion.

$\P{3}$ convexity

$\bspace$Suppose that $U\P{c_1} \geq U\P{c_2}$, and let $\beta \in \P{0,1}$, then for the first utility function, since $\log$ funciton is convex, we have

$$\begin{align}
U\P{\beta c_1 + \P{1-\beta}c_2} &= \log\P{\beta c_{\void_{0,1}} + \P{1-\beta}c_{\void_{0,2}}\,} +\ffrac{1} {2}\SB{ \log \P{\beta c_{\void_{1a,1}} + \P{1-\beta}c_{\void_{1a,2}}\,} + \log \P{\beta c_{\void_{1b,1}} + \P{1-\beta}c_{\void_{1b,2}}\,}} \\
& \geq \beta \P{\log c_{\void_{0,1}} + \ffrac{1} {2}\log c_{\void_{1a,1}} + \ffrac{1} {2}\log c_{\void_{1b,1}}} + \P{1-\beta} \P{ \log c_{\void_{0,2}} + \ffrac{1} {2}\log c_{\void_{1a,2}} + \ffrac{1} {2}\log c_{\void_{1b,2}}} \\
&= \beta U\P{c_1} + \P{1-\beta}U\P{c_2} \geq U\P{c_2}
\end{align}$$

$\bspace$Similarly, for the next two utility functions, $\P{\ffrac{1} {1-\gamma}x^{1-\gamma}}'' = -\ffrac{1} {\gamma}x^{-\gamma - 1} < 0$ and $\P{-e^{-\alpha x}}'' = -\alpha^2 e^{-\alpha x} <0$, thus all three are convex and we can prove the convexity for the later two utility functions.

$$
\begin{array}{cl}

$\bspace$We don't need lagrangian multipliersðŸ˜‚ðŸ˜‚ðŸ˜‚. $$s = \theta = \ffrac{100\rho \P{1+r_F} - 1} {\P{1+\rho}\P{1+r_F}}\\

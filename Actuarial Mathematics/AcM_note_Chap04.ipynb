# Insurance Benefits
## Summary
Formulae for the valuation of traditional insurance benefits, for whole life, term and endowment. For a basic review, $T_x$ is the future life time continuous random variable for life $(x)$, and $K_x$ is the curtate future life time. And a new random variable is $K_{x}^{(m)}$ as the benefits which depends on the number of complete period of length $\displaystyle \frac{1} {m}$ years the insured person $(x)$ lived.

## Intro
Assume that the interest rate is a constant. Assume a continuous benefit, like $\$1$ per year would be rather paid in infinitesimal unit of $\$\mathrm{d}t$ in every interval $(t,t + \mathrm{d}t)$.

## Assumption
$Def$

**Basis**: set of assumptions used in life insurance or pension calculations

And in this chapter we use the *Standard Ultimate Survival Model*: **Makeham’s law** with $A = 0.00022$, $B = 2.7 \times 10^{− 6}$, $c = 1.124$, an **ultimate basis**.

We assume a constant interest rate, the mortality model of the *ultimate basis*.

- The **effective annual rate of interst**, $i$, *Compound Interest*
- **annual accumulation factor**, $\newcommand{\ffrac} {\displaystyle \frac} 1+i$
- **discount factor**, $\boxed{ \upsilon = \ffrac{1} {1+i} }$
- **force of interest** (instantameous rate of growth of value), $\boxed{ \delta = \log(1+i) }$
- **nominal annual rate of interest compounded $p$ times per year**, $\boxed{ i^{\,(p)} = p\left( \left( 1+i \right)^{1/p} -1 \right) \Leftrightarrow 1+i = \left( 1+ \ffrac{i^{\,(p)}} {p} \right)^p }$
- **periodic rate**, $\ffrac{i^{\,(p)}} {p}$
- **The effective rate of discount per year**: $\boxed{ d = 1 - \upsilon = i\upsilon = 1 - e^{-\delta} }$
- **the nominal rate of discount compounded $p$ times per year**: $\boxed{ d^{\,(p)} = p\left( 1 - \upsilon^{1/p} \right) \Leftrightarrow \left( 1 - \ffrac{d^{\,(p)}} {p} \right)^p = \upsilon }$

The payment is contingent of the death of insured.

## Valuation of insurance benefits


### Whole life insurance: the continuous case, $\bar{A}_x$
$Def$

**Whole life insurance**: insurance contract that pays benefit IMMEDIATELY upon **death** of the insured, no matter how far in the future.

But the present value of that future payment, which is ONE DOLLAR, depends on payment date, or the date of death, which is *UNKNOWN*, thus it is an $r.v.$.

$$\boxed{ Z = \upsilon^{T_x}=e^{-\delta T_x} = \frac{1} {(1+i)^{T_x}} }$$

**EPV**: the expected present value, or the actuarial present value:

$$\boxed{ \bar{A}_x = E[Z] = E[e^{-\delta T_x}] = \int_{0} ^{\infty} e^{-\delta t} {_{t}}p_x \mu_{x+t} \, \mathrm{d}t }$$

$\dagger$ The bar upon $A$ means that the one dollar benefit is paid right on death. $\ddagger$

$\dagger$ And the EPV is the integral, of the product of the following
- presnet value of $1$ dollar: $1 \cdot e^{-\delta t}$
- probability of payting that at time $t$: ${_{t}}p_{x}\mu_{x+t}\mathrm{d}t$ $\ddagger$
***
$$\boxed{ E[Z^2] = E[e^{-2\delta T_x}]= \int_{0} ^{\infty} e^{-2\delta t} {_{t}}p_x \mu_{x+t} \, \mathrm{d}t = {^2}{\bar{A}_x} }$$

$\odot$ the $2$ on the left upper side of $\bar{A}$ meaning that calculation is at force of interest $2 \delta$, which is equivalent, at a rate of interest $j$, where $1 + j = e^{2\delta} = (1+i)^2$. $\triangleleft$

Then if the death benefit is $S$ then we have

$$E[SZ]=\int_{0} ^{\infty} Se^{-\delta t} {_{t}}p_x \mu_{x+t} \, \mathrm{d}t = S\bar{A}_x \\
Var[SZ]:=V[SZ] = S^2 \left( {^2}{\bar{A}_x} - \left( \bar{A}_x \right)^2 \right)$$

$\odot$ Low values of $Z$ are associated with high values of $T_x$. $\triangleleft$

### Whole life insurance: the annual case, $A_x$
Suppose now that the benefit of $1$ dollar is payable at the end of the year of death of $(x)$, rather than immediately on death. Since we use $K_x$ to denote curtate future life time, the time to the end of the year of death of $(x)$ is $K_x + 1$. So that now

$$\boxed{ Z = \upsilon^{K_x + 1} } \\
\begin{align}
A_x &= E[\upsilon^{K_x + 1}] = \boxed{ \sum_{k = 0}^{\infty} \upsilon^{k+1} \cdot {_{k \mid}}q_{x} } \\
&= \upsilon \cdot q_x + \upsilon^2 \cdot {_{1 \mid}}q_{x} + \upsilon^3 \cdot {_{2 \mid}}q_{x} + \cdots
\end{align}$$

Each term represents the EPV of a death benefit of a dollar paid at time $k$ conditional on the death of $(x)$ in $(k-1,k]$. 

And for the second moment, we have ${^2}A_x = E[\upsilon^{2(K_x + 1)}] = \boxed{ \sum_{k = 0}^{\infty} \upsilon^{2(k+1)} \cdot {_{k \mid}}q_{x} }$. And the variance is $S^2\left( {^2}A_x - \left( A_x \right)^2 \right)$

### Whole life insurance: the $1 /m \text{thly}$ case, $A_x^{(m)}$
$Def$

$K_x^{(m)}$, $r.v.$, where $m > 1$, as the future life time of $(x)$ in years rounded to lower $\ffrac{1} {m} \text{th}$ of a year. Commonly $m = 2,4,12$.

$$\boxed{ K_x^{(m)} = \frac{1} {m} \left \lfloor mT_x \right \rfloor }$$

For example:

$$\begin{array}{ccccc} \hline
T_x & K_x & K_x^{(2)} & K_x^{(4)} & K_x^{(12)} \\ \hline
23.675 & 23 & 23.5 & 23.5& {23\frac{8}{12}} \\
\upsilon^{T_x} & \upsilon^{K_{x}+1} & \upsilon^{K_{x}^{(2)} \, +1/2} & \upsilon^{K_{x}^{(4)} \, +1/4} & \upsilon^{K_{x}^{(12)} \, +1/12} \\ \hline
\end{array}$$
And for the probability:

$$Pr[K_x^{(m)} = k] = Pr[k \leq T_x < k + \frac{1} {m}] = {_{k \mid \frac{1} {m}}}q_x$$

For the expected value:

$$E[Z] = \boxed{ A_x ^{(m)} = \sum_{k=0}^{\infty} \upsilon^{\frac{k+1} {m}} {_{\frac{k} {m} \mid \frac{1} {m}}}q_x} = \upsilon^{1/m} {_{\frac{1} {m}}}q_x + \upsilon^{2/m} {_{{\frac{1} {m}} \mid {\frac{1} {m}}}}q_x + \upsilon^{3/m} {_{{\frac{2} {m}} \mid {\frac{1} {m}}}}q_x + \cdots $$

### Recursions
Although paying the death benefit at the year end, but sometimes we have to deal with the annual life tables. We can use **backwards recursion** to find the $A_x$.

Let the highest age in the table be $\omega$, so that we can assume that all lives expire by age $\omega$, expecially that $q_{\omega - 1} = 1$. Or if there's not a highest age, just choose one so that $q_{\omega - 1}$ is close enough to $1$.

And in that case, $K_{\omega - 1} = 0$

$$\boxed{ A_{\omega - 1} = E\left[ \upsilon ^{1+K_{\omega - 1}} \;\right] = \upsilon }$$

So that 

$$\begin{align}
A_x &= \sum_{k=1}^{\omega -x-1} \upsilon ^{k+1} \cdot {_{k}}p_x \cdot q_{x+k} \\
&= \upsilon \cdot q_x + \upsilon^2 \cdot p_x \cdot q_{x+1} + \upsilon^3 \cdot {_{2}}p_{x} \cdot q_{x+2} + \upsilon^4 \cdot {_{3}}p_{x} \cdot q_{x+3} + \cdots \\
&= \upsilon \cdot q_x + \upsilon \cdot p_x \left( \upsilon \cdot q_{x+1} + \upsilon^2 \cdot p_{x+1} \cdot q_{x+2} + \upsilon^3 \cdot {_{2}}p_{x+1} \cdot q_{x+3} + \cdots \right) \\
&= \boxed{\upsilon \cdot q_x + \upsilon \cdot p_x \cdot A_{x+1}}
\end{align}$$

And with this formula we can calculate $A_{x_{_0}}$ of the minimum age, backwards from $A_{\omega - 1} = \upsilon$.

And more terms backward are
0. $q_{\omega - 1} = 1$
1. $A_{\omega}$ not exists
2. $A_{\omega - 1} = v$
3. $A_{\omega - 2} = v q_{\omega - 2} + v \cdot p_{\omega - 2} \cdot v$
4. $A_{\omega - 3} = v q_{\omega - 3} + v^2 \cdot p_{\omega - 3} q_{\omega - 2} + v^3 p_{\omega - 3} p_{\omega - 2}$
5. $A_{\omega - i} = v \cdot q_{w-i} + \displaystyle \sum_{j=2}^i \left( v^j \left( \prod_{k=i-j+2}^{i} p_{\omega - k} \right) q_{\omega - (i-j+1)}\right) $
***
And for monthly benefits, the similar formula can be derived like:

$$\begin{align}
A_{x}^{(m)} &= \upsilon^{1/m}\cdot{_{\frac{1} {m}}}q_x + \upsilon^{2/m} \cdot {_{\frac{1} {m}}}p_x \cdot {_{\frac{1} {m}}} q_{x+\frac{1} {m}} + \upsilon^{3/m} \cdot {_{\frac{2} {m}}}p_x \cdot {_{\frac{1} {m}}} q_{x+\frac{2} {m}} + \cdots \\
&= \upsilon^{1/m} \cdot {_{\frac{1} {m}}}q_x + \upsilon^{1/m} \cdot {_{\frac{1} {m}}}p_x \left( \upsilon^{1/m} \cdot {_{\frac{1} {m}}}q_{x+\frac{1} {m}} + \upsilon^{2/m} \cdot {_{\frac{1} {m}}}p_{x+\frac{1} {m}} \cdot {_{\frac{1} {m}}}q_{x+\frac{2} {m}} + \cdots \right) \\
&= \boxed{\upsilon^{1/m} \cdot {_{\frac{1} {m}}}q_x + \upsilon^{1/m} \cdot {_{\frac{1} {m}}}p_x \cdot A^{(m)}_{x + \frac{1} {m}}}
\end{align}$$

### Term insurance
#### The continuous case $\bar {A}_{{x:\overline {n}|}}^{1}$

Now the death benefit is payable only if the policyholder dies *within* a fixed term of $n$ years. So now the present value, of a benefit of $1$ dollar, is

$$Z = \begin{cases}
\upsilon^{T_x} = e^{-\delta T_x} & \text{ if } T_x \leq n \\ 
0 & \text{ if } T_x > n 
\end{cases}$$

And here the EPV is

$$\boxed{ E[Z] = \bar {A}_{{x:\overline {n}|}}^{1} = \int_{0} ^{n} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t \,}$$

And similarly, the expected value of the square of the present value is

$$\boxed{ E[Z^2] = \sideset{^2}{_{{x:\overline {n}|}}^{1}}{\bar{A}} = \int_{0} ^{n} e^{-2\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t \,}$$

#### The annual case ${A}_{{x:\overline {n}|}}^{1}$

Now the death benefit is payable at the *end* of the year of death, provided that the policyholder dies *within* a fixed term of $n$ years. So now the present value, of a benefit of $1$ dollar, is

$$Z = \begin{cases}
\upsilon^{K_x + 1} & \text{ if } K_x \leq n - 1 \\ 
0 & \text{ if } K_x \geq n 
\end{cases}$$

And here the EPV is

$$\boxed{ E[Z] = {A}_{{x:\overline {n}|}}^{1} = \sum_{k = 0} ^{n-1} \upsilon^{k + 1} {_{k \mid}}q_{x} }$$

And similarly, the expected value of the square of the present value is

$$\boxed{E[Z^2] = {^2}{A}_{{x:\overline {n}|}}^{1} = \sum_{k = 0} ^{n-1} \upsilon^{2( k + 1)} {_{k \mid}}q_{x} }$$

#### The $1 /m \text{thly}$ case, ${A^{(m)}}{_{x:\overline {n}|}^{1}}$
Now the death benefit is payable at the *end* of the $1 /m \text{thly}$ year of death, provided that the policyholder dies *within* a fixed term of $n$ years. So now the present value, of a benefit of $1$ dollar, is

$$Z = \begin{cases}
\upsilon^{K_x^{(m)} \; +  \frac{1} {m}} & \text{ if } K_x^{(m)} \leq n - \frac{1} {m} \\ 
0 & \text{ if } K_x^{(m)} \geq n 
\end{cases}$$

And here the EPV is

$$\boxed{ E[Z] = {A^{(m)}}{_{x:\overline {n}|}^{1}} = \sum_{k = 0} ^{mn-1} \upsilon^{(k + 1)/m} {_{\frac{k} {m} \mid \frac{1} {m}}}q_{x} }$$

And similarly, the expected value of the square of the present value is

$$\boxed{ E[Z^2] = {^2}{A^{(m)}}{_{{x:\overline {n}|}}^{1}} = \sum_{k = 0} ^{n-1} \upsilon^{2( k + 1 )/m} {_{k \mid}}q_{x} }$$

### Pure Endowment
$\odot$ "Pure Endowment" is a term. And "Endowment" normally means "Endowment Insurance". $\triangleleft$

$Def$

A $n\text{-year}$ pure endowment with sum insured $100000$ dollars, issued to $(x)$, will pay $100000$ dollars in $n$ years if $(x)$ is still alive at that time

So now the present value, of a benefit of $1$ dollar, is

$$Z = \begin{cases}
0 & \text{ if } T_x < n & P = 1 - {_{n}}p_{x}\\ 
\upsilon^{T_x} & \text{ if } T_x \geq n & P = {_{n}}p_{x}
\end{cases}$$

So that the EPV is

$${_n}E_{x}:= \boxed{ E[Z] = {A}{_{{x:\overline {n}|}}^{\;\;1}} = \upsilon^n \cdot {_{n}}p_{x}  }$$

### Endowment insurance
It's a combination of a *term insurance* and a *pure endowment*. The sum insured is payable on the death of $(x)$ should $(x)$ die within a fixed term, say $n$ years, but if $(x)$ survives for $n$ years, the sum insured is payable at the end of the $n$th year.

So the present value is

$$\begin{align}
Z &= \begin{cases}
\upsilon^{T_x} = e^{-\delta T_x} & \text{ if } T_x < n \\ 
\upsilon^n & \text{ if } T_x \geq n 
\end{cases} \\
&= \upsilon^{\min{(T_x \, , \, n)}} \\
&= e^{-\delta \min{(T_x \, , \, n)}}
\end{align}
$$

Then the EPV should be

$$\begin{align}
E[Z] &= \int_{0} ^{n} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t &+ &\int_{n} ^{\infty} e^{-\delta n} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t \\
&= \int_{0} ^{n} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t &+ &e^{-\delta n} {_{n}}p_{x} \\
&= \bar {A}_{{x:\overline {n}|}}^{1} &+ &{A}{_{{x:\overline {n}|}}^{\;\;1}}
\end{align}$$

In complete actuarial notation, it would be like

$$\boxed{ \bar {A}_{{x:\overline {n}|}} = \bar{A}_{{x:\overline {n}|}}^{1} + {A}{_{{x:\overline {n}|}}^{\;\;1}} \, }$$

And similarly, the expected value of the squared present value of the benefit is 

$$\boxed{ {^{2}}\bar{A}_{{x:\overline {n}|}} = \int_{0} ^{n} e^{-2\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t + e^{-2\delta n} {_{n}}p_{x} = {^{2}}\bar{A}_{{x:\overline {n}|}}^{1} + {^{2}}{A}{_{{x:\overline {n}|}}^{\;\;1}} \, }$$

***

And if the benefit is paid at the end of the year of death, then the present value would be like

$$\begin{align}
Z &= \begin{cases}
\upsilon^{K_x + 1} & \text{ if } K_x \leq n - 1 \\ 
\upsilon^n & \text{ if } K_x \geq n 
\end{cases} \\
&= \upsilon^{\min{(K_x + 1,n)}}
\end{align}$$

So that the EPV is like

$$\boxed{ \sum_{k=0} ^{n-1} \upsilon^{k+1} {_{k\mid}}q_{x} + \upsilon^n  P[K_x \geq n] = {A}_{{x:\overline {n}|}}^{1} + \upsilon^n {_{n}}p_{x} \, }$$

And in complete actuarial notation, it will be like

$$\boxed{ {A}_{{x:\overline {n}|}} = {A}_{{x:\overline {n}|}}^{1} + {A}{_{{x:\overline {n}|}}^{\;\;1}} \, }$$

Similarly, the expected value of the squared present value of the benefit is

$$\boxed{ {^{2}}{A}_{{x:\overline {n}|}} = \sum_{k=0} ^{n-1} \upsilon^{2(k+1)} {_{k\mid}}q_{x} + \upsilon^{2n} {_{n}}p_{x} = {^{2}}{A}_{{x:\overline {n}|}}^{1} + {^{2}}{A}{_{{x:\overline {n}|}}^{\;\;1}} \, }$$

***

Finally if the death benefit is payable at the end of the $1 /m \text{th}$ year of death, then the present value of the benefit is

$$\begin{align}
Z &= \begin{cases}
\upsilon^{K_x^{(m)} \; +  \frac{1} {m}} & \text{ if } K_x^{(m)} \leq n - \frac{1} {m} \\ 
\upsilon^n & \text{ if } K_x^{(m)} \geq n 
\end{cases}\\
&= \upsilon^{\min{(K_x^{(m)} \; +  \frac{1} {m}\:,\: n)}} 
\end{align}$$

Here the EPV of the benefit is

$$\sum_{k = 0} ^{mn-1} \upsilon^{(k + 1)/m} {_{\frac{k} {m} \mid \frac{1} {m}}}q_{x} + \upsilon^n P[K_x^{(m)} \geq n] = {A^{(m)}}{_{x:\overline {n}|}^{1}} + \upsilon^n {_{n}}p_{x}$$

And in complete actuarial notation we have

$$\boxed{ A^{(m)}_{x:\overline {n}|} = {A^{(m)}}{_{x:\overline {n}|}^{1}} + {A}{_{{x:\overline {n}|}}^{\;\;1}} \, }$$

### Deferred insurance benefits
Deferred insurance refers to insurance which does not begin to offer death benefit cover until the end of a deferred period. 

For example, a death benefit will be paid immediately on the death of $(x)$, provided that $(x)$ dies between ages $x+u$ and $x + u + n$. So in this condition, the present value random variable ($p.v.\;r.v.$ later) is

$$Z = \begin{cases}
0 & \text{ if } T_x < u \text{ or } T_x \geq u+n \\ 
e^{-\delta T_x} & \text{ if } u \leq T_x < u+n
\end{cases}$$

And the actuarial notation for the EPV of the deferred term insurance benefit is

$$\begin{align}
{_{u\mid}}\bar{A}_{{x:\overline {n}|}}^{1} &= \int_{u} ^{u+n} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t \\
&\stackrel{s = t - u}= \int_{0} ^{n} e^{-\delta(s+u)} {_{s+u}}p_{x} \mu_{x+s+u} \, \mathrm{d}s \\
&= \boxed{ e^{-\delta u} {_{u}}p_{x} \int_{0} ^{n} e^{-\delta s} {_{s}}p_{x+u} \mu_{x+s+u} \, \mathrm{d}s \, }\\
&= \boxed{ e^{-\delta u} {_{u}}p_{x} \bar{A}_{x+u:\overline {n}|}^{1} \,} \\
&= \boxed{ {_{u}}E_{x}\bar{A}_{x+u:\overline {n}|}^{1} \,}
\end{align}$$

Another fomula is

$$\begin{align}
{_{u\mid}}\bar{A}_{{x:\overline {n}|}}^{1} &= \int_{u} ^{u+n} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t \\
&= \int_{0} ^{u+n} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t - \int_{0} ^{u} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t \\
&= \boxed{ \bar{A}_{{x:\overline {u+n}\displaystyle{|}}}^{1} - \bar{A}_{{x:\overline {u}|}}^{1}}
\end{align}$$

$\odot$ Thus, the EPV of a deferred term insurance benefit can be found by differencing the EPVs of term insurance benefits for terms $u + n$ and $u$. $\triangleleft$

Then we can use deffered term insurance polices to decompose the $n\text{-year}$ term insurance

$$\begin{align}
\bar {A}_{{x:\overline {n}|}}^{1} &= \int_{0}^{n} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t \\
&= \sum_{r=0} ^{n-1} \int_{r}^{r+1} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t \\
&= \boxed{ \sum_{r=0} ^{n-1} {_{r\mid}}\bar{A}_{{x:\overline {1}|}}^{1}}
\end{align}$$

and similarly for whole life insurance policy.

$$\begin{align}
\bar {A}_{x} &= \int_{0}^{\infty} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t \\
&= \sum_{r=0} ^{\infty} \int_{r}^{r+1} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \, \mathrm{d}t \\
&= \sum_{r=0} ^{\infty} {_{r\mid}}\bar{A}_{{x:\overline {1}\displaystyle{|}}}^{1}
\end{align}$$

In particular, we should note that

$$\begin{align} 
A_x &= \boxed{{A}_{{x:\overline {n}|}}^{1} + {_{n\mid}}A_x }  \\
{A}_{{x:\overline {n}|}}^{1} &= A_x - {_{n\mid}}A_x  \\
&= A_x - \upsilon^n {_{n}}p_{x} A_{x+n}
\end{align}$$

And we can also get that

$$\begin{align} 
{_{n\mid}}A_x &= A_x - {A}_{{x:\overline {n}|}}^{1}\\
&= \sum_{k=0}^{\infty} \upsilon^{k+1} {_{k\mid}}q_{x} - \sum_{k=0}^{n-1} \upsilon^{k+1} {_{k\mid}}q_{x} \\
&= \sum_{k=n}^{\infty} \upsilon^{k+1} {_{k\mid}}q_{x} \\
&= \upsilon^{n} \cdot {_{n}}p_{x} \cdot A_{x+n}
\end{align}$$

## Relating $\bar{A}_x$, $A_x$, and $A_{x}^{(m)}$
### UDD
$${_{s}}q_{y} \overset{\mathrm{UDD}}{ = } s\cdot q_{y} \Rightarrow \frac{\mathrm{d}} {\mathrm{d}s}\left({_{s}}q_{y}\right) = q_{y} \\
f_{y}(s) = {_{s}}p_{y} \cdot \mu_{y+s}$$
$$\Rightarrow {_{s}}p_{y} \cdot \mu_{y+s} \overset{\mathrm{UDD}}{ = } \:q_y$$

Then we have

$$\begin{align}
\bar{A}_x &= \int_{0}^{\infty} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \; \mathrm{d}t = \sum_{k=0} ^{\infty} \int_{k}^{k+1} e^{-\delta t} {_{t}}p_{x} \mu_{x+t} \; \mathrm{d}t \\
&\overset{t=k+s}{ = } \sum_{k=0} ^{\infty} \int_{0}^{1} e^{\delta}e^{-\delta s}e^{-\delta}e^{-\delta k} {_{k+s}}p_{x} \mu_{x+k+s} \; \mathrm{d}(k+s) \\
&= \sum_{k=0} ^{\infty} \int_{0}^{1}e^{(1-s)\delta}e^{-(1+k)\delta} {_{k}}p_{x}\cdot {_{s}}p_{x+k} \mu_{x+k+s} \mathrm{d}s \\
&= \sum_{k=0} ^{\infty} {_{k}}p_{x} \upsilon^{k+1} \int_{0}^{1}e^{(1-s)\delta} {_{s}}p_{x+k} \mu_{x+k+s} \mathrm{d}s \\
&\overset{\mathrm{UDD}}{ = }  \sum_{k=0} ^{\infty} {_{k}}p_{x} \cdot q_{x+k} \upsilon^{k+1} \int_{0}^{1}e^{(1-s)\delta} \mathrm{d}s =A_x \frac{e^{\delta} - 1} {\delta}
\end{align}$$

So finally since $e^{\delta} = 1 + i$, so that

$$\boxed{\bar{A}_x \approx \frac{i} {\delta}A_x}$$

Similarly, we have a table

$$\begin{array}{ccc} \hline
&\text{UDD Method} & \text{Similar idea} \\ \hline
\text{Whole life} & \bar{A}_x \approx \frac{i} {\delta}A_x & A_x^{(m)} = \frac{i} {i^{(m)}} \; A_x \\
n\text{-year term} &  \bar {A}_{{x:\overline {n}|}}^{1} \approx \frac{i} {\delta}{A}_{{x:\overline {n}|}}^{1} & {A^{(m)}}{_{x:\overline {n}|}^{1}} = \frac{i} {i^{(m)}} \; {A}_{{x:\overline {n}|}}^{1} \\
\text{Endowment} & \bar {A}_{{x:\overline {n}|}}\approx \frac{i} {\delta} \bar {A}_{{x:\overline {n}|}}^{1} + \upsilon^n {_{n}}p_{x} & A^{(m)}_{x:\overline {n}|} = \frac{i} {i^{(m)}} \; {A}_{{x:\overline {n}|}}^{1} + \upsilon^n {_{n}}p_{x} \\ \hline
\end{array}$$

And the deferred one: ${_{n|}}\bar{A}_x \approx \ffrac{i} {\delta} {_{n|}}{A}_x$

$\dagger$ The $i^{(m)}$ is defined by

$$1+i = \left( 1+ \frac{i^{(m)}} {m} \right)^m$$

### Using the claims acceleration approach
>**e.g.** From the relation of $A_x$ and $A_x^{(4)}$
>
>From year $K_x$ to $K_x + 1$, four different payment may happen for $A_x^{(4)}$, which are $K_x + 1/4$, $K_x + 2/4$, $K_x + 3/4$, $K_x + 4/4$. So that if we assume that the deaths occur evenly over the year, then the average payment time is

>$$\frac{1} {4} \left( K_x + 1/4 + K_x + 2/4 + K_x + 3/4 + K_x + 4/4\right) = K_x + \frac{5} {8}$$
>
>NOT $\ffrac{2} {4}$

$Conclusion$

For $K_{x}^{(m)}$, we have the average is 

$$\frac{1} {m} \left( K_x + 1/m + K_x + 2/m + \cdots + K_x + m/m\right) = K_x + \frac{m+1} {2m}$$

So that in general, for an $m-\text{thly}$ death benefit, assuming deaths are uniformly distributed over the year of age, the average time of payment of the death benefit is $( m + 1 )/ 2m$ in the year of death. So we have the approximation

$$\begin{align}
A_x^{(m)} &\approx q_x \cdot \upsilon^{\frac{m+1} {2m}} + {_{1|}}q_x \cdot \upsilon^{1+ \frac{m+1} {2m}} + {_{2|}}q_x \cdot \upsilon^{2+ \frac{m+1} {2m}} + \cdots \\
&= \sum_{k=0}^{\infty} {_{k|}}q_x \cdot \upsilon^{k+ \frac{m+1} {2m}} \\
&= \left( 1+i \right)^{\frac{m-1} {2m}} \sum_{k=0}^{\infty} {_{k|}}q_x \cdot \upsilon^{k+1}
\end{align}$$

That is 

$$\boxed{ A_x^{(m)} \approx \left( 1+i \right)^{\frac{m-1} {2m}} A_x }$$

$\dagger$ Note that

$$A_x^{(m)} = \sum_{k=0}^{\infty} {_{\frac{k} {m}|\frac{1} {m}}}q_x \cdot \upsilon^{(k+1)/m} \neq \sum_{k=0}^{\infty} {_{k|}}q_x \cdot \upsilon^{k+ \frac{m+1} {2m}}$$ $\ddagger$

And for the continuous benefit EPV, $\bar{A}_x$, we can just let $m \to \infty$ so that

$$\boxed{ \bar{A}_x \approx \left( 1+i \right)^{1/2} A_x }$$

This implies that assume that lives die uniformly through the year, the benefit paid immediately on death on average is half a year ealier than the benefit valued by $A_x$.

$$\boxed{ \bar{A}_{x:\overline{n}|} \approx (1+i)^{1/2} A^1_{x:\overline{n}|} + \upsilon^n \cdot {_{n}}p_x}$$

## Variable insurance benefits
Consider an event $E$, then the indicator random variables is defined as

$$I(E) = \begin{cases}
1 & \text{ if } E \text{ is true} \\ 
0 & \text{ if } E \text{ is not true} 
\end{cases}$$

Now we can see that we use the expectation of this variable to replace the probability of the event.

$$\newcommand{\EE}{\mathbb{E}}\EE[I(E)] = \Pr[E]$$

However in continuous condition, the indicator variable is defined as $I(t < T_x \leq t + \mathrm{d}t)$, so that 

$$\begin{align}
\EE[I(t < T_x \leq t + \mathrm{d}t)] &= \Pr[t < T_x \leq t + \mathrm{d}t] \\
&= \Pr[t < T_x] \cdot \Pr[T_x \leq t + \mathrm{d}t | t < T_x] \\
&\approx {_{t}}p_{x} \cdot \mu_{x+t} \mathrm{d} t
\end{align}$$

## Delta Neutral Strategy
Risk exposure $H$ per one call option is $H = c\P{0} - c\P{T} +y\P{S_T-S_0}$. Here the ***delta hedging*** means: *to hedge your position by holding the stock with the number of shares equal to the option's delta*.

Under the **single-period binomial tree** model, $H = x-xe^{rT}$, and with more periods, we can hedge dynamically. Then the **continuous-time** model, we define the ***Delta***:

$$\Delta_c \equiv \ffrac{\partial c\P{t,S_t}}{\partial S_t}$$

Commonly, an option written on an underlying asset $S$ is most sensitive to the changes in the value of $S$. Hence, the largest part of the risk comes from the price movements of asset $S$.

And the **first order approximation** for $c$:

$$c\P{t,S_t} \approx c\P{0,S_t} + \Delta_c\P{S_t-S_0}$$

Then the ***Delta*** of a portfolio: *the change in the value of the portfolio divided by the change in the value of the underlying asset*.

And the position with a **Delta** of $0$ is called ***delta neutral***, meaning that the portfolio doesn't change if the price of the underlying asset changes.

**e.g.**

We are short $c\P{t,S_t}$ and we want to hedge the associated risk with respect to the movements of $S_t$ by holding a position in $S_t$. We should find the number of shares of $S_t$ we should hold, namely $y$, in the complete portfolio $\Pi$ in order to make $\Delta_\Pi = 0$:

$$\Delta_\Pi = -\Delta_c + y \times 1 = 0 \Rightarrow y = \Delta_c$$
***

A portfolio is called ***long*** (***short***) ***Delta*** if its **Delta** is positive, $\Delta_\Pi > 0$ (negative with $\Delta_\Pi < 0$), in which case its value will *increase* (*decrease*) if the value of $S_t$ increases, all other variables being constant.

***Static Hedging***: The hedge is set up initially and *never adjusted*. That's exactly what we do under the **single-period binomial model**, where a perfect replication of a payoff $c\P{T,S_T}$ is achieved by holding exactly $\Delta_c$ shares of the underlying at the initial time. All the risk is eliminated keeping the portfolio **Delta-Neutral** under this model. This result is not necessarily true in other models. However in practice, $S_t$ changes frequently thus to neutralize the **Delta**, we need to adjust the number of shares of $S_t$ held in the portfolio. We called this the ***dynamic hedging*** strategy.

And for the **perfect hedge**, rebalancing must take *continuously*. This requires that the model and its parameters are exactly correct.

## Delta under the Black-Scholes Model
The **Black-Scholes-Mertion equation** is derived by setting up a **delta-neutral position** (*short* $1$ option and *long* $\Delta_c$ shares of the stock) and arguing that the return on the position should be the *risk-free interest rate*.

For the European call option, BS equation gives

$$c\P{t,S_t} = S_t\N{d_1} - Ke^{-r\P{T-t}}\N{d_2}\\
d_{1,2} = \ffrac{1}{\sigma\sqrt{T-t}}\P{\log\ffrac{S_t}{K}+\P{r\pm\ffrac{\sigma^2}{2}}\P{T-t}}$$

and we have $\Delta_c = \N{d_1}$ and then using the put-call parity we have $\Delta_p = \N{d_1} -1$. What's more

- $\Delta_c > 0$: the **replicating** portfolio of the call, never goes short in the stock
- $\Delta_c < 1$: holding one share or more of the stock is *more than enough* to **hedge** the call option payoff.
- $\Delta_p < 0$: the **replicating** portfolio, of the put, never goes long in the stock
- $\Delta_p > -1$: holding one share or more of the stock is *more than enough* to **hedge** the put option payoff.

## Option Sensitivities
*Partial derivatives* of the portfolio value with respect to parameters are called ***Greeks***. The five standard **Greeks** for a *portfolio* with value $V$ are

$$\begin{array}{c|ccccc}\hline
\text{name} & \text{Delta} & \text{Theta} & \text{Gamma} & \text{Vega} & \text{Rho}\\ \hline
\text{expression} & \Delta = \ffrac{\partial V}{\partial S_t} & \Theta = \ffrac{\partial V}{\partial t} & \Gamma = \ffrac{\partial^2 V}{\partial S_t^2} & {\large\nu} = \ffrac{\partial V}{\partial \sigma} & {\large\rho} = \ffrac{\partial r}{\partial S_t}\\\hline
\end{array}$$

### $\Theta$, derivative with respect to time
**Theta** is sometimes referred to as the ***time decay*** of the portfolio.

$$\begin{align}
\Theta_c &= S_t\mathcal{N}'\P{d_1}\ffrac{\partial d_1}{\partial t} - rKe^{-r\P{T-t}}\N{d_2}-Ke^{-r\P{T-t}}\mathcal N'\P{d_2}\ffrac{\partial d_2}{\partial t}\\
\Theta_p &= rKe^{-r\P{T-t}}\N{-d_2}-Ke^{-r\P{T-t}}\mathcal N'\P{-d_2}\ffrac{\partial d_2}{\partial t} + S_t\mathcal{N}'\P{-d_1}\ffrac{\partial d_1}{\partial t} 
\end{align}$$

and using the facts that $S_t\mathcal{N}'\P{\pm d_1} = Ke^{-r\P{T-t}}\mathcal{N}'\P{\pm d_2}$ and $\ffrac{\partial d_1}{\partial t} = \ffrac{\partial d_2}{\partial t} - \ffrac{\sigma}{2\sqrt{T-t}}$ we have

$$\begin{align}
\Theta_c &= - \ffrac{\sigma}{2\sqrt{T-t}}S_t\mathcal{N}'\P{d_1} - rKe^{-r\P{T-t}}\N{d_2}\\
\Theta_p &= - \ffrac{\sigma}{2\sqrt{T-t}}S_t\mathcal{N}'\P{d_1} + rKe^{-r\P{T-t}}\N{-d_2}
\end{align}$$

Usually $\Theta$ is negative, since option tends to be less valuable.

### $\Gamma$, second order derivative with respect to asset price
If the *absolute value* of $\Gamma$ is large, $\Delta$ would be highly sensitive to the price of the underlying asset, meaning that adjustments to keep a portfolio **delta neutral** need to be made relatively frequently. 

And since $\Delta_c = \N{d_1}$ and $\Delta_p = \N{d_1}-1$, we have, for European call/put

$$\Gamma = \ffrac{\partial \N{d_1}}{\partial S_t} = \mathcal{N}'\P{d_1} \ffrac{1}{\sigma\sqrt{T-t}}\ffrac{1}{S_t}>0$$

Up to this point, we can already draw two very useful conclusions.

### Approximation
Using **Taylor Expansion**:

$$\begin{align}
V\P{t+\Delta t,S_{t+\Delta t}} &= V\P{t,S_t} + \Delta t \ffrac{\partial V\P{t,S_t}}{\partial t} + O\P{\P{\Delta t}
^2} + \P{S_{t+\Delta t} - S_t}\ffrac{\partial V\P{t,S_t}}{\partial S_t} \\
&\bspace + \ffrac{1}{2}\P{S_{t+\Delta t} - S_t}^2 \ffrac{\partial^2 V\P{t,S_t}}{\partial S_t\partial S_t}+\cdots\\
&\approx V\P{t,S_t} + \color{brown}{\Theta}\Delta t + \color{brown}{\Delta} \P{S_{t+\Delta t} - S_t} + \ffrac{1}{2}\color{brown}{\Gamma} \P{S_{t+\Delta t} - S_t}^2
\end{align}$$

Further, if the portolio is **Delta Neutral**, then

$$V\P{t+\Delta t,S_{t+\Delta t}} \approx V\P{t,S_t} + \color{brown}{\Theta}\Delta t + \ffrac{1}{2}\color{brown}{\Gamma} \P{S_{t+\Delta t} - S_t}^2$$

### Reaching Gamma Neutral
Suppose a **Delta Neutral** portfolio has $\Gamma$ as its **Gamma**, and another traded opion has **Gamma** equal to $\Gamma_T$. If the number of traded options added to the portfolio is $w_T$, gamma of the new portfolio is $w_T \Gamma_T + \Gamma$ and thus buying $w_T = -\ffrac{\Gamma}{\Gamma_T}$ would lead to **Gamma Neutral**. 

However this changes the $\Delta$ of the portfolio, we need to trade more shares to neutralize our $\Delta$ after that.

**e.g.**

Suppose that a portfolio is **Delta Neutral** and has a $\Gamma$ of $-3000$. The $\Delta_c$ and $\Gamma_c$ of a particular traded call option are $0.62$ and $1.50$, respectively. How to make the portfolio **Delta Neutral** and **Gamma Neutral**?

> The portfolio can be made **Gamma Neutral** by including in the portfolio a *long position* of $-\ffrac{\Gamma}{\Gamma_c} = -\ffrac{-3000}{1.5} = 2000$ call option. However, the **Delta** of the portfolio will then change fram $0$ to $\Delta_v = 2000\times 0.62 = 1240$. Thus, $1240$ units of the underlying asset must be *sold* from the portfolio to keep it **Delta Neutral**.
***

***delta hedging error***: when the stock price moves from $S_{1}$ to $S_{2}$ the option price moves from $C_1$ to $C_2$, while **delta hedging** assumes that it moves to $C_2'$. The difference between $C_2$ and $C_2'$ is the **delta hedging error**.

- **Delta neutrality** provides protection against *relatively small* stock price moves between rebalancing.
- **Gamma neutrality** provides protection against *larger* movements in this stock price between hedge rebalancing.

### $\large\nu$, derivative with respect to volatility

For a European call or put option, **Vega** is given by

$$\begin{align}{\large{\nu}} &= S_t \mathcal N'\P{d_1}\ffrac{\partial d_1}{\partial \sigma} - Ke^{-r\P{T-t}}\mathcal N'\P{d_2}\ffrac{\partial d_2}{\partial \sigma}\\
&= S_t \mathcal N'\P{d_1}\ffrac{\partial d_1}{\partial \sigma} - S_t \mathcal N'\P{d_1}\P{\ffrac{\partial d_1}{\partial \sigma} - \sqrt{T-t}}\\
&= S_t \sqrt{T-t} \mathcal N'\P{d_1}>0
\end{align}$$

Thus, both the European call and put prices *increase with increasing volatility*. And this doesn't violate our intuition. Higher volatility lead to a higher chance that the option may end up either **deeper in-the-money** or **deep out-of-the-money**, while there's no extra peanlty for the option to be deeper out-of-the-money but the payoff increases when the option expires deeper in the money.

### $\large\rho$, derivative with respect to interest rate

$$\begin{align}
{\large\rho}_c &= S_t \mathcal N'\P{d_1}\ffrac{\partial d_1}{\partial r} + K\P{T-t}e^{-r\P{T-t}}\N{d_2} - Ke^{-r\P{T-t}}\mathcal N'\P{d_2}\ffrac{\partial d_2}{\partial r}\\
&= K\P{T-t}e^{-r\P{T-t}}\N{d_2} + S_t \mathcal N'\P{d_1}\ffrac{\partial d_1}{\partial \sigma} - S_t \mathcal N'\P{d_1}\P{\ffrac{\partial d_1}{\partial \sigma} - 0}\\
&= K\P{T-t}e^{-r\P{T-t}}\N{d_2} > 0\\
{\large\rho}_p &= -K\P{T-t}e^{-r\P{T-t}} \N{-d_2} < 0
\end{align}$$

A *higher interest* rate lowers the present value of the cost of exercising the European call option at expiration (the effect is similar to the *lowering of the strike price*), in turn this increases the call price. Reverse effect holds for the put option price.

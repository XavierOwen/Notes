# Volatility Smile
In the Black-Scholes model, note that the *volatility* $\sigma$ is the only unobservable parameter, we need to consider an estimate of the volatility. We'll some covered material in Chap_04.
1. **Historic volatility**: use historical data for a period of the *same length as the time to maturity* to find the volatility. However, to price, the input value of $\sigma$ should be the *forecast volatility* of the value *over the remaining life of the option* rather than an estimated volatility value from the *past market data* of the *asset price* (which is the **historic volatility**).
2. **Implied volatility**: the value of $\sigma$ that matches the *theoretical price from BS equation* with the *observed market price* of the option. Mathematically,$\\[0.5em]$

## Volatility Smiles
A plot of the **implied volatility** $\sigma_{\text{imp}}$ of an option as a function of its **strike price** $K$ is known as a ***volatility smile***. 

![](./figs/volatilitySmile.png)

### For Calls and Puts
First, the parity: $p + S_0 \color{brown}{e^{-qT}} = c + Ke^{-rT}$. then

$$\left.\begin{align}
p_{\text{BS}} + S_0 \color{brown}{e^{-qT}} &= c_{\text{BS}} + Ke^{-rT}\\
p_{\text{imp}} + S_0 \color{brown}{e^{-qT}} &= c_{\text{imp}} + Ke^{-rT}
\end{align}\right\} \Rightarrow p_{\text{BS}} - p_{\text{imp}} = c_{\text{BS}} - c_{\text{imp}}$$

and this implies that $\sigma_{\text{imp}}$ are the same for both: suppose that the **implied volatility** of the put option is $22\%$, then $p_{\text{BS}}\P{t,S_t,T,r,22\%,K} = p_{\text{mkt}}$ and thus $c_{\text{BS}}\P{t,S_t,T,r,22\%,K} - c_{\text{mkt}} = 0$. Namely, the **implied volatility** of the call option is also $22\%$.

### Risk neutral density function
Let $g\P{T,S_T,t,S_t}$ (or simply $g\P{S_T}$) denote the transition density function of the asset price, from $T$ to $t$. The time-$t$ price of a *European call* with **maturity date** $T$ and **strike price** $K$ is given by

$$c\P{t,S_t,T,K} = e^{-r\P{T-t}}\int_K^\infty \P{S_T-K} g\P{S_T}\;\dd S_T$$

If we differentiate $c$ with respect to $K$, twise, we obtain 

$$\ffrac{\partial c}{\partial K} = -e^{-r\P{T-t}}\int_K^\infty g\P{S_T}\;\dd S_T, \bspace \ffrac{\partial^2 c}{\partial K^2} = e^{-r\P{T-t}} g\P{K}$$

$$\Rightarrow\color{brown}{g\P{T,K,t,S_t} = g\P{K} = e^{r\P{T-t}}\ffrac{\partial^2 c}{\partial K^2}}$$

We introduce this function to declare that:

$\bspace$Suppose that market European option prices *at all strikes* are available, the **risk neutral density function** can be inferred *completely* from the *market prices* of options with *the same **maturity*** and
*different **strikes***, *without* knowing the **volatility** function.

**e.g.**

Assume that $c_1$, $c_2$ and $c_3$ are the prices of $T$-year *European call options* with **strike** $K-\delta$, $K$ and $K+\delta$, respectively. Assuming $\delta$ is small, then an estimate of $g\P K$ is

$$g\P{T,K,t,S_t} = g\P K = e^{r\P{T-t}} \ffrac{\partial^2 c}{\partial K^2} \approx e^{r\P{T-t}} \ffrac{c_1 + c_3 - 2c_2}{\delta^2}$$

$Remark$

>$$\ffrac{\partial f\P x}{\partial x} \approx \ffrac{f\P{x+\Delta x} - f\P x}{\Delta x} \Rightarrow \ffrac{\partial^2 f\P x}{\partial x^2} \approx \ffrac{f\P{x+\Delta x} - f\P x - \P{f\P{x} - f\P{x-\Delta x}}}{\Delta x^2}$$

## Options

The bounds:

$$\begin{align}
&\max\P{S_0-Ke^{-rT},0}\leq c\leq C \leq S_0\\
&\max\P{Ke^{-rT}-S_0,0}\leq p\leq P \leq K
\end{align}$$

The put-call parity, $c+Ke^{-rT} = p+S_0$; and if there's dividend paid, it becomes

$$c+Ke^{-rT} = p+S_0e^{-qT}\Rightarrow c\P{t,S_t} +Ke^{-r\P{T-t}} = p\P{t,S_t}+S_te^{-q\P{T-t}}$$

If American options are involved, we have

$$c+Ke^{-rT}-S_0 = p \leq P\leq c+K-S_0 = C+K-S_0$$

All preceding can be proved by constructing two portfolios with certain payoffs at time $T$.

## Binomial Tree

**Arbitrage**: NOTHING at time $0$, all possible payoffs at $T$ is non-negative, the expectation of the payoff under the actual probability measure, is positive; or mathematically

- $V_0^h = 0\\[0.5em]$
- $V_T^h\P\omega \geq0\\[0.5em]$
- $\Exp{V_T^h\P\omega}>0$

**Dynamic replication**: or **No arbitrage**: Let $c$ denote a European call option price at time $0$ (or $V\P{0,S_0}$ if necessary); and $c_u$, $c_d$ the call option price at time $T$. Then we construct the portfolio $h$ that replicates the *long position* of the call option (derivative), with $x$ units of bonds and $y$ shares of stocks. Denote $c_u = \P{S_T-K}^+$ as the payoff from the call (derivative) if the price moves up

$$\begin{cases}
x \cdot e^{rT} + y\cdot uS_0 = c_u\\
x \cdot e^{rT} + y\cdot dS_0 = c_d
\end{cases} \Rightarrow \begin{cases}
x = \ffrac{u\cdot c_d - d\cdot c_u}{\P{u-d}e^{rT}}\\[0.6em]
y = \ffrac{c_u - c_d}{\P{u-d}S_0}
\end{cases}$$

then $c = x+yS_0 = e^{-rT}\P{\ffrac{e^{rT}-d}{u-d}c_u + \ffrac{u - e^{rT}}{u-d}c_d}$

***

**Riskless hedging principle**: we now need a portfolio $h$ with a *short* position of $\Delta$ shares of stocks and $1$ units of the call option (derivative), so that the *terminal payoff* of $h$ is a constant (riskless).

$$c_u - \Delta \cdot u\cdot S_0 = c_d - \Delta \cdot d\cdot S_0 = \P{c-\Delta \cdot S_0}e^{rT}$$

***

**Risk-neutral valuation**: As we can assume the existence of the risk-neutral probability value $q_u$ and $q_d$, we first find them.

$$S_0 = e^{-rT}\ExpH^\QQ\SB{S_T} = e^{-rT}\P{q_u\cdot uS_0 + \P{1-q_u}dS_0},\bspace q_d = 1-q_u$$

After obtaining the two we can calculate the price of the call option (derivative)

$$c = e^{-rT} \P{q_u\cdot c_u + q_d \cdot c_d}\equiv \ExpH^\QQ\SB{c\P{T,S_T}}$$

**Relationship**

- Let $\Delta = y$ and we'll see that the first two methods are equivalent
- Write $\ffrac{e^{rT}-d}{u-d} = q_u$ and we'll see the first and the last methods are equivalent

***

**Multiperiod binomial model**

Let $\P{t,k}$ denote the node at time $t$ with $k$ steps up and $\tau$ the time unit. Totally $T$ units of time.

1. step by step: $u = \ffrac{S_t}{S_{t-1}}$, then $c_t\P{k}=e^{-r\tau}\ExpH^\QQ\SB{q \cdot c_{t+1}\P{k+1}+\P{1-q} \cdot c_{t+1}\P{k}}$
2. path probability: 

$$\begin{align}
c &= e^{-rT}\ExpH^\QQ\SB{c\P{T,S_T}}\\
&= e^{-rT}\SB{\binom{T}{T}q^T c_T\P{T} + \binom{T}{T-1}q^{T-1}\P{1-q}c_T\P{T-1} + \binom{T}{T-2}q^{T-2}\P{1-q}^2c_T\P{T-2}+\cdots}
\end{align}$$

and a little bit modification for European put option. As for American options, be careful.

***

**Replication**: or **Hedging**. For *Multiperiod binomial model*, we need to dynamically replicate at each time period. Here's a $Proposition$: Consider a claim $U = \Phi\P{S_T}$. Then this claim can be replicated using a self-financing portfolio. Still $V_t\P{k}$ denotes the option price at the node $\P{t,k}$, where $k$ is the number of up-moves that have occurred, then $V_t\P{k}$ can be computed recursively by the scheme:

$$\begin{align}
V_T\P{k} &= \phi\P{S_T} \\
V_t\P{k} &= e^{-r \times 1}\SB{q V_{t+1}\P{t+1} + \P{1-q} V_{t+1}\P{k}}
\end{align}$$

and to hedge this, we construct the portfolio as 

$$y_t\P{k} = \ffrac{V_{t+1}\P{k+1} - V_{t+1}\P{k}} {\P{u-d}S_t},\bspace x_t\P{k} = V_t\P k - y_t\P k \cdot S_t = V_{t}\P{k}-\ffrac{V_{t+1}\P{k+1} - V_{t+1}\P{k}} {u-d}$$

here $\P{x_t\P{k},y_t\P{k}}$ for the amount of stock and bond held in the portfolio at node $(t, k)$ over the time interval $[t, t + 1)$

## Stochastic Process

**Martingale**

1. $\forall\, t$, $\ExpH^\PP\SB{\abs{X_t}}<\infty$. Try using the fundamental inequlity: $\Exp{\abs{W_t}}\leq\Exp{\ffrac{1+W_t^2}{2}} = \ffrac{1+t}{2}$
2. $\forall\, s\leq t$, $\ExpH^\PP\SB{X_{t}\mid\FcF_s}\equiv \ExpH^\QQ_s\SB{X_t} = X_s$
3. $\CB{X_t}_{t\geq 0}$ is a stochastic process adapted to $\CB{\FcF_t}_{t\geq 0}$

***

**Brownian motion**

1. $W_t-W_s\sim\N{0,t-s}$, $\P{t-s}^{-1}\P{W_t-W_s}^2\sim\chi^2\P1$, $\Exp{\chi^2\P k} = k$ and $\Var{\chi^2\P k} = 2k$
2. $\Exp{\exp\CB{\mu t + \sigma W_t}} = \exp\CB{\mu t + \ffrac{1}{2}\sigma^2 \cdot t}$
3. $\ExpH_s\SB{W_t} = W_s + \ExpH_s\SB{W_{t-s}} = W_s + \Exp{W_{t-s}} = W_s$

***

**quadratic variation**

$$\SB{W}_t = \lim_{\norm{\Pi}\to0}Q_\Pi = \lim_{\norm{\Pi}\to0}\sum_{j=0}^{n-1} \P{W_{t_{\void_{j+1}}} - W_{t_{\void_j}}}^2 $$

1. $\SB{W}_t = t$, by verifying the expectation is $t$ and variance is $0$
2. Write $\d{t = \lim_{\norm{\Pi}\to0}Q_\Pi = \lim_{\norm{\Pi}\to0}\sum_{j=0}^{n-1} \P{W_{t_{\void_{j+1}}} - W_{t_{\void_j}}}^2 = \int_0^t \;\P{\dd W_s}^2}$, then $\P{\dd W_t }^2 = \dd t$
3. Similarly, $\d{0 = \lim_{\norm{\Pi}\to0} \sum_{j=0}^{n-1} \P{t_j - t_{j-1}}^2 = \int_0^t \;\P{\dd s}^2}$ and thus $\dd t \dd t = 0$
4. Using Cauchy-Schwartz inequality, $\dd W_t\dd t = 0$

***

**Itô**

$$I\P t = \int_0^t h\P{s,W_s}\;\dd W_s,\bspace I_n\P t =\sum_{j=0}^{n-1} h\P{t_j,W_{t_{\void_j}}}\P{W_{t_{\void_{j+1}}} - W_{t_{\void_j}}}$$

1. $\ExpH_s\SB{I\P t-I\P s} = \ExpH_s\SB{\d{\lim_{n\to\infty} \sum_{j=0}^{n-1} h\P{t_j,W_{t_{\void_j}}}\P{W_{t_{\void_{j+1}}} - W_{t_{\void_j}}}}} = 0$ and thus $I\P t$ is a martingale
2. Itô isometry: $\Exp{I\P t^2} = \d{\int_0^t \Exp{h^2\P{s,W_s}}\;\dd s}$
3. **Itô process**: $\dd X_t = \mu\P{t,X_t}\dd t + \sigma\P{t,X_t}\dd W_t$
4. **Itô's formula**: for $Y_t = f\P{t,X_t}$ we have
$$\begin{align}\dd f &= \ffrac{\partial f}{\partial t}\dd t + \ffrac{\partial f}{\partial X_t}\dd X_t + \ffrac{1}{2}\ffrac{\partial^2 f}{\partial X_t^2}\P{\dd X_t}^2\\
&= \P{\ffrac{\partial f}{\partial t} + \mu\P{t,X_t}\ffrac{\partial f}{\partial X_t} + \ffrac{1}{2} \sigma^2 \P{t,X_t} \ffrac{\partial^2 f}{\partial X_t^2}}\dd t + \sigma\P{t,X_t}\ffrac{\partial f}{\partial X_t}\dd W_t
\end{align}$$

**e.g.**

The basic reference model for stock price in continuous time is

$$\ffrac{\dd S_t}{S_t} = \mu \dd t + \sigma \dd W_t \P{\neq \dd \log S_t}$$

First write $\dd S_t = \mu \cdot S_t \dd t + \sigma \cdot S_t \dd W_t \equiv \mu_t \dd t + \sigma_t \dd W_t$. Then using the Itô's formula we have

$$\begin{align}
\dd \log S_t &= \dd f\P{S_t} \\
&= \P{0 + \P{\mu S_t}\ffrac{1}{S_t} + \ffrac{1}{2}\P{\sigma S_t}^2 \P{-\ffrac{1}{S_t^2}}}\dd t + \P{\sigma S_t} \ffrac{1}{S_t} \dd W_t \\
&= \P{\mu - \ffrac{1}{2}\sigma^2}\dd t + \sigma \dd W_t
\end{align}$$
***
Here we can also apply the first form of Itô's formula, which yields

$$\begin{align}
\dd \log S_t &= \ffrac{1}{S_t} \dd S_t - \ffrac{1}{2S_t^2} \P{\dd S_t}^2 \\
&= \mu \dd t + \sigma \dd W_t - \ffrac{1}{2}\P{\ffrac{\dd S_t}{S_t}}^2\\
&= \mu \dd t + \sigma \dd W_t - \ffrac{1}{2}\sigma^2\dd t = \P{\mu - \ffrac{1}{2}\sigma^2}\dd t + \sigma \dd W_t\\
\end{align}$$

take the intergral, we have

$$\log\ffrac{S_t}{S_0} = \P{\mu - \ffrac{\sigma^2}{2}}t + \sigma W_t$$

$$\begin{align}
S_t &= S_0 \exp\CB{\P{\mu - \ffrac{\sigma^2}{2}}t + \sigma W_t}\\
&= S_s \exp\CB{\P{\mu - \ffrac{\sigma^2}{2}}\P{t-s} + \sigma \P{W_t-W_s}}
\end{align}$$

***

**Feynman-Kac**

Assume the function $f$ solves the boundary value problem:

$$\begin{cases}
-rf\P{t,x} +\ffrac{\partial f} {\partial t}\P{t,x} + \mu\P{t,x} \ffrac{\partial f} {\partial x}\P{t,x} + \ffrac{1}{2}\sigma^2 \P{t,x} \ffrac{\partial^2 f} {\partial x^2}\P{t,x} = 0\\[0.6em]
f\P{T,x} = \Phi\P{x}
\end{cases}$$

And define $\CB{X_t}_{0\leq t\leq T}$ to be the solution of the ***SDE***: $\dd{X_t} = \mu\P{t,X_t} \dd{t}+\sigma\P{t,X_t} \dd{W_t}$. Then under some regularity condition, we have

$$f\P{t,x} = \Exp{e^{-r\P{T-t}}f\P{T,X_T}\mid X_t = x} = \Exp{e^{-r\P{T-t}}\Phi\P{X_T}\mid X_t = x}$$

The two equation, are equivalent.

**e.g.**

Use the Feynman-Kac stochastic representation formula to solve

$$\ffrac{\partial F\P{t,x}}{\partial t} + \ffrac{1}{2}\sigma^2 \ffrac{\partial^2 F\P{t,x}}{\partial x^2} = 0$$

subject to the *terminal value condition* $F\P{T,x} = e^x$

>By F-K formula, we have the solution of the corresponding stochastic differential equation
>
>$$\dd X_t = \mu_t\dd t + \sigma_t \dd W_t$$
>
>where $\mu_t = 0$, $r = 0$, and $\sigma_t = \sigma$. So that the solution is, in integral form,
>
>$$X_t = \sigma W_t + X_0$$
>
>So that the solution $F\P{t,x}$
>
>$$\begin{align}
F\P{t,x} &= \Exp{e^{-r\P{T-t}} e^{X_T} \mid X_t = x} \\
&= \Exp{e^{-0\P{T-t}} e^{\sigma W_T} \mid \sigma W_t = x} \\
&= \Exp{e^{\sigma\P{ W_T - W_t} +x} \mid \sigma W_t = x}\\
&= e^{\sigma x}\Exp{e^{\sigma\P{ W_T - W_t}}}\\
&= \exp\CB{ x + \ffrac{\sigma^2}{2}\P{T-t}}
\end{align}$$

***

**Girsanov's Theorem**

Suppose that $\CB{W_t}_{t \geq0}$ is a $\PP$-Brownian motion with the natural filtration $\CB{\FcF_t}_{t\geq 0}$ and $\CB{\theta_t}_{t\geq0}$ is an $\CB{\FcF_t}_{t\geq0}$-adapted process such that

$$\Exp{\exp\CB{\ffrac{1}{2}\int_{0}^{T} \theta_t^2 \;\dd{t}}}<\infty$$

Then we define $L_t = \exp\CB{\d{\mathbf-\int_{0}^{t} \theta_s \;\dd{W_s} \mathbf-\ffrac{1}{2}\int_{0}^{t}\theta_s^2 \;\dd{s}}}$ and let $\tilde P$ be the probability measure defined by $\tilde P\SB{A} = \d{\int_A L_t\P{\omega} \PP\P{\dd{\omega}}}$. Then under the new probability measure $\tilde P$, the process defined by $\widetilde W_t = W_t \mathbf +\d{\int_{0}^{t}\theta_s\;\dd{s}}$ is a standard Brownian motion.

1. $\tilde P\SB{A} = \d{\int_A L_t\P{\omega} \PP\P{\dd{\omega}}} \iff L_t = \given{\ffrac{\dd \tilde P}{\dd \PP}}_{\FcF_t}$, and $L_t$ is a martingale
2. $\ExpH^{\tilde P} \SB{\Phi\P{t}\mid \FcF_s} = \ExpH^\PP\SB{\Phi\P{t}\ffrac{L_t}{L_s}\mid\FcF_s}$, or shortly $\ExpH^\tilde P_s\SB{\Phi\P t} = \ExpH^\PP_s\SB{\Phi\P t\ffrac{L_t}{L_s}}$
3. $\ExpH^\tilde P_s\SB{\Phi\P t\ffrac{L_s}{L_t}} = \ExpH^\PP_s\SB{\Phi\P t}$

**e.g.**

The basic reference model for stock price in continuous time is

$$\dd S_t = \mu S_t \dd t + \sigma S_t \dd W_t$$

where $\CB{W_t}_{t\geq 0}$ is a $\PP$-Brownian motion. Find a measure under which the discounted stock price $\CB{\tilde S_t}_{t\geq 0}$ is a martingale. Here, $\tilde S_t = e^{-rt}S_t$.

>By Itô's formula, under $\PP$ measure, we have
>
>$$\dd \tilde S_t =\tilde S_t \P{-r + \mu} \dd t + \sigma \tilde S_t \dd W_t$$
>
>With the intention to *zero* the **drift** $\mu_t$, we let $\widetilde W_t = W_t + \ffrac{\mu - r}{\sigma}t$:
>
>$$\dd \tilde S_t =  \sigma \tilde S_t \dd \widetilde W_t$$
>
>According to Girsanov’s Theorem, we can define, by letting $\theta = \ffrac{\mu-r}{\sigma}$
>
>$$\newcommand{\QQ}{\mathbb{Q}}
\left.\ffrac{\dd\QQ}{\dd\PP}\right|_{\FcF_t} = \exp\CB{-\theta W_t - \ffrac{1}{2}\theta^2 t}$$
>
>Then $\CB{\widetilde W_t}_{t\geq0}$ is a $\QQ$-Brownian motion and the discounted stock price $\CB{\tilde S_t}_{t\geq 0 }$ is a $\QQ$-martingale.

$Remark$

>When $S_t$ is multiplied by $e^{rt}$, the resulting process can be remembered as a conclusion that
>
>$$\dd \tilde S_t =\tilde S_t \P{r + \mu} \dd t + \sigma \tilde S_t \dd W_t$$
>
>The **drift** is increased by $r$, while the **diffusion** (volatility) remains the same.
>
>And this can be used to prove $S_t$ is martingale, given the adaptivity and finite expecation of the absolute value.

## Black-Scholes-Merton model

$$\dd S_t = \mu S_t \dd t + \sigma S_t \dd W_t,\bspace S_t = S_0 \cdot \exp\CB{\P{\mu - \ffrac{\sigma^2}{2}}t +\sigma W_t}$$

- **Dynamic Replication**: or **No arbitrage**, first, the wealth process$\\[1em]$
$$\dd V_t = \ffrac{\pi_t}{S_t} \dd S_t + \ffrac{V_t - \pi_t}{B_t}\dd B_t = \SB{rV_t + \P{\mu-r} \pi_t}\dd t + \sigma\pi_t \dd W_t\\[0.5em]$$
then use **Itô formula** on $c\\[0.5em]$
$$\dd c = \P{\ffrac{\partial c}{\partial t} + \mu S_t \ffrac{\partial c}{\partial S_t} + \ffrac{1}{2}\sigma^2 S_t^2 \ffrac{\partial^2 c}{\partial S_t^2}}\dd t + \sigma S_t\ffrac{\partial c}{\partial S_t}\;\dd W_t\\[1em]$$
Comparing the $\dd W_t$ term, we have $\ffrac{\pi_t}{S_t} = \ffrac{\partial c}{\partial S_t}$ and comparing the $\dd t$ term we have$\\[1.3em]$
$$\ffrac{\partial c}{\partial t} + \mu S_t \ffrac{\partial c}{\partial S_t} + \ffrac{1}{2}\sigma^2 S_t^2 \ffrac{\partial^2 c}{\partial S_t^2} = rc + \P{\mu-r} \pi_t \Rightarrow \ffrac{\partial c}{\partial t} + r S_t \ffrac{\partial c}{\partial S_t} + \ffrac{1}{2}\sigma^2 S_t^2 \ffrac{\partial^2 c}{\partial S_t^2} - rc = 0\\[2em]$$

- **Riskless Hedging Principle**: $\Pi\P{t,S_t} = c\P{t,S_t}-\Delta S_t$. Using Itô,$\\[1em]$
$$\begin{align}
\dd\Pi\P{t,S_t}&=\P{\ffrac{\partial c}{\partial t} + \mu S_t \ffrac{\partial c}{\partial S_t} + \ffrac{1}{2}\sigma^2 S_t^2 \ffrac{\partial^2 c}{\partial S_t^2}}\dd t + \sigma S_t\ffrac{\partial c}{\partial S_t}\;\dd W_t - \Delta \P{\mu S_t\dd t + \sigma S_t \dd W_t}\\
&= \P{ \ffrac{\partial c}{\partial t} + \mu S_t\P{\ffrac{\partial c}{\partial S_t} - \Delta} + \ffrac{1}{2}\sigma^2 S_t^2 \ffrac{\partial^2 c}{\partial S_t^2}}\dd t + \sigma S_t\P{\ffrac{\partial c}{\partial S_t} - \Delta}\dd W_t
\end{align}\\[1.2em]$$
Then from the **riskless hedging principle**, we have $\dd\Pi\P{t,S_t} = r\cdot\Pi\P{t,S_t}\dd t$ and thus$\\[1em]$
$$\P{\ffrac{\partial c}{\partial t} + \mu S_t\P{\ffrac{\partial c}{\partial S_t} - \Delta} + \ffrac{1}{2}\sigma^2 S_t^2 \ffrac{\partial^2 c}{\partial S_t^2}}\dd t + \sigma S_t\P{\ffrac{\partial c}{\partial S_t} - \Delta}\dd W_t = r\cdot\Pi\P{t,S_t}\dd t = r\cdot\P{c-\Delta S_t}\dd t\\[1em]$$
Mathing the terms, we have
$$\ffrac{\partial c}{\partial t} + \ffrac{1}{2} \sigma^2 S_t^2 \ffrac{\partial^2 c}{\partial S_t^2} + rS_t \ffrac{\partial c}{\partial S_t} - rc = 0,\bspace \Delta = \ffrac{\partial c}{\partial S_t}\\[2em]$$

- **Risk-Neutral Valuation**: first find the risk-neutral probabilities where the discounted stock price $\CB{\tilde S_t}_{t\geq0}$ is a martingale, by **Gisanov's Thoerem**.
    - $\given{\ffrac{\dd\QQ}{\dd\PP}}_{\FcF_t} = \exp\CB{-\theta W_t - \ffrac{1}{2}\theta^2 t}$, where $\theta = \ffrac{\mu - r}{\sigma}$
    - Under $\QQ$ measure, $\dd S_t = rS_t\dd t + \sigma S_t \dd \widetilde W_t$, $\dd \tilde S_t = \sigma \tilde S_t \dd \widetilde W_t$
    - then the risk-neutral valuation formula$\\[1em]$
$$\begin{align}
c\P{t,S_t} &= \ExpH_t^\QQ\SB{e^{-r\P{T-t}}c\P{T,S_T}}\\
&= \ExpH_t^\QQ\SB{e^{-r\P{T-t}} %\max
\P{S_T - K}^+}\\
&= \ExpH^\QQ\SB{e^{-r\P{T-t}} %\max
\P{S_T - K}^+ \mid S_t}
\end{align}$$
    - then the **Feynman-Kac theorem**, $\\[1em]$
$$\ffrac{\partial c}{\partial t} + \ffrac{1}{2} \sigma^2 S_t^2 \ffrac{\partial^2 c}{\partial S_t^2} + rS_t \ffrac{\partial c}{\partial S_t} - rc = 0$$

**BS equation**: 

- for European call option without dividend: $\ffrac{\partial c}{\partial t} + \ffrac{1}{2} \sigma^2 S_t^2 \ffrac{\partial^2 c}{\partial S_t^2} + rS_t \ffrac{\partial c}{\partial S_t} - rc = 0$

$$\begin{align}
c\P{t,S_t} &= S_t\Phi^{-1}\P{d_1} - Ke^{-r\P{T-t}} \Phi^{-1}\P{d_2}\equiv S_t\N{d_1} - Ke^{-r\P{T-t}} \N{d_2}\\
d_1 &= \ffrac{1}{\sigma\sqrt{T-t}} \P{\log\ffrac{S_t}{K} + \P{r\,\mathbf+\,\ffrac{1}{2}\sigma^2}\P{T-t}}\\
d_2 &= \ffrac{1}{\sigma\sqrt{T-t}} \P{\log\ffrac{S_t}{K} + \P{r\,\mathbf-\,\ffrac{1}{2}\sigma^2}\P{T-t}}= d_1 - \sigma \sqrt{T-t}
\end{align}$$

>We solve this using probabilistic method. Under $\QQ$,
>
>$$
c\P{t,S_t} = \ExpH^\QQ\SB{e^{-r\P{T-t}}} = e^{-r\P{T-t}}\P{\ExpH^\QQ_t\SB{S_T \cdot \mathbf{1}_{S_{\void_T}>K}}-K\cdot\ExpH^\QQ_t\SB{ \mathbf{1}_{S_{ \void_T} >K}}}
$$
>
>Note that $S_T = S_t \exp\CB{\P{r- \ffrac{1}{2}\sigma^2}\P{T-t} + \sigma\P{\widetilde W_T -\widetilde W_t}}$

>$$S_T\mid \FcF_t \sim S_t \exp\CB{\P{r- \ffrac{1}{2}\sigma^2}\P{T-t} + \sigma\sqrt{T-t}\cdot Z},\bspace Z\sim \N{0,1}$$
>
>$$\begin{align}

- **Currency Option**: Consider a currency option that give its owner the right to buy $1$ unit of a foreign currency for $K$ units of the domestic currency at maturity $T$. Thus the payoff would be, as evaluated in the domestic currency, is euqal to $\P{R_T - K}^+$, where $R_T$ is the exchange rate at time $T$, $i.e.$, the domestic value of one unit of foreign currency at time $T$. We now assume the **exchange rate process** $\CB{R_t}_{t\geq 0}$ satisfies the **Black-Scholes model**, $\dd R_t = R_t \SB{\mu_R \dd t + \sigma_R \dd W_t} = \mu_R R_t \dd t + \sigma_R R_t \dd W_t$. The wealth dynamic (in domestic currency) of a portfolio with $\pi_t$ in the foreign account and the rest in the domestic account, is$\\[0.8em]$
$$\dd V_t = \ffrac{\pi_t}{R_t^{\,f}} \dd R_t^{\,f} + \ffrac{V_t - \pi_t}{B_t} \dd B_t= \P{r_d V_t + \pi_t \P{\mu_R + r_f - r_d}}\dd t + \pi_t \sigma_R \dd W_t\\[1em]$$
where $r_d$ is the domestic risk-free rate. And this dynamics is the the same as the case of a dividend-paying with $q$ replaced by $r_f$. Thus we have the solution of the option price, by applying the risk-neutral method then,

$$\begin{align}
c\P{t,R_t} &= R_t \exp\CB{-r_f\P{T-t}}\N{d_1} - K\exp\CB{-r_d\P{T-t}}\N{d_2}\\
d_{1,2} &= \ffrac{1}{\sigma_R\sqrt{T-t}} \P{\log\P{\ffrac{R_t}{K}} + \P{r_d - r_f \pm \ffrac{\sigma^2}{2}}\P{T-t}}
\end{align}$$

## Hedging

- The hedger takes a *long position* in $\delta$ units of the futures $F_2\P{t,U}$. Here $\delta$ is called the **hedge ratio** and $F_2\P{t,U}$ is the **futures price** at time $t$ of the contract maturing at time $U$ on the asset $S^\P2$.
- **risk exposure** $H = S_t^\P1 - S_T^\P1 + \delta\P{F_2\P{T,U} - F_2\P{t,U}}$
    - $S_t^\P1 - S_T^\P1$: the change in the value of the asset to be delivered at time $T$
    - $\delta\P{F_2\P{T,U} - F_2\P{t,U}}$: the payoff of holding the *long position* in the futures contract between $t$ and $T$.
- The **basis risk**: its variance conditional at time $t$

$$\VarH_t\SB{H} = \VarH_t\SB{S_T^\P 1}+ \delta^2 \VarH_t \SB{F_2\P{T,U}}-2\delta \CovH_t\P{S_T^\P1,F_2\P{T,U}}$$

- **Optimal $\delta$**: taking the derivative with respect to $\delta$

$$\delta = \ffrac{\CovH_t\P{S_T^\P1,F_2\P{T,U}}}{\VarH_t \SB{F_2\P{T,U}}} = \rho\ffrac{\sigma_S}{\sigma_F}$$

- **Greeks neutral**: A position with a $\Delta$ ($\Gamma$) of zero is called **Delta** (**Gamma**) neutral. And pratically we first reach **Gamma Neutral** then the **Delta Neutral**.

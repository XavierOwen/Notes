## Question 1

Consider the claim which pays $S_{T_1}/S_{T_0}$ at time $T_1 > T_0 > 0$. What's the value of the claim at time $T_0$? What is the price of the claim at time zero?

>To calculate $\ExpH_{T_0}^\QQ\SB{e^{-r\P{T_1-T_0}} \ffrac{S_{T_1}}{S_{T_0}}}$, note that under $\QQ$, $e^{rt}S_t$ is a martingale. Thus
>
>$$\ExpH_{T_0}^\QQ\SB{e^{-rT_1}S_{T_1}} = \ExpH_{T_0}^\QQ\SB{e^{-rT_0}S_{T_0}} \Rightarrow \ExpH_{T_0}^\QQ\SB{e^{-r\P{T_1-T_0}} \ffrac{S_{T_1}}{S_{T_0}}} = 1$$
>
>Then at time $0$, $\ExpH_{0}^\QQ\SB{e^{-r\P{T_1-0}} \ffrac{S_{T_1}}{S_{T_0}}} =\ExpH_{0}^\QQ\SB{\ExpH_{T_0}^\QQ \SB{e^{-rT_1} \ffrac{S_{T_1}}{S_{T_0}}}} = e^{-rT_0}$
>***
>Or, if you cannot find that $e^{rt}S_t$ is a martingale, you can still solve is alternatively. Under $\QQ$, we have $\dd S_t = r S_t \dd t + \sigma S_t \dd W_t$ and thus we write
>
>$$S_{T_1} = S_{T_0} \exp\CB{\P{r-\ffrac{\sigma^2}{2}}\P{T_1-T_0} +\sigma \P{W_{T_1} - W_{T_0}}}$$
>
>Then, $\ExpH_{T_0}^\QQ\SB{e^{-r\P{T_1-T_0}} \ffrac{S_{T_1}}{S_{T_0}}} = \ExpH_{T_0}^\QQ\SB{\exp\CB{\P{-\ffrac{\sigma^2}{2}}\P{T_1-T_0} +\sigma \P{W_{T_1} - W_{T_0}}}}=1$

## Question 1.extended

Calculate $\ExpH_0^\QQ\SB{e^{-rT_1}\P{S_{T_1} - kS_{T_0}}^+}$, especially when $k = 1$, it becomes the forward start option's price.

>Actually, we are computing $\ExpH_0^\QQ\SB{e^{-rT}\P{S_{T} - kS_{T_1}}^+}$, and that's what a forward start option looks like. However here we keep the "wrong" notation.
>
>$$
\begin{align}
\ExpH_0^\QQ\SB{e^{-rT_1}\P{S_{T_1} - kS_{T_0}}^+} &= \ExpH_0^\QQ\SB{e^{-rT_0} \ExpH_{T_0}^\QQ\SB{e^{-r\P{T_1-T_0}}\P{S_{T_1} - kS_{T_0}}^+}}\\
&= \ExpH_0^\QQ\SB{e^{-rT_0} c\P{T_0,S_{T_0}, T_1,kS_{T_0}} }\\

## Question 2

Under the Black-Scholes model, calculate the value of an *Asian call option*, with strike price $K$, in which the *arithmetic average* of the stock price is calculated on the basis of just two sampling times, $0$ and $T$, where $T$ is the maturity time of the contract. Find an expression for the value of the corresponding contract when there are three sampling times, $0$, $T/2$ and $T$.

>$$\begin{align}
\ExpH_0^\QQ\SB{e^{-rT}\P{\ffrac{S_0+S_T}{2}-K}^+} &= \ffrac{1}{2}\ExpH_0^\QQ\SB{e^{-rT}\P{S_T - \P{2K-S_0}}^+}\\
&= \ffrac{1}{2}\P{S_0\N{d_1'} - \P{2K-S_0}e^{-rT}\N{d_2'}}\\
d_{1,2}' &= \ffrac{1}{\sigma\sqrt{T-t}}\P{\log\ffrac{S_t}{2K-S_0}+\P{r\pm\ffrac{\sigma^2}{2}}\P{T-t}}
\end{align}$$
>
>And the preceding formula requires that $2K-S_0>0$. If not, then $2K - S_0 \leq 0$ then
>
>$$\begin{align}
\ExpH_0^\QQ\SB{e^{-rT}\P{\ffrac{S_0+S_T}{2}-K}^+} &= \ffrac{1}{2}\ExpH_0^\QQ\SB{e^{-rT}\P{S_T - \P{2K-S_0}}}\\
&= \ffrac{1}{2}\ExpH_0^\QQ\SB{e^{-rT} S_T} - \ffrac{1}{2}e^{-rT}\P{2K-S_0},\bspace\text{martingale}\\
&= \ffrac{1}{2}S_0 - \ffrac{1}{2}e^{-rT}\P{2K-S_0}
\end{align}$$
>
>And when there are three sampling time, the corresponding Asian call option price is
>
>$$\ExpH_0^\QQ\SB{e^{-rT}\P{\ffrac{S_0+S_T+S_{T/2}}{3}-K}^+}$$

## Question 3

An *European call option* on a certain stock has a strike price of $\$30$, a time to maturity of one year, and an implied volatility of $30\%$. An *European put option* on the same stock has a strike price of $\$30$, a time to maturity of one year, and an implied volatility of $33\%$. What is the arbitrage opportunity open to a trader? Does the arbitrage work only when the *lognormal assumption* underlying Black-Scholes-Merton holds? Explain carefully the reasons for your answer.

>Put-call parity implies that European put and call options have the *same implied volatility*. 
>
>$$\left.\begin{align}
p_{\text{BS}} + S_0 e^{-qT} &= c_{\text{BS}} + Ke^{-rT}\\
p_{\text{imp}} + S_0 e^{-qT} &= c_{\text{imp}} + Ke^{-rT}\\
\end{align}\right\}\Rightarrow p_{\text{BS}} - p_{\text{imp}} = c_{\text{BS}}- c_{\text{imp}}$$
>
>Now the put has a higher implied volatility, meaning that its actual value is higher than the theoretical value, since ${\large\nu} = \ffrac{\partial c}{\sigma} = \ffrac{\partial p}{\sigma}>0$. Thus the artibtrage opportunity is to *buy the call*, *sell the put* and *short the stock*. This does not depend on the lognormal assumption underlying Black-Scholes-Merton. Put-call parity is true for any set of assumptions.

## Question 4

Consider a *European call* and a *European put* with the *same* strike price and time to maturity. Show that they change in value by the same amount when the volatility increases from a level, $\sigma_1$, to a new level, $\sigma_2$ within a short period of time.

>Define $c_1$ and $p_1$ as the values of the call and the put when the volatility is $\sigma_1$. Define $c_2$ and $p_2$ as the values of the call and the put when the volatility is $\sigma_2$. From put-call parity, we have:
>
>$$\left.\begin{align}
p_1 + S_0 e^{-qT} &= c_1 + Ke^{-rT}\\
p_2 + S_0 e^{-qT} &= c_2 + Ke^{-rT}
\end{align}\right\}\Rightarrow p_1 - p_2 = c_1 - c_2$$

## Question 5

Suppose that the stock price today is $S_t = 10$, the interest rate is $r = 0\%$, and the time to maturity is $0.5$ years. Consider an option whose Black-Scholes price is given by the function

$$c\P{t,S_t} = S_t^2 e^{2\P{T-t}}$$

where time is in annual terms. If your portfolio is long $10$ of these options and long or short $N$ shares of the stock, what should the value of $N$ be so that the portfolio is delta-neutral?

>The $\Delta$ of the portfolio is
>
>$$\ffrac{\partial \P{10c\P{t,S_t} - NS_t}}{\partial S_t} = 20 S_t e^{2\P{T-t}} - N = 200e-N$$
>
>Thus Delta-Neutral requires to short $200e$ shares of stock.

## Question 7

Let $c\P{t, S_t }$ and $p\P{t, S_t }$ denote the values of a European call and put option with the same exercise price $K$ and expiry time $T$. Show that $c\P{t, S_t }-p\P{t, S_t }$ also satisfies the Black-Scholes equation with the final data $c\P{T, S_T }-p\P{T, S_T } = S_T-K$. Deduce that $S_t-Ke^{-r\P{T-t}}$ is also a solution of the Black-Scholes equation. Interpret these results financially.

>Notice that for call, $\ffrac{\partial c}{\partial t}+rS_t \ffrac{\partial c}{\partial S_t}+\ffrac{1}{2}\sigma^2S_t^2\ffrac{\partial^2 c}{\partial S_t^2}-rc = 0$ subject to $c\P{T,S_T} = \P{S_T-K}^+$ and for put, $\ffrac{\partial p}{\partial t}+rS_t \ffrac{\partial p}{\partial S_t}+\ffrac{1}{2}\sigma^2S_t^2\ffrac{\partial^2 p}{\partial S_t^2}-rp = 0$ subject to $p\P{T,S_T} = \P{K-S_T}^+$. Hence,
>
>$$\ffrac{\partial \P{c-p}}{\partial t}+rS_t \ffrac{\partial \P{c-p}}{\partial S_t}+\ffrac{1}{2}\sigma^2S_t^2\ffrac{\partial^2 \P{c-p}}{\partial S_t^2}-r\P{c-p} = 0$$
>
>subject to $c\P{T,S_T} - p\P{T,S_T} = S_T-K$. And that's the equation, however we observe that
>
>$$\ffrac{\partial \P{S_t-Ke^{-r\P{T-t}}}}{\partial t}+rS_t \ffrac{\partial \P{S_t-Ke^{-r\P{T-t}}}}{\partial S_t}+\ffrac{1}{2}\sigma^2S_t^2\ffrac{\partial^2 \P{S_t-Ke^{-r\P{T-t}}}}{\partial S_t^2}-r\P{S_t-Ke^{-r\P{T-t}}} = 0$$
>
>And when $t=T$, $S_T-Ke^{-r\P{T-T}} = S_T-K$. Thus, $c\P{t, S_t }-p\P{t, S_t } = S_t-Ke^{-r\P{T-t}}$. Financially, this means to hold a European call is equivalent to holding a European put and a forward.

## Question 8

Consider the exchange option which entities the holder the right but not the obligation to exchange risky asset $S^\P1$ for another risky asset $S^\P2$ at maturity $T$. Let the price dynamics of $S^\P1$ and $S^\P2$ under the risk neutral measure be governed by, for $i=1,2$,

$$\ffrac{\partial S_t^\P{ i}}{S_t^\P{i}} = \P{r-q_i}\dd t + \sigma_i \dd W_t^\P i$$

The correlation of the two brownian motion has instantaneous correlation $\rho$, such that $\Exp{W_t^\P1 W_t^\P2} = \rho t$. Let $V\P{t,S_t^\P1,S_t^\P2}$ denote the price function of the exchange option, whose terminal payoff is $V\P{T,S_T^\P1,S_T^\P2} = \P{S_T^\P2 - S_T^\P1}^+$. Solve it.

> First consider the wealth process. Write the total value of holding $1$ share of stock as $G_t^\P i\equiv S_t + \d{\int_0^t q_iS_u\;\dd u}$, then
>
>$$\dd V_t = \ffrac{V_t - \P{\pi_t^\P 1 + \pi_t^\P2}}{B_t} \dd B_t + \ffrac{\pi_t^\P1}{G_t^\P1}\dd G_t^\P1 + \ffrac{\pi_t^\P2}{G_t^\P2}\dd G_t^\P2\\
\dd V_t = \P{rV_t + \pi_t^\P1\P{\mu_1 \color{brown}{+ q_1} - r} + \pi_t^\P2\P{\mu_2 \color{brown}{+ q_2} - r}}\dd t + \pi_t^\P1\sigma_1 \dd W_t^\P1 + \pi_t^\P2\sigma_2 \dd W_t^\P2$$
>
>Using **two-dimensional Itô's rule** yields
>
>$$\begin{align}\dd V &= \ffrac{\partial V}{\partial t}\dd t + \ffrac{\partial V}{\partial S_t^\P 1} \dd S_t^\P 1 + \ffrac{\partial V}{\partial S_t^\P 2} \dd S_t^\P 2 + \ffrac{1}{2}\ffrac{\partial^2 V}{\partial S_t^\P1\partial S_t^\P1} \P{\dd S_t^\P 1}^2 + \ffrac{1}{2}\ffrac{\partial^2 V}{\partial S_t^\P2\partial S_t^\P2} \P{\dd S_t^\P 2}^2\\
&\bspace+ \ffrac{1}{2}\ffrac{\partial^2 V}{\partial S_t^\P1\partial S_t^\P2}\dd S_t^\P 1\dd S_t^\P 2\\
&= \ffrac{\partial V}{\partial t}\dd t + \ffrac{\partial V}{\partial S_t^\P 1} S_t^\P 1 \SB{\mu_1 \dd t + \sigma_1 \dd W_t^\P 1} + \ffrac{\partial V}{\partial S_t^\P 2} S_t^\P 2 \SB{\mu_2 \dd t + \sigma_2 \dd W_t^\P 2}\\
&\bspace + \ffrac{1}{2}\ffrac{\partial^2 V}{\partial S_t^\P1\partial S_t^\P1} \P{S_t^\P 1}^2 \SB{\mu_1 \dd t + \sigma_1 \dd W_t^\P 1}^2 + \ffrac{1}{2}\ffrac{\partial^2 V}{\partial S_t^\P2\partial S_t^\P2} \P{S_t^\P 2}^2 \SB{\mu_2 \dd t + \sigma_2 \dd W_t^\P 2}^2\\
&\bspace + \ffrac{1}{2}\ffrac{\partial^2 V}{\partial S_t^\P1\partial S_t^\P2}S_t^\P 1 \SB{\mu_1 \dd t + \sigma_1 \dd W_t^\P 1}S_t^\P 2 \SB{\mu_2 \dd t + \sigma_2 \dd W_t^\P 2}
\end{align}$$
>
>Since $\dd t\dd t = 0$, $\dd W_t^\P 1 \dd W_t^\P 1 = \dd t$, $\dd W_t^\P 1 \dd t = 0$, and $\dd W_t^\P 1 \dd W_t^\P 2 = \rho \dd t$, the preceding it equal to

>$$\begin{align}\dd V &= \Big[ \ffrac{\partial V}{\partial t} + \mu_1 S_t^\P 1 \ffrac{\partial V}{\partial S_t^\P 1} + \mu_2 S_t^\P 2 \ffrac{\partial V}{\partial S_t^\P 2} + \ffrac{1}{2}\P{\sigma_1S_t^\P1}^2 \ffrac{\partial^2 V}{\partial S_t^\P1\partial S_t^\P1} + \ffrac{1}{2}\P{\sigma_2S_t^\P2}^2 \ffrac{\partial^2 V}{\partial S_t^\P2\partial S_t^\P2}\\
&\bspace+ \rho \sigma_1 \sigma_2 S_t^\P1 S_t^\P2 \ffrac{\partial^2 V}{\partial S_t^\P1\partial S_t^\P2}\Big]\dd t + \sigma_1 S_t^\P1 \ffrac{\partial V}{\partial S_t^\P1} \dd W_t^\P 1 + \sigma_2 S_t^\P2 \ffrac{\partial V}{\partial S_t^\P2} \dd W_t^\P 2
\end{align}$$
>
>Thus to replicate, $V\P{t,S_t^\P1,S_t^\P2} = V_t$. And similar to the one-asset case, comparing the $\dd t$ term and the $\dd W_t^\P 1$, $\dd W_t^\P 2$ terms, we have the ***Black-Scholes equation for two assets***:
>
>$$\begin{align}
\ffrac{\partial V}{\partial t} + \ffrac{1}{2}\P{\sigma_1S_t^\P1}^2 \ffrac{\partial^2 V}{\partial S_t^\P1\partial S_t^\P1} + \ffrac{1}{2}\P{\sigma_2S_t^\P2}^2 \ffrac{\partial^2 V}{\partial S_t^\P2\partial S_t^\P2} + \rho \sigma_1 \sigma_2 S_t^\P1 S_t^\P2 \ffrac{\partial^2 V}{\partial S_t^\P1\partial S_t^\P2}&\\
+ \P{r\color{brown}{- q_1}}S_t^\P1 \ffrac{\partial V}{\partial S_t^\P1} + \P{r\color{brown}{- q_2}}S_t^\P2 \ffrac{\partial V}{\partial S_t^\P2} - rV &= 0
\end{align}$$
>
>with the **replicating portfolio** $\ffrac{\pi_t^\P1}{S_t^\P 1} = \ffrac{\partial V}{\partial S_t^\P1}$ and $\ffrac{\pi_t^\P2}{S_t^\P 2} = \ffrac{\partial V}{\partial S_t^\P2}$ and **boundary condition**: $V\P{T,S_T^\P 1,S_T^\P2} = \P{S_T^\P 2 - S_T^\P1}^+ = S_T^\P1\P{\ffrac{S_T^\P 2}{S_T^\P1} - 1}^+$.
>
>Then our conjecture is $V\P{t,S_t^\P1,S_t^\P2} = S_t^\P1 h\P{t,Z_t}$, where $Z_t = \ffrac{S_t^\P2}{S_t^\P1}$. We now derive the governing equation for $h$:
>
>$$\begin{align}
0=\;&S_t^\P 1 \ffrac{\partial h}{\partial t} + \ffrac{1}{2} \P{\sigma_1 S_t^\P1}^2 \color{brown}{S_t^\P1 \ffrac{\partial^2 h}{\partial Z_t^2}\P{\ffrac{\partial Z_t}{\partial S_t^\P1}}^2} + \ffrac{1}{2} \P{\sigma_2 S_t^\P2}^2 S_t^\P1\ffrac{\partial^2 h}{\partial Z_t^2}\P{\ffrac{\partial Z_t}{\partial S_t^\P2}}^2\\
&\bspace+\rho \sigma_1 \sigma_2 S_t^\P1 S_t^\P2 \color{brown}{S_t^\P1\ffrac{\partial^2 h}{\partial Z_t^2}\ffrac{\partial Z_t^2}{\partial S_t^\P1\partial S_t^\P2}}\\
&\bspace+ \P{r\color{brown}{- q_1}}S_t^\P1 \color{brown}{\P{S_t^\P1\ffrac{\partial h}{\partial Z_t}\ffrac{\partial Z_t}{\partial S_t^\P1} + h\P{t,Z_t}}} + \P{r\color{brown}{- q_2}}S_t^\P2 S_t^\P1\ffrac{\partial h}{\partial Z_t}\ffrac{\partial Z_t}{\partial S_t^\P2} - rS_t^\P1 h\P{t,Z_t}\\
\Rightarrow 0=\;& h_t + \ffrac{1}{2} \P{\sigma_1^2 + \sigma_2^2 - 2 \rho\sigma_1\sigma_2}Z_t^2\ffrac{\partial^2 h}{\partial Z_t^2} \color{brown}{+\P{q_1-q_2}Z_t\ffrac{\partial h}{\partial Z_t} - q_1 h\P{t,Z_t}}
\end{align}$$
>
>And indeed, we can use **Feynman–Kac Formula**, with its **boundary condition**: $h\P{T,Z_T} = \P{Z_T-1}^+$ (with "strike price" $K=1$) and $r = q_1$, $\mu=\P{q_1-q_2}\color{brown}{Z_t}$, and $\sigma = \sqrt{\sigma_1^2 + \sigma_2^2-2\rho\sigma_1\sigma_2} \color{brown}{\cdot Z_t}$. So the preceding *PDE and its boundary condition* is equivalent to the following *SDE* (stochastic differential equation): $\dd Z_t = \P{q_1-q_2}Z_t \dd t + \sqrt{\sigma_1^2 + \sigma_2^2-2\rho\sigma_1\sigma_2} \cdot Z_t \dd \hat{W}_t$. Or in stardard **Black-Schole equation** form, $\mu=q_1-q_2$ and $\sigma = \sqrt{\sigma_1^2 + \sigma_2^2-2\rho\sigma_1\sigma_2}$:
>
>$$\ffrac{\dd Z_t}{Z_t} = \P{q_1-q_2} \dd t + \sqrt{\sigma_1^2 + \sigma_2^2-2\rho\sigma_1\sigma_2} \dd \hat{W}_t$$
>
>And similar to regular **BS formula with dividend**, we have:
>
>$$\begin{align}h\P{t,Z_t} &= Z_t \color{brown}{e^{-q_{\void_2}\P{T-t}}}\N{\hat d_1}-1\cdot\color{brown}{e^{-q_{\void_1}\P{T-t}}}\N{\hat d_2}\\
V\P{t,S_t^\P1,S_t^\P2} &= S_t^\P2 \color{brown}{e^{-q_{\void_2}\P{T-t}}}\N{\hat d_1} - S_t^\P1 \color{brown}{e^{-q_{\void_1}\P{T-t}}}\N{\hat d_2}\\
&= e^{-r\P{T-t}}\P{S_t^\P2 e^{\color{brown}{\P{r-q_{\void_2}}}\P{T-t}}\N{\hat d_1} - S_t^\P1 e^{\color{brown}{\P{r-q_{\void_1}}}\P{T-t}}\N{\hat d_2}}\\
\hat d_{1,2} &= \ffrac{1}{\sigma\sqrt{T-t}} \P{\log\P{\ffrac{S_t^{\P2}}{S_t^{\P1}}}+ \P{\color{brown}{q_1 - q_2} \pm \ffrac{\sigma^2}{2}}\P{T-t}}\\
&= \ffrac{1}{\sigma\sqrt{T-t}} \P{\log\P{\ffrac{S_t^{\P2}}{S_t^{\P1}}}+ \P{\color{brown}{\P{r-q_2} - \P{r-q_1}} \pm \ffrac{\sigma^2}{2}}\P{T-t}}
\end{align}$$

## Question 9

&= \ffrac{1}{\sigma\sqrt{T-t}} \P{\log\P{\ffrac{S_t^{\P1}}{S_t^{\P2}}}+ \P{\color{brown}{\P{r-q_1} - \P{r-q_2}} \pm \ffrac{\sigma^2}{2}}\P{T-t}}
\end{align}$$
>
>Now $S_t^\P1 = K$ and the corresponding $q_1 = r$, $\sigma_1 = 0$ then the preceding is updated as
>
>$$\begin{align}
p\P{t,S_t^\P1,S_t^\P2} &= e^{-r\P{T-t}}\P{K\N{\hat d_1} - S_t^\P2 e^{\color{brown}{\P{r-q_{\void_2}}}\P{T-t}}\N{\hat d_2}}\\
\hat d_{1,2} &= \ffrac{1}{\sigma_2\sqrt{T-t}} \P{\log\P{\ffrac{K}{S_t^{\P2}}}+ \P{\color{brown}{ - \P{r-q_2}} \pm \ffrac{\sigma_2^2}{2}}\P{T-t}}
\end{align}$$

$\P{\text b}$ Show how we can derive the European call price formula from the European put price formula using the put-call symmetry relation.

>The **put-call symmetry relation** with **dividend** is
>
>$$c + Ke^{-rT} = p + S_0e^{-qT}$$
>
>Thus here
>
>$$\begin{align}
c\P{t,S_t^\P1,S_t^\P2} &= e^{-r\P{T-t}}\P{K\N{\hat d_1} - S_t^\P2 e^{\color{brown}{\P{r-q_{\void_2}}}\P{T-t}}\N{\hat d_2}} + S_t^\P2 e^{-q_{\void_2}\P{T-t}} - Ke^{-r\P{T-t}}\\

\hat d_{1,2} &= \ffrac{1}{\sigma_2\sqrt{T-t}} \P{\log\P{\ffrac{K}{S_t^{\P2}}}+ \P{\color{brown}{ - \P{r-q_2}} \pm \ffrac{\sigma_2^2}{2}}\P{T-t}}
\end{align}$$

>Or directly regard the call as using $K$ in exchange of $S_t^\P2$

c\P{t,S_t^\P1,S_t^\P2} &=  e^{-r\P{T-t}}\P{S_t^\P2 e^{\color{brown}{\P{r-q_{\void_2}}}\P{T-t}}\N{\hat d_1} - K \N{\hat d_2}}\\

## Question 10

Consider a *future* on an underlying asset which pays $N$ discrete dividends between $t$ and $T$ and let $D_i$ denote the amount of the $i$th dividend paid on the ex-dividend date $t_i$, $i = 1, 2 \cdots, N$. Show that the futures price is given by 

$$F\P{t,S_t} = S_te^{r\P{T-t}} - \sum_{i=1}^N D_i e^{r\P{T-t_{\void_i}}}$$

where $S_t$ is the current asset price and $r$ is the riskless interest rate.

>This special future is not applicable to the put call parity. So here we use risk-neutral pricing method. Consider at time $t$, we sell $S_t$ units of bonds, then buy an asset, and short this forward. At time $T$, our payoff is
>
>$$K+\sum_{i=1}^N D_i e^{r\P{T-t_{\void_i}}} - S_te^{r\P{T-t}} = 0 \Rightarrow F\P{t,S_t} = K = S_te^{r\P{T-t}} - \sum_{i=1}^N D_i e^{r\P{T-t_{\void_i}}}$$

$Remark$

>The forward price is not its value. Its value is
>
>$$f\P{t,S_t} = e^{-r\P{T-t}}\ExpH^\QQ_t\SB{S_T-K} = S_t - Ke^{-r\P{T-t}} = \sum_{i=1}^N D_i e^{r\P{t-t_{\void_i}}}$$

## Q1
Prove that 
$$\|A\|_{\infty} = \max_{1 \leq i \leq m} \| \mathbf{b}_i\|_1,$$
for any real $m \times n$ matrix $A$, where $\mathbf{b}_i$ , for $i = 1, 2, \dots, m$, are the row vectors of $A$.

$Answer$

We first prove that $\|A\|_{\infty} \leq \max\limits_{1 \leq i \leq m} \| \mathbf{b}_i\|_1$, and we use $\mathbf{a}^r_i$ to express the row vectors of matrix $A$, instead of $\mathbf{b}_i$:
$$\begin{align*}
\|A\|_{\infty} &= \max_{\mathbf{x} \neq \mathbf{0}} \frac{\|A\mathbf{x}\|_{\infty}} {\|\mathbf{x}\|_{\infty}} \\
&= \max_{\mathbf{x} \neq \mathbf{0}} \frac{\|\begin{bmatrix}
\mathbf{a}_1^r \mathbf{x}, &  
\mathbf{a}_2^r \mathbf{x}, & 
\dots, & 
\mathbf{a}_m^r \mathbf{x}
\end{bmatrix}^{\mathrm{T}}\|_{\infty}} {\|\mathbf{x}\|_{\infty}}\\
&= \max_{\mathbf{x} \neq \mathbf{0}} \frac{\|\begin{bmatrix}
|\mathbf{a}_1^r \mathbf{x}|, &  
|\mathbf{a}_2^r \mathbf{x}|, & 
\dots, & 
|\mathbf{a}_m^r \mathbf{x}|
\end{bmatrix}^{\mathrm{T}}\|_{\infty}} {\|\mathbf{x}\|_{\infty}}\\
&\leq \max_{\mathbf{x} \neq \mathbf{0}} \frac{\|\begin{bmatrix}
\left( \sum_{j=1}^{n}|a|_{1j} \right) \|\mathbf{x}\|_{\infty}, &  
\left( \sum_{j=1}^{n}|a|_{2j} \right) \|\mathbf{x}\|_{\infty}, & 
\dots, & 
\left( \sum_{j=1}^{n}|a|_{mj} \right) \|\mathbf{x}\|_{\infty}
\end{bmatrix}^{\mathrm{T}}\|_{\infty}}{\|\mathbf{x}\|_{\infty}}\\
&= \left\| \begin{bmatrix}
\sum_{j=1}^{n}|a_{1j}|, &  
\sum_{j=1}^{n}|a_{2j}|, & 
\dots, & 
\sum_{j=1}^{n}|a_{mj}|
\end{bmatrix}^{\mathrm{T}} \right\|_{\infty} \\
&= \left\| \begin{bmatrix}
\| \mathbf{a}^r_1\|_1, &  
\| \mathbf{a}^r_2\|_1, & 
\dots, & 
\| \mathbf{a}^r_m\|_1
\end{bmatrix}^{\mathrm{T}} \right\|_{\infty} \\
&= \max_{1 \leq i \leq m} \| \mathbf{a}^r_i\|_1
\end{align*}$$
And now we head to the opposite direction of the inequation, that is $\|A\|_{\infty} \geq \max\limits_{1 \leq i \leq m} \| \mathbf{a}^r_i\|_1$:  
We first construct a special column vector $\mathbf{x}^{*}_i$ for each row vector $\mathbf{a}^r_i$. That is we let the $j\mathrm{-th}$ component of vector $\mathbf{x}^{*}_i$ is $\pm 1$, denoted as $x^{*}_{ij}$, and the symbol is determind by the $j\mathrm{-th}$ component of vector $\mathbf{a}^r_i$, $i.e.$, $a_{ij}$. If $a_{ij} \geq 0$, then $x^{*}_{ij}$ is $+1$, otherwise is $-1$. And obviously we always have $\|\mathbf{x}^{*}_i\|_{\infty}= 1$. Now we can start to prove. For all $i = 1,2,\dots,m$, we have
$$\begin{align*}
\|A\|_{\infty} &= \max_{\mathbf{x} \neq \mathbf{0}} \frac{\|A\mathbf{x}\|_{\infty}} {\|\mathbf{x}\|_{\infty}} \\
& \geq \frac{\|A \mathbf{x}^{*}_i\|_{\infty}} {\|\mathbf{x}^{*}_i\|_{\infty}} \\
&= \left\|\begin{bmatrix}
a_{11}x^{*}_{i1}+a_{12}x^{*}_{i2}+\cdots+a_{1n}x^{*}_{in} \\ 
a_{21}x^{*}_{i1}+a_{22}x^{*}_{i2}+\cdots+a_{2n}x^{*}_{in} \\ 
\cdots\\ 
|a_{i1}|+|a_{i2}|+\cdots+|a_{in}| \\
\cdots \\
a_{m1}x^{*}_{i1}+a_{m2}x^{*}_{i2}+\cdots+a_{mn}x^{*}_{in}
\end{bmatrix}\right\|_{\infty}
\end{align*}$$
$i.e.,$  
$$
\|A\|_{\infty}
\geq 
\left\|\begin{bmatrix}
\sum_{j=1}^n a_{1j}x^{*}_{1j} \\ 
\sum_{j=1}^n a_{2j}x^{*}_{1j} \\ 
\cdots \\
\sum_{j=1}^n a_{mj}x^{*}_{1j}
\end{bmatrix}\right\|_{\infty},
\|A\|_{\infty}
\geq 
\left\|\begin{bmatrix}
\sum_{j=1}^n a_{1j}x^{*}_{2j} \\ 
\sum_{j=1}^n a_{2j}x^{*}_{2j} \\ 
\cdots \\
\sum_{j=1}^n a_{mj}x^{*}_{2j}
\end{bmatrix}\right\|_{\infty},
\cdots,
\|A\|_{\infty}
\geq 
\left\|\begin{bmatrix}
\sum_{j=1}^n a_{1j}x^{*}_{mj} \\ 
\sum_{j=1}^n a_{2j}x^{*}_{mj} \\ 
\cdots \\
\sum_{j=1}^n a_{mj}x^{*}_{mj}
\end{bmatrix}\right\|_{\infty}
$$

Now we rearrange each row of these $m$ numbers of matrixes, and we get:
$$
\|A\|_{\infty}
\geq 
\left\|\begin{bmatrix}
\sum_{j=1}^n a_{1j}x^{*}_{1j} \\ 
\sum_{j=1}^n a_{1j}x^{*}_{2j} \\ 
\cdots \\
\sum_{j=1}^n a_{1j}x^{*}_{mj} 
\end{bmatrix}\right\|_{\infty},
\|A\|_{\infty}
\geq 
\left\|\begin{bmatrix}
\sum_{j=1}^n a_{2j}x^{*}_{1j} \\ 
\sum_{j=1}^n a_{2j}x^{*}_{2j} \\ 
\cdots \\
\sum_{j=1}^n a_{2j}x^{*}_{mj} 
\end{bmatrix}\right\|_{\infty},
\cdots,
\|A\|_{\infty}
\geq 
\left\|\begin{bmatrix}
\sum_{j=1}^n a_{mj}x^{*}_{1j} \\ 
\sum_{j=1}^n a_{mj}x^{*}_{2j} \\ 
\cdots \\
\sum_{j=1}^n a_{mj}x^{*}_{mj} 
\end{bmatrix}\right\|_{\infty}
$$

From the property of $\mathbf{x}^{*}_i$, that for $m \neq i$, $\sum_{j=1}^n a_{ij}x^{*}_{ij} = \sum_{j=1}^n |a_{ij}| \geq \sum_{j=1}^n a_{ij}x^{*}_{mj}$.
We now get that for all $i = 1,2,\dots,m$, we have
$$\|A\|_{\infty} \geq \sum_{j=1}^n |a_{ij}| = \| \mathbf{a}^r_i\|_1$$
$i.e.,$  
$$\|A\|_{\infty} \geq \max\limits_{1 \leq i \leq m} \| \mathbf{a}^r_i\|_1$$
Now two direction is proved truth so we can conclude that $\|A\|_{\infty} = \max\limits_{1 \leq i \leq m} \| \mathbf{a}^r_i\|_1$.
$\square$

## Q2
Let $Q$ be an $m \times m$ orthogonal matrix.

$(\textrm{a})$ Prove that for any given vectors $\mathbf{u}$ and $\mathbf{v}$ in $\mathbb{R}^m$, the angle between $\mathbf{u}$ and $\mathbf{v}$ is the same as the angle between $Q\mathbf{u}$ and $Q\mathbf{v}$.

$Answer$  
Let $\theta$ be the angle between $\mathbf{u}$ and $\mathbf{v}$, and $\theta '$ be the angle between $Q\mathbf{u}$ and $Q\mathbf{v}$.  
Since
$$\left( Q\mathbf{u},Q\mathbf{v} \right) = \left( \mathbf{u},\mathbf{v} \right)$$
then $\|Q\mathbf{u}\| = \left( Q\mathbf{u},Q\mathbf{u} \right) = \left( \mathbf{u},\mathbf{u} \right) = \|\mathbf{u}\|$, so that we have
$$\begin{align*}
\cos \theta &= \frac{\left( \mathbf{u}, \mathbf{v} \right)} {\left\| \mathbf{u} \right\| \left\| \mathbf{v} \right\|} \\
&= \frac{\left( Q\mathbf{u}, Q\mathbf{v} \right)} {\left\| Q\mathbf{u} \right\| \left\| Q\mathbf{v} \right\|} \\
&= \cos \theta'
\end{align*}$$
And since $0^{\circ} \leq \theta, \theta' \leq 180^{\circ}$, so we finally get $\theta = \theta'$.

$(\textrm{b})$ Prove that $\|Q\|_2 = 1$.

$Answer$  
So we already know that orthogonal matrices preserves the length of a vector from $(a)$, then

$$
\begin{align*}
\|Q\|_2 &= \max_{\|\mathbf{x}\|_2 = 1}\|Q\mathbf{x}\|_2 \\
&= \max_{\|\mathbf{x}\|_2 = 1}\|\mathbf{x}\|_2 \\
&= 1
\end{align*}
$$

$(\textrm{c})$ Give an example $2 \times 2$ orthogonal matrix $Q$, such that $\|Q\|_1 \neq 1$ and $\|Q\|_{\infty} \neq 1$.

$Answer$  
$$Q = \begin{bmatrix}
\frac{\sqrt{2}}{2} & \frac{\sqrt{2}}{2}\\ 
\frac{\sqrt{2}}{2} & -\frac{\sqrt{2}}{2}
\end{bmatrix}$$
Here $\|Q\|_1 = \sqrt{2}$ and $\|Q\|_{\infty} = \sqrt{2}$.

$(\textrm{d})$ Prove that $\|QA\|_2 = \|A\|_2$ , for any $m \times m$ matrix $A$.

$Answer$  
$$
\begin{align*}
\|QA\|_2 &= \max_{\|\mathbf{x}\|_2 = 1} \|QA\mathbf{x}\|_2 \\ 
&= \max_{\|\mathbf{x}\|_2 = 1} \left( QA\mathbf{x},QA\mathbf{x} \right) \\
&= \max_{\|\mathbf{x}\|_2 = 1} \mathbf{x}^{\mathrm{T}}A^{T}Q^{T}QA\mathbf{x} \\
&= \max_{\|\mathbf{x}\|_2 = 1} \mathbf{x}^{\mathrm{T}}A^{T}A\mathbf{x} \\
&= \max_{\|\mathbf{x}\|_2 = 1} \|A\mathbf{x}\|_2 = \|A\|_2
\end{align*}$$

## Q3
Given
$$\mathbf{u} = \left[
\begin{array}{r}
\sqrt{6} \\
\sqrt{3} \\
\sqrt{2} \\
\end{array} \right],Q = \left[
\begin{array}{rrr}
\sqrt{6}/6 & \sqrt{3}/3 & \sqrt{2}/2 \\
\sqrt{6}/3 & -\sqrt{3}/3 & 0 \\
\sqrt{6}/6 & \sqrt{3}/3 & -\sqrt{2}/2
\end{array}
\right]$$

$(\textrm{a})$ Compute $\|\mathbf{u}\|_1,\|\mathbf{u}\|_2,\|\mathbf{u}\|_{\infty}$.

$Answer$  
$\|\mathbf{u}\|_1 = \sqrt{6} + \sqrt{3} + \sqrt{2}$

$\|\mathbf{u}\|_2 = \sqrt{6+3+2} = \sqrt{11}$

$\|\mathbf{u}\|_{\infty} = \sqrt{6}$

$(\textrm{b})$ Show that $Q$ is an orthogonal matrix.

$Answer$  
$\begin{align*}
QQ^{\mathrm{T}} &=
\begin{bmatrix}
\sqrt{6}/6 & \sqrt{3}/3 & \sqrt{2}/2 \\
\sqrt{6}/3 & -\sqrt{3}/3 & 0 \\
\sqrt{6}/6 & \sqrt{3}/3 & -\sqrt{2}/2
\end{bmatrix}
\cdot
\begin{bmatrix}
\sqrt{6}/6 & \sqrt{6}/3 & \sqrt{6}/6 \\
\sqrt{3}/3 & -\sqrt{3}/3 & \sqrt{3}/3 \\
\sqrt{2}/2 & 0 & -\sqrt{2}/2
\end{bmatrix}\\
&= 
\begin{bmatrix}
1/6+1/3+1/2 & 1/3-1/3+0 & 1/6+1/3-1/2\\ 
1/3-1/3+0 & 2/3+1/3+0 & 1/3-1/3+0\\ 
1/6+1/3-1/2 & 1/3-1/3+0 & 1/6+1/3+1/2
\end{bmatrix}\\
&= I_3
\end{align*}$  
Thus, matrix $Q$ is orthogonal.

$(\textrm{c})$ Compute $\|Q\mathbf{u}\|_1,\|Q\mathbf{u}\|_2,\|Q\mathbf{u}\|_{\infty}$. Are they the same as $\|\mathbf{u}\|_1,\|\mathbf{u}\|_2,\|\mathbf{u}\|_{\infty}$, respectively? Which are and which are not?

$Answer$  
First, we calculate $Q\mathbf{u}$:
$$Q\mathbf{u} = \begin{bmatrix}
\sqrt{6}/6 & \sqrt{3}/3 & \sqrt{2}/2 \\
\sqrt{6}/3 & -\sqrt{3}/3 & 0 \\
\sqrt{6}/6 & \sqrt{3}/3 & -\sqrt{2}/2
\end{bmatrix}
\cdot
\begin{bmatrix}
\sqrt{6} \\
\sqrt{3} \\
\sqrt{2} \\
\end{bmatrix} = 
\begin{bmatrix}
3\\ 
1\\ 
1
\end{bmatrix}$$
Then,

$\|Q\mathbf{u}\|_1 = 3+1+1 = 5$

$\|Q\mathbf{u}\|_2 = \sqrt{9+1+1} = \sqrt{11}$

$\|Q\mathbf{u}\|_{\infty} = 3$

$i.e.,$  
Only $\|Q\mathbf{u}\|_2 = \|\mathbf{u}\|_2$.

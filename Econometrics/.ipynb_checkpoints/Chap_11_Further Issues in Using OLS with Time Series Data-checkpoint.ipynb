## Stationary and Nonstationary Time Series

A ***stationary time series process*** is one whose probability distributions are stable over time in the following sense: *If we take any collection of $r.v.$ in the sequence and then shift that sequence ahead $h$ time periods, the joint probability distribution must remain unchanged.*

$Def$ Stationary Stochastic process

>The stochastic process $\CB{x_t:t=1,2,\dots}$ is ***stationary*** if, for every collection of time indices $1\leq t_1 \leq t_2 \leq \cdots \leq t_m$, the joint distribution of $\P{x_{t_1},x_{t_2},\dots, x_{t_m}}$ is the same as the joint distribution of $\P{x_{t_1 + h},x_{t_2 + h},\dots, x_{t_m + h}}$ for all integers $h\geq 1$.

```python
import numpy as np
import time
import matplotlib
from matplotlib import colors
import matplotlib.pyplot as plt
```

```python
xmin, xmax, xn = -2.25, +0.75, 3000/2
ymin, ymax, yn = -1.25, +1.25, 2500/2
maxiter = 200
horizon = 2.0 ** 40
log_horizon = np.log(np.log(horizon))/np.log(2)
```

```python
X = np.linspace(xmin, xmax, xn)
Y = np.linspace(ymin, ymax, yn)
```

```python
C = X + Y[:, None]*1j
N = np.zeros(C.shape, dtype=int)
Z = np.zeros(C.shape, np.complex64)
```

```python
for n in range(maxiter):
    I = np.less(abs(Z), horizon)
    N[I] = n
    Z[I] = Z[I]**2 + C[I]
N[N == maxiter-1] = 0
```

```python
with np.errstate(invalid='ignore'):
    M = np.nan_to_num(N + 1 -np.log(np.log(abs(Z)))/np.log(2) +log_horizon)
```

```python
dpi = 128
width = 20
height = 20*yn/xn
fig = plt.figure(figsize=(width, height), dpi=dpi)
ax = fig.add_axes([0.0, 0.0, 1.0, 1.0], frameon=False, aspect=1)

light = colors.LightSource(azdeg=315, altdeg=10)
M = light.shade(M, cmap=plt.cm.hot, vert_exag=1.5, norm=colors.PowerNorm(0.3), blend_mode='hsv')
plt.imshow(M, extent=[xmin, xmax, ymin, ymax], interpolation="bicubic")
ax.set_xticks([])
ax.set_yticks([])

year = time.strftime("%Y")
major, minor, micro = matplotlib.__version__.split('.', 2)
text = ("The Mandelbrot fractal set\n"
        "Rendered with matplotlib %s.%s, %s - http://matplotlib.org"
        % (major, minor, year))
ax.text(xmin+.025, ymin+.025, text, color="white", fontsize=12, alpha=0.5)
```
